##### Recall: Basic Rules for Integration

1. $\int kf(x) \, dx = k \int f(x) dx$
2. $\int [f(x) \pm g(x)] \, dx = \int f(x) \, dx \pm \int g(x) \, dx$
3. $\int dx = x + C$
4. $\int x^n \, dx = \frac{x^{n+1}}{n+1} + C, n \ne -1$
5. $\displaystyle \int \cfrac{dx}{x} = \int \cfrac{1}{x} \, dx = \ln{|x|} + C$
6. $\int e^{x} \, dx = e^x + C$
7. $\int a^x \, dx = \left( \cfrac{1}{\ln{a}} \right) a^x + C$
8. $\int \sin{x} \, dx = - \cos{x} + C$
9. $\int \cos{x} \, dx = \sin{x} + C$
10. $\int \tan{x} \, dx = -\ln{|\cos{x}|} + C$
11. $\int \cot{x} \, dx = \ln{|\sin{x}|} + C$
12. $\int \sec{u} \, dx = \ln{|\sec{x} + \tan{x}|} + C$
13. $\int \csc{x} \, dx = - \ln{|\csc{x} + \cot{x}|} + C$
14. $\int \sec^2{x} \, dx = \tan{x} + C$
15. $\int \csc^2{x} \, dx = - \cot{x} + C$
16. $\int \sec{x} \tan{x} \, dx = \sec{x} + C$
17. $\int \csc{x} \cot{x} \, dx = - \csc{x} + C$
18. $\displaystyle \int \cfrac{dx}{\sqrt{a^2 - x^2}} = \arcsin{\left( \cfrac{x}{a} \right)} + C$
19. $\displaystyle \int \cfrac{dx}{a^2 + x^2} = \cfrac{1}{a} \arctan{\left( \cfrac{x}{a} \right)} + C$
20. $\displaystyle \int \cfrac{dx}{x \sqrt{x^2 - a^2}} = \cfrac{1}{a} \text{arcsec} \, {\left( \cfrac{|x|}{a} \right)} + C$

##### Common Integration Techniques:

1. ##### Expansion (multiply out)
     * Example: \begin{align} (1 + e^x)^2 = 1 + 2e^x + e^{2x} \end{align}
2. ##### Separation of numerator(s)
    * Example: \begin{align} \cfrac{1 + x}{x^2 + 1} = \cfrac{1}{x^2 + 1} + \cfrac{x}{x^2 + 1} \end{align}
3. ##### Completing the square
    * Example: \begin{align} \cfrac{1}{\sqrt{2x - x^2}} = \cfrac{1}{\sqrt{1 - (x-1)^2}} \end{align}
4. ##### Dividing improper rational function(s)
    * Example: \begin{align} \cfrac{x^2}{x^2 + 1} = 1 - \cfrac{1}{x^2 + 1} \end{align}
5. ##### Adding/subtracting terms that add to zero
    * Example: \begin{align} \cfrac{2x}{x^2 + 2x + 1} = \cfrac{2x + 2 - 2}{x^2 + 2x + 1} = \cfrac{2x+2}{x^2 + 2x + 1} - \cfrac{2}{x^2 + 2x + 1} \end{align}
6. ##### Using trigonometric identities
    * Example: \begin{align} \cot^2{x} = \csc^2{x} - 1 \end{align}
7. ##### Multiplying and dividing by Pythagorean conjugate
    * Example: \begin{align} \cfrac{1}{1 + \sin{x}} & = \left( \cfrac{1}{1+\sin{x}} \right) \left( \cfrac{1 - \sin{x}}{1 - \sin{x}} \right) \\ & = \cfrac{1 - \sin{x}}{1 - \sin^2{x}} \\ & = \cfrac{1 - \sin{x}}{\cos^2{x}} \\ & = \sec^2{x} - \cfrac{\sin{x}}{\cos^2{x}} \end{align}

##### Example 1:

Find the indefinite integrals:

a. $ \displaystyle \int \cfrac{4}{x^2 + 9} \, dx$

b. $ \displaystyle \int \cfrac{4x}{x^2 + 9} \, dx$

c. $ \displaystyle \int \cfrac{4x^2}{x^2 + 9} \, dx$

a. Find the indefinite integral: $ \displaystyle \int \cfrac{4}{x^2 + 9} \, dx.$

##### Solution:

a.

\begin{align}
\displaystyle \int \cfrac{4}{x^2 + 9} \, dx & = 4 \int \cfrac{1}{x^2 + 9} \, dx & \text{(from 1 of basic integration rules above)} \\
& = 4 \int \cfrac{1}{x^2 + 3^2} \, dx = 4 \int \cfrac{1}{3^2 + x^2} \, dx & \text{(rewrite)} \\
& = \boxed{\cfrac{4}{3} \arctan{\left( \cfrac{x}{3} \right)} + C} & \text{(from 19 of basic integration rules above)}
\end{align}

b. Find the indefinite integral: $ \displaystyle \int \cfrac{4x}{x^2 + 9} \, dx.$

##### Solution:

b.

\begin{align}
\displaystyle \int \cfrac{4x}{x^2 + 9} \, dx & = 4 \int \cfrac{x}{x^2 + 9} \, dx & \text{(from 1 of basic integration rules above)}
\end{align}

Now let $u=x^2+9$, then $du = 2x \, dx \Rightarrow \cfrac{1}{2} du = x \, dx$.

Rearranging and substituting, we get

\begin{align}
\displaystyle \int \cfrac{4x}{x^2 + 9} \, dx & = 4 \int \cfrac{1}{x^2 + 9} \cdot x \, dx & \text{(rearranging)}\\
& = 4 \int \cfrac{1}{u} \cdot \cfrac{1}{2} \, du & \text{(substituing )} \\
& = \cfrac{4}{2} \int \cfrac{1}{u} \, du & \text{(factor $1/2$)} \\
& = 2 \int \cfrac{1}{u} \, du & \text{(simplify)} \\
& = 2 \ln{|u|} + C & \text{(from 5 of basic integration rules)} \\
& = \boxed{2 \ln{(x^2 + 9)} + C} & \text{(back substitute; notice no absolute value)}
\end{align}

c. Find the indefinite integral: $ \displaystyle \int \cfrac{4x^2}{x^2 + 9} \, dx.$

##### Solution:

c.

\begin{align}
\displaystyle \int \cfrac{4x^2}{x^2 + 9} \, dx & = 4 \int \cfrac{x^2}{x^2 + 9} \, dx & \text{(factor out the $4$)}
\end{align}

\begin{align}
\displaystyle \int \cfrac{4x^2}{x^2 + 9} \, dx & = 4 \int \cfrac{x^2 + 9 - 9}{x^2 + 9} \, dx & \text{(add and subtract 9; i.e. add zero)} \\
& = 4 \int \left( \cfrac{x^2 + 9}{x^2 + 9} - \cfrac{9}{x^2+9} \right) \, dx & \text{(separate numerator)} \\
& = 4 \int \cfrac{x^2 + 9}{x^2 + 9} \, dx - 4 \int \cfrac{9}{x^2+9} \, dx & \text{(from 2 of basic integration rules)} \\
& = 4 \int \, dx - 4 \cdot 9 \int \cfrac{1}{x^2 + 9} \, dx & \text{(simplify and factor 9)} \\
& = \int \, dx - 36 \int \cfrac{1}{3^2 + x^2} \, dx & \text{(rewrite and simplify)} \\
& = 4x - 36 \cdot \cfrac{1}{3} \arctan{\left( \cfrac{x}{3} \right)} + C & \text{(integrate; i.e. use 3 and 19 from rules)} \\
& = \boxed{4x - 12 \arctan{\left( \cfrac{x}{3} \right)} + C} & \text{(simplify)}
\end{align}

##### Example 2:

Find the definite integral: $\displaystyle \int_0^1 \cfrac{x+3}{\sqrt{4 - x^2}} \, dx.$

##### Solution:

\begin{align}
\displaystyle \int_0^1 \cfrac{x+3}{\sqrt{4 - x^2}} \, dx & = \int_0^1 \cfrac{x}{\sqrt{4 - x^2}} \, dx + \int_0^1 \cfrac{3}{\sqrt{4 - x^2}} \, dx & \text{(separate numerator)}
\end{align}

\begin{align}
\displaystyle \int_0^1 \cfrac{x+3}{\sqrt{4-x^2}} \, dx & = \underbrace{\int_0^1 \cfrac{x}{\sqrt{4-x^2}} \, dx}_{I_1} + \underbrace{\int_0^1 \cfrac{3}{\sqrt{4 - x^2}} \, dx}_{I_2}
\end{align}

To solve $I_1$ we can use a $u$ substitution:

Let $u = 4 - x^2$. Then $du = -2x \, dx$. Therefore, $-\cfrac{1}{2} \, du = x \, dx$.

Then

\begin{align}
\displaystyle I_1 & = \int_0^1 \cfrac{1}{\sqrt{4 - x^2}} x \, dx \\
& = \int_{x=0}^{x=1} \cfrac{1}{\sqrt{u}} \cdot \left( - \cfrac{1}{2} \right) \, du \\
& = - \cfrac{1}{2} \int_{x=0}^{x=1} \cfrac{1}{u^{1/2}} \, du
\end{align}

We could continue in this way, however, mixing $x$'s and $u$'s can get very confusing and is generally not the best practice...

With a definite integral where we have substituted, as a best practice, we will also substitute the limits of integration:

Since $u = 4 - x^2$, then if $x = 0 \Rightarrow u = 4$ and if $x = 1 \Rightarrow u = 4 - 1^2 = 3$. Therefore, the lower limit of integration $x=0$ becomes $u = 4$ and the upper limit of integration $x=1$ becomes $u = 3$.

\begin{align}
I_1 & = - \cfrac{1}{2} \int_4^3 u^{-1/2} \, du
\end{align}

Since $3 < 4$, we generally prefer the lower limit of integration to be less than the upper limit of integration. So we need to switch the order. Recall that $\displaystyle \int_b^a f(x) \, dx = - \int_a^b f(x) \, dx$. Then we have

\begin{align}
I_1 & = \cfrac{1}{2} \int_3^4 u^{-1/2} \, du
\end{align}

Now integrating (using power rule):

\begin{align}
I_1 & = \cfrac{1}{2} \cfrac{u^{1/2}}{1/2} \Big|_3^4 \\
& = \cfrac{1}{2} 2 u^{1/2} \Big|_3^4 \\
& = u^{1/2} \Big|_3^4 \\
& = \sqrt{4} - \sqrt{3} \\
& = 2 - \sqrt{3}
\end{align}

This completes the computation for $I_1$ only. We still need $I_2$.

So we need to solve $I_2 = \displaystyle \int_0^1 \cfrac{3}{\sqrt{4 - x^2}} \, dx$.

This is straightforward from 18 of the basic integration rules listed before:

\begin{align}
\displaystyle I_2 & = 3 \int_0^1 \cfrac{1}{\sqrt{4 - x^2}} \, dx \\
& = 3 \arcsin{ \left( \cfrac{x}{2} \right)} \Big|_0^1 \\
& = 3 \arcsin{(1/2)} - 3 \arcsin{(0)} \\
& = 3 \left( \cfrac{\pi}{6} \right) \\
& = \cfrac{\pi}{2}
\end{align}

Therefore, our final answer is

\begin{align}
\displaystyle \int_0^1 \cfrac{x+3}{\sqrt{4 - x^2}} \, dx& = I_1 + I_2 \\
& = \boxed{2 - \sqrt{3} + \cfrac{\pi}{2}}
\end{align}

##### Example 3:

Find the indefinite integral: $\displaystyle \int \cfrac{x^2}{\sqrt{16 - x^6}} \, dx.$

##### Solution:

First, we'll rewrite:

\begin{align}
\displaystyle \int \cfrac{x^2}{\sqrt{16 - x^6}} \, dx & = \int \cfrac{x^2}{\sqrt{16 - (x^3)^2}} \, dx.
\end{align}

Next, we use a $u$ substitiion. Let $u = x^3$. Then $du = 3 x^2 \, dx \Rightarrow 1/3 \, du = x^2 \, dx$.

\begin{align}
\displaystyle \int \cfrac{x^2}{\sqrt{16 - x^6}} \, dx & = \int \cfrac{1}{\sqrt{16 - u^2}} \cdot 1/3 \, du \\
& = 1/3 \int \cfrac{1}{\sqrt{16 - u^2}} \, du \\
& = 1/3 \arcsin{\left( \cfrac{u}{4} \right)} + C \\
& = \boxed{1/3 \arcsin{\left( \cfrac{x^3}{4} \right)} + C}
\end{align}

##### Example 4:

Find the indefinite integral: $\displaystyle \int \cfrac{1}{1 + e^{x}} \, dx.$

##### Solution:

One possible approach is to add zero in the numerator (add and subtract $e^x$):

\begin{align}
\displaystyle \int \cfrac{1}{1 + e^{x}} \, dx & =  \int \cfrac{1 + e^x - e^x}{1 + e^x} \, dx
\end{align}

Now separate the numerator and simplify:

\begin{align}
\displaystyle \int \cfrac{1}{1 + e^{x}} \, dx & =  \int \left( \cfrac{1 + e^x}{1 + e^x} - \cfrac{e^x}{1 + e^x} \right) \, dx \\
& = \int \left( 1 - \cfrac{e^x}{1 + e^x} \right) \, dx \\
& = \int dx - \int \cfrac{e^x}{1 + e^x} \, dx
\end{align}

The first integral is straightforward; for the second integral we use a $u$ substitution. For the second integral, let $u = 1 + e^x$. Then $du = e^x \, dx$.

Substituting we have

\begin{align}
\displaystyle \int \cfrac{1}{1+e^x} \, dx & = x - \int \cfrac{1}{1 + e^x} \cdot e^x \, dx \\
& = x - \int \cfrac{1}{u} \, du \\
& = x - \ln{|u|} + C \\
& = \boxed{x - \ln{\left( 1 + e^x \right)} + C}
\end{align}

##### Example 5:

Find the indefinite integral: $\displaystyle \int \cot{x} \ln{(\sin{x})} \, dx.$

##### Solution:

Perhaps a $u$ substitution?

Let $u = \ln{(\sin{x})}$. Then $du = \cfrac{1}{\sin{x}} \cdot \cos{x} \, dx$ (from *chain rule*).

Note that $du = \cfrac{\cos{x}}{\sin{x}} \, dx = \cot{x} \, dx$.

Therefore, if $u = \ln{(\sin{x})}$, then $du = \cot{x} \, dx$.

So we have

\begin{align}
\displaystyle \int \cot{x} \ln{(\sin{x})} \, dx & = \int \ln{\sin{x}} \cdot \cot{x} \, dx \\
& = \int u \, du \\
& = u^2 / 2 + C \\
& = \boxed{\cfrac{[\ln(\sin{x})]^2}{2} + C}
\end{align}

##### Example 6:

Find the indefinite integral: $\displaystyle \int \tan^2{(2x)} \, dx.$

##### Solution:

First, we use a simple $u$ subsitution for the argument of tangent. Let $u = 2x$. Then $du = 2 \, dx \Rightarrow \cfrac{1}{2} \, du = dx$.

This yields

\begin{align}
\displaystyle \int \tan^2{(2x)} \, dx & = \int \tan^2{u} \cdot \cfrac{1}{2} \, du \\
& = \cfrac{1}{2} \int \tan^2{u} \, du
\end{align}

Recall that $\tan^2{u} = \sec^2{u} - 1$. So we get

\begin{align}
\displaystyle \cfrac{1}{2} \int \tan^2{u} \, dx & = \cfrac{1}{2} \int (\sec^2{u} - 1) \, dx
\end{align}

From basic integration rules (number 14):

\begin{align}
\displaystyle \cfrac{1}{2} \int \tan^2{u} \, dx & = \cfrac{1}{2} \left( \tan{u} - u \right) + C \\
& = \cfrac{1}{2} \left( \tan{(2x)} - 2x \right) + C \\
& = \boxed{\cfrac{1}{2}\tan{(2x)} - x + C \\}
\end{align}
