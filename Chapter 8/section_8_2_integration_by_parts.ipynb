##### Motivating Examples

Find the indefinite integrals of each of the following.

**(a)** $\displaystyle \int \cfrac{e^{1/t}}{t^2} \, dt$

**(b)** $\displaystyle \int x \sqrt{x - 1} \, dx$

**(c)** $\displaystyle \int \cfrac{\left( \ln{x} \right)^2}{x} \, dx$

**(d)** $\displaystyle \int x \ln{x} \, dx$

**(e)** $\displaystyle \int x e^x \, dx$

**(f)** $\displaystyle \int e^{2x} \cos{2x} \, dx$

For **(a)** through **(c)** we can use some sort of $u$-sub. For **(d)** through **(f)** things get a bit trickier...

Something we may notice in parts **(d)** through **(f)** is theres some sort of product involved. Even though we don't have a produce rule for integration, a common technique that helps with integrating function products is *integration by parts.*

##### Derivation of Integration by Parts

Let $u$ and $v$ be functions of $x$. From the **product rule** for differentiation we know

$$\frac{d}{dx} \left( u \cdot v \right) = u' v + v' u.$$

Written another way:

$$\frac{d}{dx} \left( u \cdot v \right) = \frac{du}{dx} v + \frac{dv}{dx} u.$$

Multiplying both sides by $dx$ (treating $dx$ as a differential):

$$d(u \cdot v) = v \, du + u \, dv$$

Now integrating both sides of the previous equation:

$$\int d(u \cdot v) = \int v \, du + \int u \, dv$$

Subtracting from both sides:

$$\int u \, dv = \int d(u \cdot v) - \int v \, du$$

Integrating the left integral on the left-hand side:

$$\boxed{\int u \, dv = u \cdot v - \int v \, du}$$

Note that we cannot directly integrate the remaining integrals since we do not know how $u$ and $v$ are defined relative to each other. The above boxed formula is known as **integration by parts (IBP).**

##### Theorem: Integration by Parts

Let $u$ and $v$ be continuously differentiable functions of $x$. Then

$$\int u \, dv = u \cdot v - \int v \, du.$$

##### Example 1

Find $\int x \ln{x} \, dx.$

##### Solution:

Let $u = \ln{x}$ and $dv = x \, dx$. Then $du = \frac{1}{x} \, dx$ and $v = \int x \, dx = x^2/2$.

Note that we temporarily leave off the constant of integration $C$ as the constant of integration will factor in at a later time.

Rearranging and substituting:

$$\int \underbrace{\left( \ln{x} \right)}_{u} \overbrace{x \, dx}^{dv} = \underbrace{u \cdot v - \int v \, du}_{\text{by IBP}}$$

So we end up getting

\begin{align}
\int x \ln{x} \, dx & = \underbrace{(\ln{x})}_{u} \underbrace{\left( \frac{x^2}{2} \right)}_{v} - \int \underbrace{\left( \frac{x^2}{2} \right)}_{v} \underbrace{\left(\frac{1}{x} \, dx \right)}_{du} \\
& = \frac{x^2}{2} \ln{x} - \int \frac{x}{2} \, dx \\
& = \boxed{\frac{x^2}{2} \ln{x} - \frac{x^2}{4} + C}
\end{align}

##### Example 2

Find $\int x e^x \, dx$.

##### Solution:

Let $u = x$ and $dv = e^x \, dx$. Then $du = dx$ and $v = \int e^x \, dx = e^x$.

Rearranging and substituting:

$$\int \underbrace{\left( x \right)}_{u} \overbrace{e^x \, dx}^{dv} = \underbrace{u \cdot v - \int v \, du}_{\text{by IBP}}$$

So we end up getting

\begin{align}
\int x e^x \, dx & = \underbrace{x}_{u} \underbrace{ e^x }_{v} - \int \underbrace{ e^x }_{v} \underbrace{ dx }_{du} \\
& = \boxed{x e^x - e^x + C}
\end{align}

So how do we know what to substitute as $u$ or $dv$?

##### Guidelines for Choice of $u$

Consider the mnemonic **LIATE**:

* **L**ogarithm
* **I**nverse trig
* **A**lgebraic
* **T**rig
* **E**xponentional

If there are any functions/parts of the integral that have the above (in order), try letting $u$ equal that part. That is, if there is a logarithmic part of the integrand, try letting $u$ be the logarithmic part before any others and etc.

*This is only a guideline and not a rule. In general, try letting $dv$ be a reasonably integrable function with known methods and/or identities; then $u$ will be the leftovers. Sometimes a clever choice of $u$ and $dv$ is needed.*

##### Example 3

Let's try a fun one!

Find $\int e^{2x} \cos{(2x)} \, dx$

##### Solution:

Before we even consider IBP, let's first perform a simple $u$ substitution. The $2x$ in the integrand can be easily taken care of by performing such a substitution.

Let $u = 2x$. Then $du = 2 \, dx \Rightarrow dx = \frac{1}{2} \, du$.

Substituting and factoring:

$$\int e^{2x} \cos{(2x)} \, dx = \frac{1}{2} \int e^u \cos{u} \, du$$

Now we're ready for IBP. But we've already used $u$! So we'll either have to use a different variable or change the current variable of integration.

Let's just go ahead and change the variable of integration (basically just let $u = t$):

$$\frac{1}{2} \int e^u \cos{u} \, du = \frac{1}{2} \int e^t \cos{t} \, dt$$

Alright, so now what? IBP? What should we use for our substitutions?

Let's use **LIATE**:

There's no **L**ogarithmic part; no **I**nverse trig; no **A**lgebraic part... however, there is a **T**rig function in there.

So let's let $u = \cos{t}$ and $dv = e^t \, dt$. Then $du = -\sin{t} \, dt$ and $v = e^t$.

Rearranging and substituting:

$$\frac{1}{2} \int \underbrace{(\cos{t})}_{u} \overbrace{e^t \, dt}^{dv} = \frac{1}{2} \left( u \cdot v - \int v \, du \right)$$

So now we have

$$\frac{1}{2} \int e^t \cos{t} \, dt = \frac{1}{2} \left( e^t\cos{t} - \int \underbrace{e^t}_{v} \underbrace{(-\sin{t}) \, dt}_{du} \right)$$

Simplifying:

\begin{align}
\frac{1}{2} \int e^t \cos{t} \, dt & = \frac{1}{2} e^t\cos{t} - \frac{1}{2} \int e^t (-\sin{t}) \, dt \\
& = \frac{1}{2} e^t \cos{t} + \frac{1}{2} \int e^t \sin{t} \, dt
\end{align}

Okay, so we have another integral on the right-hand side that looks like it might be a candidate for integration by parts... so let's do it again!

Let $u = \sin{t}$ and $dv = e^t \, dt$. Then $du = \cos{t} \, dt$ and $v = e^t$.

Again, we rearrange and substitute:

$$\frac{1}{2} \int e^t \cos{t} \, dt = \frac{1}{2} e^t \cos{t} + \frac{1}{2} \int \underbrace{(\sin{t})}_{u} \underbrace{e^t \, dt}_{v} = \frac{1}{2} e^t \cos{t} + \frac{1}{2} \left( u \cdot v - \int v \, du \right)$$

So now we have

\begin{align}
\frac{1}{2} \int e^t \cos{t} \, dt & = \frac{1}{2} e^t \cos{t} + \frac{1}{2} \left(  e^t \sin{t} - \int e^t \cos{t} \, dt \right) \\
& = \frac{1}{2} e^t \cos{t} + \frac{1}{2} e^t \sin{t} - \frac{1}{2} \int e^t \cos{t} \, dt
\end{align}

So now what?

Notice:

$$\underline{\frac{1}{2} \int e^t \cos{t} \, dt} = \frac{1}{2} e^t \cos{t} + \frac{1}{2} e^t \sin{t} - \underline{\frac{1}{2} \int e^t \cos{t} \, dt}$$

Since this is an equation, let's treat the two underlined terms as a variable to be solved for and add the underlined term on the right-hand side to both sides:

$$\int e^t \cos{t} \, dt = \frac{1}{2} e^t \cos{t} + \frac{1}{2} e^t \sin{t}$$

BUT WAIT! We need our constant of integration:

$$\int e^t \cos{t} \, dt = \frac{1}{2} e^t \cos{t} + \frac{1}{2} e^t \sin{t} + C$$

As much as we might like to think we are done... we're not. Afterall, we did start with $x$'s and we currently have $t$'s.

Back substituting:

\begin{align}
\int e^{2x} \cos{2x} \, dx & = \frac{1}{2} \int e^t \cos{t} \, dt \\
& \Downarrow \\
\int e^{2x} \cos{2x} \, dx & = \frac{1}{2} \left( \frac{1}{2} e^t \cos{t} + \frac{1}{2} e^t \sin{t} + C \right) \\
& \Downarrow \text{(recall that $t = 2x$)} \\
\int e^{2x} \cos{(2x)} \, dx & = \boxed{\frac{1}{4} e^{2x} \cos{(2x)} + \frac{1}{4} e^{2x} \sin{(2x)} + C}
\end{align}

See!? Told you it'd be fun!

##### Example 4

Find $\int \ln{x} \, dx$

##### Solution:

We don't have a formula for this integral.

A direct $u$ substitution won't work since if $u = \ln{u}$, then $du = 1/x$ and we don't have $1/x$ anywhere in the integral.

IBP?

Certainly! But what do we substitute for $u$ and $dv$?

We really onlt have one or two parts:

$\ln{x} \, dx$ taken as a whole part or $\ln{x}$ and $dx$ as two separate parts.

So we'll let $u = \ln{x}$ and $dv = dx$. This means that $du = 1/x \, dx$ and $v = x$.

Substituting:

\begin{align}
\int \underbrace{\ln{x}}_{u} \, \underbrace{dx}_{dv} & = uv - \int v \, du \\
& = x \ln{x} - \int x \cdot 1/x \, dx \\
& = x \ln{x} - \int dx \\
& = \boxed{x \ln{x} - x + C}
\end{align}

##### Example 5

Let's try a definite integral.

Find $\displaystyle \int_{0}^{1} \arcsin{x} \, dx = \int_{0}^{1} \sin^{-1}{x} \, dx$

Recall: $\arcsin{x} = \sin^{-1}{x}$

##### Solution:

We will approach this integral in a similar way as the previous example:

Let $u = \sin^{-1}{x}$ and $dv = dx$. Then $du = \cfrac{1}{\sqrt{1 - x^2}} \, dx$ and $v = x$.

Substituting:

\begin{align}
\int_{0}^{1} \underbrace{\sin^{-1}{x}}_{u} \, \underbrace{dx}_{dv} & = uv \Big|_{x=0}^{x=1} - \int_{x=0}^{x=1} v \, du \\
& = (x \sin^{-1}{x}) \Big|_0^1 - \int_0^1 x \cdot \cfrac{1}{\sqrt{1-x^2}} \, dx \\
& = (1 \cdot \sin^{-1}(1) - 0) - \int_0^1 x \cdot \cfrac{1}{\sqrt{1-x^2}} \, dx \\
& = \cfrac{\pi}{2} - \int_0^1 x \cdot \cfrac{1}{\sqrt{1-x^2}} \, dx
\end{align}

To handle the integral that remains we can use a simple substitution:

Let $t = 1-x^2$. Then $dt = -2x \, dx$.

Rearranging and substituting:

\begin{align}
\cfrac{\pi}{2} - \int_0^1 x \cdot \cfrac{1}{\sqrt{1-x^2}} \, dx & = \cfrac{\pi}{2} - \int_0^1 \cfrac{1}{\underbrace{\sqrt{1-x^2}}_{\text{denominator is }\sqrt{t}}} \cdot \underbrace{x \,dx}_{-\frac{1}{2}dt} \\
& = \cfrac{\pi}{2} - \int_0^1 \cfrac{1}{\underbrace{\sqrt{1-x^2}}_{\text{denominator is }\sqrt{t}}} \cdot \underbrace{x \,dx}_{-\frac{1}{2}dt} \\
& = \cfrac{\pi}{2} + \cfrac{1}{2} \int_1^0 \cfrac{1}{\sqrt{t}} \, dt \\
& = \cfrac{\pi}{2} + \cfrac{1}{2} \int_1^0 t^{-1/2} \, dt \\
& = \cfrac{\pi}{2} - \cfrac{1}{2} \int_0^1 t^{-1/2} \, dt \\
& = \cfrac{\pi}{2} - \cfrac{1}{2} \left( \cfrac{t^{1/2}}{1/2} \right) \Big|_0^1 \\
& = \boxed{\cfrac{\pi}{2} - 1}
\end{align}

##### Tabular Method

The tabular method for IBP is a convenient method *provided the choice for $u$ has a* **terminating derivative.** By terminating derivative I mean that repeated derivatives of $u$ is *eventually* zero.

##### Example 6

Find $\int x^4 e^x \, dx$.

##### Solution:

We first let $u = x^4$ and $dv = e^x \, dx$. Then the fifth derivative of $u$ is *zero* (i.e. $u$ has a terminating derivative). So we use the tabular method for IBP:

|$+$|$\overbrace{u = x^4}^{\text{This column find derivatives}}$|$\overbrace{dv = e^x \, dx}^{\text{This column find integrals}}$|
|:-:|:-:|:-:|
|$-$|$4x^3$|$e^x$|
|$+$|$12x^2$|$e^x$|
|$-$|$24x$|$e^x$|
|$+$|$24$|$e^x$|
|$-$|$0$|$e^x$|

Once you obtain a $0$ for your derivatives column stop differentiating/integrating. Then multiply *diagonally* left-to-right. The alternating signs at the far left are multiples of $1$ for $+$ and multiples of $-1$ for $-$ to be applied to the middle column (derivatives column) *before* multiplying diagonally.

So we end up with

$$\int x^4 e^x \, dx = \boxed{x^4e^x-4x^3 e^x + 12x^2 e^x - 24xe^x + 24 e^x + C}$$

The alternative to this is performing IBP four times... your choice. Choose wisely.

##### Example 7

Find $\int x^3 \cos{(3x)} \, dx.$

##### Solution:

Let $u = x^3$ and $dv = \cos{(3x)} \, dx$. Then $u$ has a terminating derivative. So we can use IBP. Now before we proceed with IBP, let's first note we need to integrate $v = \int \cos{(3x)} \, dx$. So we can use a simple substitution: let $t = 3x$, then $dt = 1/3 \, dx \Rightarrow dx = 1/3 \, dt$. This means we'll be integrating over $\cos{t}$ or $\sin{t}$ with a factor of $1/3$ repeatedly.

So the tabular method for IBP gives us

|$+$|$\overbrace{u = x^3}^{\text{This column find derivatives}}$|$\overbrace{dv = \cos{(3x)} \, dx}^{\text{This column find integrals}}$|
|:-:|:-:|:-:|
|$-$|$3x^2$|$1/3 \sin{(3x)}$|
|$+$|$6x$|$- 1/9 \cos{(3x)}$|
|$-$|$6$|$-1/27 \sin{(3x)}$|
|$+$|$0$|$1/81 \cos{(3x)}$|

So we get

\begin{align}
\int x^3 \cos{(3x)} \, dx & = x^3 \left(1/3 \sin{(3x)} \right) + (-3x^3) \left(- 1/9 \cos{(3x)} \right) + (6 x) \left(- 1/27 \sin{(3x)} \right) + (-6) \left(1/81 \cos{(3x)} \right) + C \\
& = \boxed{1/3 x^3 \sin{(3x)} + 1/3 x^3 \cos{(3x)} - 2/9 x \sin{(3x)} - 2/27 \cos{(3x)} + C}
\end{align}
