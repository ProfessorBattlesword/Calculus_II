##### Objective 1: Solve trigonometric integrals involving powers of sine and cosine

* In this section we will investigate integrals involving various trigonometric forms

* We first consider integrals of the form $$\int \sin^{m}{x} \cos^{n}{x} \, dx$$ where either $m$ or $n$ are *positive* integers
* Examples to consider:
    * **(a)** $\int \sin^{4}{x} \cos^{3}{x} \, dx$
    * **(b)** $\int \sin^{3}{x} \cos^{4}{x} \, dx$
    * **(c)** $\int \cos^{5}{x} \sin^{4}{x} \, dx$
    * **(d)** $\int \sin^{4}{x} \, dx$

##### Recall Common Trig Identities Involving Sines and Cosines

* Pythagorean identities:
    * $\sin^{2}{x} + \cos^{2}{x} = 1$
    * $\sin^{2}{x} = 1 - \cos^{2}{x}$
    * $\cos^{2}{x} = 1 - \sin^{2}{x}$

* Power-reducing formulas:
    * $\sin^{2}{x} = \cfrac{1 - \cos{(2x)}}{2}$
    * $\cos^{2}{x} = \cfrac{1 + \cos{(2x)}}{2}$

##### Guidelines for Evaluating Integrals Involving Powers of Sine and Cosine

1. For *positive, odd powers of sine*, save one factor of sine for substitution and convert the remaining sines to cosines. Then expand and integrate.

2. For *positive, odd powers of cosine*, save one factor of cosine for substitution and convert the remaining cosines to sines. Then expand and integrate.

3. For *both powers of sine and cosine that are nonnegative and even*, make repeated use of the identities $\sin^{2}{x} = \cfrac{1 - \cos{(2x)}}{2}$ and $\cos^{2}{x} = \cfrac{1 + \cos{(2x)}}{2}$ as needed to convert the integrand to odd powers of cosine. Then use guideline 2 above.

##### Example 1

Find $\int \sin^{4}{x} \cos^{3}{x} \, dx$.

##### Solution:

Since the cosine part has an odd and positive power, we break off one factor of cosine and save for substitution:

$$\int \sin^{4}{x} \cos^{3}{x} \, dx = \int \sin^{4}{x} \cos^{2}{x} \cos{x} \, dx$$

Making use of a Pythagorean identity:

$$\int \sin^{4}{x} \cos^{3}{x} \, dx = \int \sin^{4}{x} (1 - \sin^{2}{x}) \cos{x} \, dx$$

Distributing:

$$\int \sin^{4}{x} \cos^{3}{x} \, dx = \int (\sin^{4}{x} - \sin^{6}{x}) \cos{x} \, dx$$

Now let $u = \sin{x}$. Then $du = \cos{x} \, dx$.

Substituting:

\begin{align}
\int \sin^{4}{x} \cos^{3}{x} \, dx & = \int (\underbrace{\sin^{4}{x}}_{u^4} - \underbrace{\sin^{6}{x}}_{u^6}) \underbrace{\cos{x} \, dx}_{du} \\
& = \int u^4 - u^6 \, du
\end{align}

And now we integrate:

\begin{align}
\int \sin^{4}{x} \cos^{3}{x} \, dx & = \cfrac{u^5}{5} - \cfrac{u^7}{7} + C \\
& = \boxed{\cfrac{\sin^{5}{x}}{5} - \cfrac{\sin^{7}{x}}{7} + C}
\end{align}

##### Example 2

Find $\int \sin^{3}{x} \cos^{4}{x} \, dx$.

##### Solution:

Since now the sine part has an odd and positive power, we break off one factor of sine and save for substitution:

$$\int \sin^{3}{x} \cos^{4}{x} \, dx = \int \sin^{2}{x} \cos^{4}{x} \sin{x} \, dx$$

Pythagorean identity:

$$\int \sin^{2}{x} \cos^{4}{x} \sin{x} \, dx = \int (1 - \cos^{2}{x}) \cos^{4}{x} \sin{x} \, dx$$

Let $u = \cos{x}$. Then $du = - \sin{x} \, dx \Rightarrow - du = \sin{x} \, dx$.

Substituting:

$$\int (1 - \underbrace{\cos^{2}{x}}_{u^2}) \underbrace{\cos^{4}{x}}_{u^4} \underbrace{\sin{x} \, dx}_{-du} = - \int (1 - u^2) u^4 \, du$$

Distributing and integrating:

\begin{align}
\int \sin^{3}{x} \cos^{4}{x} \, dx & = - \int u^4 - u^6 \, du \\
& = - \left( \cfrac{u^5}{5} - \cfrac{u^7}{7} \right) + C \\
& = \cfrac{u^7}{7} - \cfrac{u^5}{5} + C \\
& = \boxed{\cfrac{\cos^{7}{x}}{7} - \cfrac{\cos^{5}{x}}{5} + C}
\end{align}

##### Example 3

Find $\int \cos^{5}{x} \sin^{4}{x} \, dx$.

##### Solution:

The power on cosine is odd and positive, so we break off a factor of cosine and save it for substitution:

$$\int \cos^{5}{x} \sin^{4}{x} \, dx = \int \cos^{4}{x} \sin^{4}{x} \cos{x} \, dx$$

Rewrite our cosine term:

$$\int \cos^{4}{x} \sin^{4}{x} \cos{x} \, dx = \int \left( \cos^{2}{x} \right)^2 \sin^{4}{x} \cos{x} \, dx$$

Pythagorean identity:

$$\int \left( \cos^{2}{x} \right)^2 \sin^{4}{x} \cos{x} \, dx = \int \left( 1 - \sin^{2}{x} \right)^2 \sin^{4}{x} \cos{x} \, dx$$

Let $u = \sin{x}$. Then $du = \cos{x} \, dx$.

Substituting:

$$\int \left( 1 - \underbrace{\sin^{2}{x}}_{u^2} \right)^2 \underbrace{\sin^{4}{x}}_{u^4} \underbrace{\cos{x} \, dx}_{du} = \int (1-u^2)^2 u^4 \, du$$

Expanding and distributing:

\begin{align}
\int (1-u^2)^2 u^4 \, du & = \int (1 - 2u^2 + u^4) u^4 \, du \\
& = \int u^4 - 2u^6 + u^8 \, du
\end{align}

Integrating:

\begin{align}
\int \cos^{5}{x} \sin^{4}{x} \, dx & = \int u^4 - 2u^6 + u^8 \, du \\
& = \cfrac{u^5}{5} - 2 \cfrac{u^7}{7} + \cfrac{u^9}{9} + C \\
& = \boxed{\cfrac{\sin^{5}{x}}{5} - 2 \cfrac{\sin^{7}{x}}{7} + \cfrac{\sin^{9}{x}}{9} + C}
\end{align}

##### Example 4

Find $\int \sin^{4}{x} \, dx$.

##### Solution:

We only have a sine portion and the power is even and nonnegative, so we will need rewrite in terms of $\sin^{2}{x}$ to use the power-reducing formula for sine and go from there:

$$\int \sin^{4}{x} \, dx = \int \left( \sin^{2}{x} \right)^2 \, dx$$

Power-reducing formula for sine:

$$\int \left( \sin^{2}{x} \right)^2 \, dx = \int \left( \cfrac{1 - \cos{(2x)}}{2} \right)^2 \, dx$$

Simplifying:

\begin{align}
\int \left( \cfrac{1 - \cos{(2x)}}{2} \right)^2 \, dx & = \frac{1}{4} \int \left( 1 - \cos{(2x)} \right)^2 \, dx \\
& = \frac{1}{4} \int \left( 1 - 2 \cos{(2x)} + \cos^{2}{(2x)} \right) \, dx
\end{align}

Again, we apply power-reducing formula, but this time for cosine:

\begin{align}
\frac{1}{4} \int 1 - 2 \cos{(2x)} + \underbrace{\cos^{2}{(2x)}}_{\frac{1 + \cos{(4x)}}{2}} \, dx \\ = \frac{1}{4} \int \left( 1 - 2 \cos{(2x)} + \cfrac{1 + \cos{(4x)}}{2} \right) \, dx
\end{align}

Simplifying:

\begin{align}
\frac{1}{4} \int \left( 1 - 2 \cos{(2x)} + \cfrac{1 + \cos{(4x)}}{2} \right) \, dx \\
= \frac{1}{4} \int \left( 1 - 2 \cos{(2x)} + \cfrac{1}{2} + \cfrac{\cos{(4x)}}{2} \right) \, dx \\
= \frac{1}{4} \int \left( \cfrac{3}{2} - 2 \cos{(2x)} + \cfrac{1}{2} \cos{(4x)} \right) \, dx \\
= \frac{1}{8} \int \left( 3 - 4 \cos{(2x)} + \cos{(4x)} \right) \, dx
\end{align}

And finally we integrate:

\begin{align}
\int \sin^{4}{x} \, dx & = \frac{1}{8} \int \left( 3 - 4 \cos{(2x)} + \cos{(4x)} \right) \, dx \\
& = \frac{1}{8} \left( 3x - \cfrac{4 \sin{(2x)}}{2} + \cfrac{\sin{(4x)}}{4} \right) + C \\
& = \boxed{\frac{1}{8} \left( 3x - 2 \sin{(2x)} + \frac{1}{4} \sin{(4x)} \right) + C}
\end{align}

##### Objective 2: Solve trigonometric integrals involving powers of secant and tangent

Now we consider integrals of the form $$\int \sec^{m}{x} \tan^{n}{x} \, dx$$ where either $m$ or $n$ are *positive* integers



##### Recall Common Trig Identities Involving Secant and Tangent

* Pythagorean identities:
    * $\sec^{2}{x} = \tan^{2}{x} + 1$
    * $\tan^{2}{x} = \sec^{2}{x} - 1$

##### Guidelines for Evaluating Integrals Involving Powers of Secant and Tangent

1. For *positive, even powers of secant*, save a factor of $\sec^{2}{x}$ for substitution and convert the remaining factors to tangents. Then expand and integrate.

2. For *positive, odd powers of tangent*, save a factor of $\sec{x}\tan{x}$ for substitution and convert the ramaining factors to secants. Then expand and integrate.

3. If no secant and tangent has positive, even powers, then convert a factor of $\tan^{2}{x}$ to $\sec^{2}{x} - 1$. Then expand and repeat if necessary.

4. If no secant and tangent has positive, odd powers, use integration by parts

5. If methods 1 through 4 do not apply, then convert to sines and cosines

##### Example 5

Find $\displaystyle \int \cfrac{\tan^{3}{x}}{\sqrt{\sec{x}}} \, dx$.

##### Solution:

We have a positive, odd power of tangent so we need to save a factor of $\sec{x}\tan{x}$ for substitution. But how do we break off a factor of $\sec{x}$?

First, rewrite:

\begin{align}
\int \cfrac{\tan^{3}{x}}{\sqrt{\sec{x}}} \, dx & = \int \cfrac{\tan^{3}{x}}{\sec^{1/2}{x}} \, dx \\
& = \int \sec^{-1/2}{x} \tan^{3}{x} \, dx
\end{align}

Next, add zero in the power of the secant term and then we can break off a secant:

\begin{align}
\int \sec^{-1/2}{x} \tan^{3}{x} \, dx & = \int \sec^{(-1/2 + 1 - 1)}{x} \tan^{3}{x} \, dx \\
& = \int \sec^{-3/2}{x} \tan^{3}{x} \sec{x} \, dx
\end{align}

And now break off a tangent:

$$\int \sec^{-3/2}{x} \tan^{3}{x} \sec{x} \, dx  = \int \sec^{-3/2}{x} \tan^{2}{x} \sec{x} \tan{x} \, dx$$

Pythagorean identity:

\begin{align}
\int \sec^{-3/2}{x} \underbrace{\tan^{2}{x}}_{\sec^{2}{x} - 1} \sec{x} \tan{x} \, dx & = \int \sec^{-3/2}{x} (\sec^{2}{x} - 1) \sec{x} \tan{x} \, dx
\end{align}

Now let $u = \sec{x}$. Then $du = \sec{x} \tan{x} \, dx$.

Substituting:

$$\int \underbrace{\sec^{-3/2}{x}}_{u^{-3/2}} (\underbrace{\sec^{2}{x}}_{u^2} - 1) \underbrace{\sec{x} \tan{x} \, dx}_{du} = \int u^{-3/2} (u^2 - 1) \, du$$

Distributing:

$$\int u^{-3/2} (u^2 - 1) \, du = \int u^{1/2} - u^{-3/2} \, dx$$

Integrating:

\begin{align}
 \int \cfrac{\tan^{3}{x}}{\sqrt{\sec{x}}} \, dx & = \int u^{1/2} - u^{-3/2} \, dx \\
 & = \cfrac{u^{3/2}}{3/2} - \cfrac{u^{-1/2}}{-1/2} + C \\
 & = \frac{2}{3} u^{3/2} + 2 u^{-1/2} + C \\
 & = \frac{2}{3} \sec^{3/2}{x} + 2 \sec^{-1/2}{x} + C \\
 & = \boxed{\frac{2}{3} \sec^{3/2}{x} + \frac{2}{\sqrt{\sec{x}}} + C}
\end{align}

##### Example 6

Find $\displaystyle \int \sec^{4}{(3x)} \tan^{3}{(3x)} \, dx$.

##### Solution:

First, we note that we have an argument of $3x$ in both secant and tangent parts. So we will use a $u$ substitution to handle this.

Let $u= 3x$. Then $du = 3 \, dx \Rightarrow \frac{1}{3} \, du = dx$.

Substituting:

$$\int \sec^{4}{\underbrace{3x}_{u}} \tan^{3}{\underbrace{3x}_{u}} \, \underbrace{dx}_{\frac{1}{3} \, du} = \frac{1}{3} \int \sec^{4}{u} \tan^{3}{u} \, du$$

Here, we actually can use either guideline 1 or 2 from before. We have positive, even power of secant *and* we also have positive, odd power of tangent.

We will go ahead and choose to follow guideline 1 since we have already seen an example of using guideline 2. So we break off a factor of $\sec^{2}{x}$ and save for substitution:

$$\frac{1}{3} \int \sec^{4}{u} \tan^{3}{u} \, du = \frac{1}{3} \int \sec^{2}{u} \tan^{3}{u} \sec^{2}{u} \, du$$

Pythagorean identity:

$$\frac{1}{3} \int \underbrace{\sec^{2}{u}}_{1+\tan^{2}{u}} \tan^{3}{u} \sec^{2}{u} \, du = \frac{1}{3} \int (1 + \tan^{2}{u}) \tan^{3}{u} \sec^{2}{u} \, du$$

Now let $t = \tan{u}$. Then $dt = \sec^{2}{t} \, dt$.

Substituting:

$$\frac{1}{3} \int (1 + \underbrace{\tan^{2}{u}}_{t^2}) \underbrace{(\tan^{3}{u})}_{t^3} \underbrace{\sec^{2}{u} \, du}_{dt} = \frac{1}{3} \int (1 + t^2) t^3 \, dt$$

Distributing and integrating:

\begin{align}
 \int \sec^{4}{3x} \tan^{3}{3x} \, dx & = \frac{1}{3} \int (1 + t^2) t^3 \, dt \\
 & = \frac{1}{3} \int t^3 + t^5 \, dt \\
 & = \frac{1}{3} \left( \cfrac{t^4}{4} + \cfrac{t^6}{6} \right) + C \\
 & = \frac{1}{3} \left( \cfrac{\tan^{4}{u}}{4} + \cfrac{\tan^{6}{u}}{6} \right) + C \\
 & = \boxed{\frac{1}{6} \left( \cfrac{\tan^{4}{(3x)}}{2} + \cfrac{\tan^{6}{(3x)}}{3} \right) + C} \\
\end{align}
