##### Objective 1: Understand the concept of partial fraction decomposition

##### Motivating Example

Consider the integral $\displaystyle \int \cfrac{1}{x^2 - 5x + 6} \, dx$.

One possible approach:

* First, complete the square:
    * Recall the formula $(x + b)^2 = x^2 + 2bx + b^2$
    * Let $2b = -5$ (since this is the coefficient on the $x$ term)
    * Then $b = -\cfrac{5}{2}$ and $b^2 = \cfrac{25}{4}$
    * To complete the square we add and subtract $b^2 = \cfrac{25}{4}$ in the denominator: $$\int \cfrac{1}{\left( x^2 - 5x + \frac{25}{4} \right) + 6 - \frac{25}{4}} \, dx$$
    * Then factor and simplify: $$\int \cfrac{1}{\left( x - \frac{5}{2} \right)^2 - \frac{1}{4}} \, dx$$

Next, let $u = x - \frac{5}{2}$. Then $du = dx$.

Substituting:

$$\int \cfrac{1}{\left( x - \frac{5}{2} \right)^2 - \frac{1}{4}} \, dx = \int \cfrac{1}{u^2 - \frac{1}{4}} \, du$$

Although we don't have a square root, we could still use a trig substitution here. We have something of the form $u^2 - a^2$ with $a = \cfrac{1}{2}$ which resembles $\sec^{2}{\theta} - 1$. 

So we let $u = \cfrac{1}{2} \sec{\theta}$. Then $du = \cfrac{1}{2} \sec{\theta} \tan{\theta} \, d\theta$.

\begin{align}
\int \cfrac{1}{u^2 - \frac{1}{4}} \, du & = \int \cfrac{1}{\left( \frac{1}{2} \sec{\theta} \right)^2 - \frac{1}{4}} \cdot \cfrac{1}{2} \sec{\theta} \tan{\theta} \, d\theta
\end{align}

\begin{align}
\int \cfrac{1}{\left( \frac{1}{2} \sec{\theta} \right)^2 - \frac{1}{4}} \cdot \cfrac{1}{2} \sec{\theta} \tan{\theta} \, d\theta & = \cfrac{1}{2} \int \cfrac{1}{\frac{1}{4} \sec^{2}{\theta} - \frac{1}{4}} \cdot  \sec{\theta} \tan{\theta} \, d\theta \\
& = \cfrac{1}{2} \int \cfrac{1}{ \frac{1}{4} \left( \sec^{2}{\theta} - 1 \right)} \cdot  \sec{\theta} \tan{\theta} \, d\theta \\
& = \frac{1/2}{1/4} \int \cfrac{1}{\left( \sec^{2}{\theta} - 1 \right)} \cdot  \sec{\theta} \tan{\theta} \, d\theta \\
& = 2 \int \cfrac{1}{\tan^{2}{\theta}} \cdot  \sec{\theta} \tan{\theta} \, d\theta \\
& = 2 \int \cfrac{1}{\tan{\theta}} \cdot  \sec{\theta} \, d\theta
\end{align}

Now we change to sines and cosines and simplify:

\begin{align}
2 \int \cfrac{1}{\tan{\theta}} \cdot  \sec{\theta} \, d\theta = 2 & = 2 \int \cfrac{1}{\frac{\sin{\theta}}{\cos{\theta}}} \cdot  \cfrac{1}{\cos{\theta}} \, d\theta \\
& = 2 \int \cfrac{\cos{\theta}}{\sin{\theta}} \cdot  \cfrac{1}{\cos{\theta}} \, d\theta \\
& = 2 \int \cfrac{1}{\sin{\theta}} \, d\theta \\
& = 2 \int \csc{\theta} \, d\theta
\end{align}

Integrating:

\begin{align}
2 \int \csc{\theta} \, d\theta & = -2 \ln \left| \csc{\theta} + \cot{\theta} \right| + C
\end{align}

Looking at a triangle with $u = \cfrac{1}{2} \sec{\theta} \Rightarrow \cos{\theta} = \cfrac{1}{2u}$:

```python
import warnings
warnings.filterwarnings('ignore')

import numpy as np
import matplotlib.patches as patches
import matplotlib.pyplot as plt
%matplotlib inline

plt.style.use('fivethirtyeight')
```

```python
def right_triangle(A, B, C, axis, hyp, leg1, leg2):
    x = [A[0],B[0],C[0]]
    y = [A[1],B[1],C[1]]
    axis.plot(x,y, color ='b')
    axis.plot([A[0],C[0]], [A[1],C[1]], color = 'b')
    axis.annotate(leg1,(.9,-.4), fontsize=25)
    axis.annotate(leg2,(2.05,1), fontsize=25)
    axis.annotate(hyp,(.5,1), fontsize=25, rotation=45)
    rect = patches.Rectangle(
        (1.8,0), width=.2, height=.2, fill=False, edgecolor='b', alpha=1)
    arc = patches.Arc(
        (0,0), width=.4, height=.4, theta1=0, theta2=45,
        fill=False, edgecolor='b', alpha=1)
    axis.annotate(r'$\theta$', (.3,.05), fontsize=25)
    axis.add_artist(rect)
    axis.add_artist(arc)
    axis.set_xticks([])
    axis.set_xlim(min(x)-1,max(x)+1)
    axis.set_yticks([])
    axis.set_ylim(min(y)-1,max(y)+1)
    axis.set_aspect('equal')
    
A = (0,0)
B = (2,2)
C = (2,0)
```

So $\csc{\theta} = \cfrac{2u}{\sqrt{4u^2 - 1}}$ and $\cot{\theta} = \cfrac{1}{\sqrt{4u^2 - 1}}$

\begin{align}
-2 \ln \left| \csc{\theta} + \cot{\theta} \right| + C & = -2 \ln \left| \cfrac{2u}{\sqrt{4u^2 - 1}} + \cfrac{1}{\sqrt{4u^2 - 1}} \right| + C \\
& = -2 \ln \left| \cfrac{2u + 1}{\sqrt{4u^2 - 1}} \right| + C
\end{align}

Back substituting from $u = x - \cfrac{5}{2}$:

\begin{align}
\int \cfrac{1}{x^2 - 5x + 6} \, dx & = -2 \ln \left| \cfrac{2u + 1}{\sqrt{4u^2 - 1}} \right| + C \\
& = -2 \ln{\left| \cfrac{2(x - 5/2) + 1}{\sqrt{4(x - 5/2)^2 - 1}} \right|}  + C \\
& = -2 \ln{\left| \cfrac{2x - 4}{\sqrt{4(x^2 - 5x + 25/4) - 1}} \right|} + C \\
& = -2 \ln{\left| \cfrac{2x - 4}{\sqrt{4x^2 - 20x + 25 - 1}} \right|} + C \\
& = -2 \ln{\left| \cfrac{2(x - 2)}{\sqrt{4x^2 - 20x + 24}} \right|} + C \\
& = -2 \ln{\left| \cfrac{2(x - 2)}{2 \sqrt{x^2 - 5x + 6}} \right|} + C \\
& = -2 \ln{\left| \cfrac{(x - 2)}{\sqrt{x^2 - 5x + 6}} \right|} + C \\
& = -2 \ln{\left| \cfrac{(x - 2)}{\sqrt{(x - 2)(x-3)}} \right|} + C \\
& = -2 \left( \ln{\left| x - 2 \right|} - \ln{\sqrt{(x - 2)(x-3)}} \right) + C \\
& = -2 \left( \ln{\left| x - 2 \right|} - \left( \ln{\sqrt{x - 2}} + \ln{\sqrt{x-3}} \right) \right) + C \\
& = -2 \left( \ln{\left| x - 2 \right|} - \ln{\sqrt{x - 2}} - \ln{\sqrt{x-3}} \right) + C \\
& = -2 \left( \ln{\left| x - 2 \right|} - \ln{(x - 2)^{1/2}} - \ln{(x-3)^{1/2}} \right) + C \\
& = -2 \left( \ln{\left| x - 2 \right|} - \cfrac{1}{2} \ln{\left| x - 2 \right|} - \cfrac{1}{2} \ln{\left| x-3 \right|} \right) + C \\
& = -2 \ln{\left| x - 2 \right|} + \ln{\left| x - 2 \right|} + \ln{\left| x - 3 \right|} + C \\
\int \cfrac{1}{x^2 - 5x + 6} \, dx & = \boxed{\ln{\left| x-3 \right|} - \ln{\left| x - 2 \right|} + C}
\end{align}

##### Alternative Approach: Partial Fraction Decomposition

Let's try an alternative approach. We might notice that we can factor the denomitator right away:

\begin{align}
\int \cfrac{1}{x^2 - 5x + 6} \, dx & = \int \cfrac{1}{(x - 2)(x - 3)} \, dx
\end{align}

If we have two linear factors multiplied in the denomitator of an already reduced fraction we can decompose the fraction into two separate terms for constants $A$ and $B$ to be determined:

\begin{align}
\cfrac{1}{(x - 2)(x - 3)} & = \cfrac{A}{x - 2} + \cfrac{B}{x - 3}
\end{align}

To solve for $A$ and $B$ we multiply both sides by the common denominator and then choose convenient values for $x$ to get a system of equations witht he same number of equations as unknowns.

Multiplying by $(x-2)(x-3)$:

\begin{align}
(x - 2)(x - 3) \cdot \cfrac{1}{(x - 2)(x - 3)} & = (x - 2)(x - 3) \cdot \left( \cfrac{A}{x - 2} + \cfrac{B}{x - 3} \right) \\
& \Downarrow \\
1  & = A (x - 3) + B(x - 2)
\end{align}

Now if $x = 3$:

\begin{align}
1 & = A(0) + B (3 - 2) \\
& \Downarrow \\
B & = 1
\end{align}

If $x = 2$:

\begin{align}
1 & = A (2 - 3) + B(0) \\
& \Downarrow \\
A & = -1
\end{align}

So with $A = -1$ and $B = 1$:

\begin{align}
\int \cfrac{1}{x^2 - 5x + 6} \, dx & = \int \cfrac{1}{(x - 2)(x - 3)} \, dx \\
& = \int \cfrac{-1}{x - 2} + \cfrac{1}{x - 3} \, dx \\
& = \int \cfrac{1}{x - 3} - \cfrac{1}{x - 2} \, dx \\
& = \int \cfrac{1}{x - 3} \, dx - \int \cfrac{1}{x - 2} \, dx 
\end{align}

Now let $u = x - 3$ for the left integral and $u = x - 2$ for the right integral. Then $du = dx$ for both and we end up with the following.

\begin{align}
\int \cfrac{1}{x^2 - 5x + 6} \, dx & = \boxed{\ln{\left| x - 3 \right|} - \ln{\left| x - 2 \right|} + C}
\end{align}

##### Objective 2: Use partial fraction decomposition with linear factors to integrate rational functions

##### Partial Fraction Decomposition

1. **Divide when improper:** For rational functions with numerators of greater or equal degree than the denominator use long division to simplify.

2. **Factor denominator:** Completely factor the denominator into linear ($ax+b$) and quadratic ($ax^2 + bx + c$) factors.

3. **Linear factors:** For each linear factor in the denominator (i.e. for factors of the form $(ax + b)^m$) the partial fraction decomposition is of the form $$\frac{A_1}{(ax + b)} + \frac{A_2}{(ax + b)^2} + \frac{A_3}{(ax + b)^2} + \cdots + \frac{A_m}{(ax + b)^m}$$

4. **Quadratic factors:** For each prime quadratic factor in the denominator (i.e. for prime factors of the form $(ax^2 + bx + c)^n$) the partial fraction decomposition is of the form $$\cfrac{B_1x+C_1}{(ax^2 + bx + c)} + \cfrac{B_2x+C_2}{(ax^2 + bx + c)^2} + \cfrac{B_3x+C_3}{(ax^2 + bx + c)^3} + \cdots + \cfrac{B_n x+C_n}{(ax^2 + bx + c)^n}$$

##### Example 1

Find $\displaystyle \int \cfrac{5x^2 + 20x + 6}{x^3 + 2x^2 + x} \, dx$.

##### Solution:

Going through the steps of partial fraction decomposition:

1. We do not have an improper fraction, so we skip this step.
2. We factor the denominator:

\begin{align}
    x^3 + 2 x^2 + x & = x(x^2 + 2x + 1) \\
    & = x(x + 1)(x + 1) \\
    & = x(x + 1)^2
\end{align}

3. We only have linear factors so we decompose the fraction in the following way.

\begin{align}
    \cfrac{5x^2 + 20x + 6}{x^3 + 2 x^2 + x} & = \cfrac{5x^2 + 20x + 6}{x(x+1)^2} \\
    & \Downarrow \\
    \cfrac{5x^2 + 20x + 6}{x(x+1)^2} & = \cfrac{A}{x} + \cfrac{B}{x + 1} + \cfrac{C}{(x + 1)^2}
\end{align}

4. There are no quadratric factors, so we skip this step.

Now we need to find values $A, B$, and $C$ to determine the specific sum of fractions that combine into our integrand.

Multiplying by common denominator and simplifying:

\begin{align}
x(x+1)^2 \cdot \cfrac{5x^2 + 20x + 6}{x(x+1)^2} & = x(x+1)^2 \cdot \left( \cfrac{A}{x} + \cfrac{B}{x + 1} + \cfrac{C}{(x + 1)^2} \right) \\
5x^2 + 20x + 6 & = A(x + 1)^2 + B x (x + 1) + C x
\end{align}

Now let $x = 0$:

\begin{align}
5(0)^2 + 20(0) + 6 & = A(1)^2 + B (0) + C(0) \\
A & = 6 \\
\end{align}

Now let $x = -1$:

\begin{align}
5(-1)^2 + 20(-1) + 6 & = A(0)^2 + B(0) + C (-1) \\
5 - 20 + 6 & = -C \\
C & = 9
\end{align}

Finally, let $x = 1$ and sub in the values for $A$ and $C$:

\begin{align}
5(1)^2 + 20(1) + 6 & = 6(2)^2 + B(1)(2) + 9(1) \\
5 + 20 + 6 & = 24 + 2B + 9 \\
31 & = 33 + 2B \\
2B & = -2 \\
B & = -1
\end{align}

Going back to the integral we then have

$$\int \cfrac{5x^2 + 20x + 6}{x^3 + 2x^2 + x} \, dx = \int \left( \cfrac{6}{x} - \cfrac{1}{x + 1} + \cfrac{9}{(x + 1)^2} \right) \, dx$$

Using integration properties for sums and differences:

\begin{align}
\int \left( \cfrac{6}{x} - \cfrac{1}{x + 1} + \cfrac{9}{(x + 1)^2} \right) \, dx & = \int \cfrac{6}{x} \, dx - \int \cfrac{1}{x + 1} \, dx + \int \cfrac{9}{(x + 1)^2} \, dx
\end{align}

The first integral is straightforward. The middle and last integral we can use $u = x + 1$ and $du = dx$ for both:

\begin{align}
\int \left( \cfrac{6}{x} - \cfrac{1}{x + 1} + \cfrac{9}{(x + 1)^2} \right) \, dx & = \int \cfrac{6}{x} \, dx - \int \cfrac{1}{x + 1} \, dx + \int \cfrac{9}{(x + 1)^2} \, dx \\
& = 6 \ln|x| - \int \cfrac{1}{u} \, du + \int \cfrac{9}{u^2} \, du \\
& = 6 \ln|x| - \ln{|u|} + 9\int u^{-2} \, du \\
& = 6 \ln|x| - \ln{|u|} + 9 \frac{u^{-2+1}}{-2 + 1} + C \\
& = 6 \ln|x| - \ln{|u|} - 9 u^{-1} + C \\
& = 6 \ln|x| - \ln{|u|} - \cfrac{9}{u} + C 
\end{align}

Then back substitute:

$$\int \cfrac{5x^2 + 20x + 6}{x^3 + 2x^2 + x} \, dx = \boxed{6 \ln|x| - \ln{|x + 1|} - \cfrac{9}{x + 1} + C}$$

##### Objective 3: Use partial fraction decomposition with quadratic factors to integrate rational functions

##### Example 2

Find $\displaystyle \int \cfrac{2x^3 - 4x - 8}{(x^2 - x)(x^2 + 4)} \, dx$.

##### Solution:

Going through the steps of partial fraction decomposition:

1. We do not have an improper fraction, so we skip this step.
2. We factor the denominator:

\begin{align}
    (x^2 - x)(x^2 + 4) & = x(x - 1)(x^2 + 4)
\end{align}

3. and 4. We have two linear factors and one prime quadratic factor so we need to decompose the fraction as follows: 

\begin{align}
    \cfrac{2x^3 - 4x - 8}{(x^2 - x)(x^2 + 4)} & = \cfrac{2x^3 - 4x - 8}{x(x - 1)(x^2 + 4)} \\
    & \Downarrow \\
    \cfrac{2x^3 - 4x - 8}{x(x - 1)(x^2 + 4)} & = \cfrac{A}{x} + \cfrac{B}{x - 1} + \cfrac{Cx + D}{(x^2 + 4)}
\end{align}

Multiply both sides by the common denominator:

$$
x(x - 1)(x^2 + 4) \cdot \cfrac{2x^3 - 4x - 8}{x(x - 1)(x^2 + 4)} \\
= x(x - 1)(x^2 + 4) \cdot \left( \cfrac{A}{x} + \cfrac{B}{x - 1} + \cfrac{Cx + D}{(x^2 + 4)} \right) \\
 \Downarrow \\
2x^3 - 4x - 8 = A (x - 1)(x^2 + 4) + B x(x^2 + 4) + (Cx + D)x(x - 1)
$$

Now let $x = 0$:

\begin{align}
2(0)^3 - 4(0) - 8 & = A(-1)(4) + B(0) + (C(0) + D)(0) \\
-8 & = -4A \\
A & = 2
\end{align}

Now let $x = 1$:

\begin{align}
2(1)^3 - 4(1) - 8 & = A(0) + B(1)(5) + (C(1) + D)(0) \\
2 - 4 - 8 & = 5B \\
-10 & = 5B \\
B & = -2
\end{align}

Now we can't eliminate any terms with any other real value of $x$, however, we can use two arbitrary values of $x$ to get a system of two equations and two unknowns ($C$ and $D$).

We let $x = 2$ and plug in the known values $A = 2$ and $B = -2$ first:

\begin{align}
2(2)^3 - 4(2) - 8 & = 2(1)(8) - 2(2)(8) + (C(2) + D)(2)(1) \\
16 - 8 - 8 & = 16 - 32 + 4C + 2D \\
0 & = -16 + 4C + 2D \\
4C + 2D & = 16 \\
2C + D & = 8 \qquad \text{(equation 1)}
\end{align}

Now let $x = -1$ and plug in the known values $A = 2$ and $B = -2$:

\begin{align}
2(-1)^3 - 4(-1) - 8 & = 2(-2)(5) - 2(-1)(5) + (C(-1) + D)(-1)(-2) \\
-2 + 4 - 8 & = -20 + 10 - 2C + 2D \\
-6 & = -10 - 2C + 2D \\
-2C + 2D & = 4 \\
-C + D & = 2 \qquad \text{(equation 2)}
\end{align}

So now we need to solve the system of equations given by $$2C + D = 8 \\ -C + D = 2.$$

If we subtract the second equation from the first:

\begin{align}
2C + D = 8 \\ \underline{-(-C + D = 2)} \\
3C = 6 \\
C = 2
\end{align}

Now using this value for $C$ in the second equation:

\begin{align}
-2 + D & = 2 \\
D & = 4
\end{align}

So now we can go to our integral:

\begin{align}
\int \cfrac{2x^3 - 4x - 8}{(x^2 - x)(x^2 + 4)} \, dx & = \int \left( \cfrac{A}{x} + \cfrac{B}{x - 1} + \cfrac{Cx + D}{(x^2 + 4)} \right) \, dx \\
& = \int \left( \cfrac{2}{x} - \cfrac{2}{x - 1} + \cfrac{2x + 4}{(x^2 + 4)} \right) \, dx \\
& = 2 \int \left( \cfrac{1}{x} - \cfrac{1}{x - 1} + \cfrac{x + 2}{(x^2 + 4)} \right) \, dx 
\end{align}

Distributing the integral using integration properties:

\begin{align}
2 \int \left( \cfrac{1}{x} - \cfrac{1}{x - 1} + \cfrac{x + 2}{(x^2 + 4)} \right) \, dx & = 2 \int \cfrac{1}{x} \, dx \\
& \qquad - 2 \int \cfrac{1}{x - 1} \, dx + 2 \int \cfrac{x + 2}{(x^2 + 4)} \, dx \\
& = 2 \int \cfrac{1}{x} \, dx - 2 \int \cfrac{1}{x - 1} \, dx \\
& \qquad + 2 \int \cfrac{x}{(x^2 + 4)} \, dx + 2 \int \cfrac{2}{x^2 + 4} \, dx \\
& = 2 \int \cfrac{1}{x} \, dx - 2 \int \cfrac{1}{x - 1} \, dx \\
& \qquad + 2 \int \cfrac{x}{(x^2 + 4)} \, dx + 4 \int \cfrac{1}{x^2 + 4} \, dx \\
\end{align}

The first and the last integrals we have formulas for:

\begin{align}
2 \int \cfrac{1}{x} \, dx - 2 \int \cfrac{1}{x - 1} \, dx + 2 \int \cfrac{x}{(x^2 + 4)} \, dx + 4 \int \cfrac{1}{x^2 + 4} \, dx \\
= 2 \ln{|x|} - 2 \int \cfrac{1}{x - 1} \, dx + 2 \int \cfrac{x}{(x^2 + 4)} \, dx + 4 \left( \frac{1}{2} \tan^{-1}{\frac{x}{2}} \right) \\
= 2 \ln{|x|} + 2 \tan^{-1}{\frac{x}{2}} - 2 \int \cfrac{1}{x - 1} \, dx + 2 \int \cfrac{x}{(x^2 + 4)} \, dx 
\end{align}

We have two remaining integrals:

* We let $u = x - 1$ for the first remaining integral. Then $du = dx$.
* We let $t = x^2 + 4$ for the second remaining integral. Then $dt = 2x \, dx \Rightarrow \frac{1}{2} \, dt = x \, dx$.

\begin{align}
2 \ln{|x|} + 2 \tan^{-1}{\frac{x}{2}} - 2 \int \cfrac{1}{\underbrace{x - 1}_{u}} \, \underbrace{dx}_{du} + 2 \int \cfrac{1}{\underbrace{(x^2 + 4)}_{t}} \underbrace{x \, dx}_{\frac{1}{2} \, dt} \\
= 2 \ln{|x|} + 2 \tan^{-1}{\frac{x}{2}} - 2 \int \cfrac{1}{u} \, du + 2 \cdot \frac{1}{2} \int \cfrac{1}{t} \, dt \\
\end{align}

Now we integrate the remaining two integrals and back substitute:

\begin{align}
\int \cfrac{2x^3 - 4x - 8}{(x^2 - x)(x^2 + 4)} \, dx & = 2 \ln{|x|} + 2 \tan^{-1}{\frac{x}{2}} - 2 \int \cfrac{1}{u} \, du + 2 \cdot \frac{1}{2} \int \cfrac{1}{t} \, dt \\
& = 2 \ln{|x|} + 2 \tan^{-1}{\frac{x}{2}} - 2 \ln{|u|} + \ln{|t|} + C \\
& = 2 \ln{|x|} + 2 \tan^{-1}{\frac{x}{2}} - 2 \ln{|x - 1|} + \ln{(x^2 + 4)} + C \\
& = \boxed{2 \ln{|x|} - 2 \ln{|x - 1|} + \ln{(x^2 + 4)} + 2 \tan^{-1}{\frac{x}{2}} + C}
\end{align}

##### Example 3

Find $\displaystyle \int \cfrac{8x^3 + 13x}{(x^2 + 2)^2} \, dx$.

##### Solution:

Going through the steps of partial fraction decomposition:

1. We do not have an improper fraction, so we skip this step.
2. The denominator cannot be factored any further.
3. We do not have any linear factors.
4. We have a single prime quadratic factor, but it's squared, so we need to decompose the fraction as follows: 

\begin{align}
    \cfrac{8x^3 + 13x}{(x^2 + 2)^2} & = \cfrac{Ax + B}{(x^2 + 2)} + \cfrac{Cx + D}{(x^2 + 2)^2}
\end{align}

Multiply both sides by the common denominator:

\begin{align}
(x^2 + 2)^2 \cdot \cfrac{8x^3 + 13x}{(x^2 + 2)^2} & = (x^2 + 2)^2 \cdot \left( \cfrac{Ax + B}{(x^2 + 2)} + \cfrac{Cx + D}{(x^2 + 2)^2} \right) \\
& \Downarrow \\
8x^3 + 13x & = (Ax + B)(x^2 + 2) + Cx + D 
\end{align}

If $x = 0$:

\begin{align}
8(0)^3 + 13(0) & = (A(0) + B)((0)^2 + 2) + C(0) + D \\
& \Downarrow \\
2 B + D & = 0 \qquad \text{(equation 1)}
\end{align}

If $x = 1$:

\begin{align}
8(1)^3 + 13(1) & = (A(1) + B)((1)^2 + 2) + C(1) + D \\
& \Downarrow \\
21 & = 3(A + B) + C + D \\
3 A + 3 B + C + D & = 21 \qquad \text{(equation 2)}
\end{align}

If $x = -1$:

\begin{align}
8(-1)^3 + 13(-1) & = (A(-1) + B)((-1)^2 + 2) + C(-1) + D \\
& \Downarrow \\
- 21 & = 3(-A + B) - C + D \\
-3A + 3B - C + D & = -21 \qquad \text{(equation 3)}
\end{align}

If $x = 2$:

\begin{align}
8(2)^3 + 13(2) & = (A(2) + B)((2)^2 + 2) + C(2) + D \\
& \Downarrow \\
90 & = 6(2A + B) + 2C + D \\
12A + 6B + 2C + D & = 90 \qquad \text{(equation 4)}
\end{align}

So we need to solve the system of four equations and four unknowns:

\begin{align}
2 B + D & = 0 \\
3 A + 3 B + C + D & = 21 \\
-3A + 3B - C + D & = -21 \\
12A + 6B + 2C + D & = 90 
\end{align}

We could use elimination several times to solve or we could use a calculator to solve this system. We would end up with $A = 8, B = 0, C = -3$, and $D = 0$.

Alternatively, we may notice that since no value of $x$ can be chosen conveniently to eliminate several variables. When this occurs, it may be easier to use the method of undetermined coefficients: after we multiply by our common denominator, instead of choosing values for $x$ to plug in and obtain a system, we multiply and compare coefficients:

\begin{align}
8x^3 + 13x & = (Ax + B)(x^2+2) + Cx + D \\
8x^3 + 13x & = Ax^3 + Bx^2 + 2Ax + Cx + 2B + D \qquad \text{(multiplying out)} \\
8x^3 + 13x & = Ax^3 + Bx^2 + (2A + C)x + 2B + D \qquad \text{(factoring out $x$)}
\end{align}

Now we can compare coefficinets on each side.

$$\color{RoyalBlue}{8}x^3 + \color{orange}{0}x^2 + \color{red}{13}x + \color{grey}{0} = \color{RoyalBlue}{A}x^3 + \color{orange}{B}x^2 + \color{red}{(2A + C)}x + \color{grey}{2B + D}$$

Matching the coefficients by color: $$A = 8 \\ B = 0 \\ 2A + C = 13 \\ 2B + D = 0$$

Since $A= 8$ and $B=0$ are known, we then have that $2A + C = 13 \Rightarrow C = -3$ and $2B + D = 0 \Rightarrow D = 0$

Therefore, either we have our integral as

\begin{align}
\int \cfrac{8x^3 + 13x}{(x^2 + 2)^2} \, dx & = \int \left( \cfrac{8x}{(x^2 + 2)} - \cfrac{3x}{(x^2 + 2)^2} \right) \, dx \\
& = \int \cfrac{8x}{(x^2 + 2)} \, dx - \int \cfrac{3x}{(x^2 + 2)^2} \, dx \\
\end{align}

Now let $u = x^2 + 2$ for both integrals. Then we get $du = 2x \, dx \Rightarrow \frac{1}{2} \, du = x \, dx$ for both integrals.

So we get

\begin{align}
\int \cfrac{8x^3 + 13x}{(x^2 + 2)^2} \, dx & = 8 \int \cfrac{1}{\underbrace{(x^2 + 2)}_{u}} \cdot \overbrace{x \, dx}^{\frac{1}{2} \, du} - 3 \int \cfrac{1}{\underbrace{(x^2 + 2)^2}_{u^2}} \cdot \overbrace{x \, dx}^{\frac{1}{2} \, du} \\
& = 8 \int \cfrac{1}{u} \cdot \frac{1}{2} \, du - 3 \int \cfrac{1}{u^2} \cdot \frac{1}{2} \, du \\
& = 4 \int \cfrac{1}{u} \, du - \frac{3}{2} \int u^{-2} \, du \\
& = 4 \ln{|u|} - \frac{3}{2} \left( \cfrac{u^{-2+1}}{-2 + 1} \right) + C \\
& = 4 \ln{|u|} - \frac{3}{2} \left( \cfrac{u^{-1}}{-1} \right) + C \\
& = 4 \ln{|u|} + \frac{3}{2} \left( \cfrac{1}{u} \right) + C 
\end{align}

Back substituting:

\begin{align}
\int \cfrac{8x^3 + 13x}{(x^2 + 2)^2} \, dx & = \boxed{4 \ln{(x^2 + 2)} + \cfrac{3}{2(x^2 + 2)} + C}
\end{align}
