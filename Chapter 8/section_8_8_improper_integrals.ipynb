##### Objective 1: Evaluate an improper integral that has an infinite limit of integration

##### Improper Integrals

* Recall that the value of a definite integral $\displaystyle \int_a^b f(x) \, dx$ calculates the area between the $x$-axis and the graph of the function $f(x)$ over the interval $[a, b]$ 

    * In calculating definite integrals, up until now, we have always worked with functions that satisfy the Fundamental Theorem of Calculus:
        * Let $f(x)$ be continuous over the interval $[a,b]$ and $F(x)$ be the antiderivative of $f(x)$. Then $\displaystyle \int_a^b f(x) \, dx = F(b) - F(a)$.
    
    * One of the key requirements for the Fundamental Theorem of Calculus is that $f(x)$ be *continous* over the interval $[a, b]$; but what if this is not the case?
    
    * An integral of $f(x)$ is said to be an **improper integral** if...
        * Its lower limit of integration tends to $-\infty$
        * Its upper limit of integration tends to $+\infty$
        * Both limits of integration are infinite
        * $f(x)$ has finitely many infinite discontinuities over the interval of integration $[a, b]$

##### Improper Integrals with Limits of Integraion Tending to $\pm \infty$

* If $f$ is continous on the interval $[a, \infty)$, then

\begin{align}
    \int_a^\infty f(x) \, dx & = \lim_{b \rightarrow \infty}{\int_a^b f(x) \, dx}
\end{align}

* If $f$ is continous on the interval $(-\infty, b]$, then

\begin{align}
    \int_{-\infty}^b f(x) \, dx & = \lim_{a \rightarrow -\infty}{\int_a^b f(x) \, dx}
\end{align}

* If $f$ is continous on the interval $(-\infty, \infty)$, then

\begin{align}
    \int_{-\infty}^\infty f(x) \, dx & = \int_{-\infty}^c f(x) \, dx + \int_c^\infty f(x) \, dx
\end{align}

For each individual case above, provided *all* of the limits exist on the right hand side we say that the improper integral on the left hand side **converges.** Otherwise, the improper integral **diverges.**

##### Example 1

Find $\displaystyle \int_1^\infty \cfrac{1}{x} \, dx$.

##### Solution:

* Since we have our upper limit of integration tending to $+\infty$ we have an improper integral and we have

\begin{align}
    \int_1^\infty \cfrac{1}{x} \, dx & = \lim_{b \rightarrow \infty} \int_1^b \cfrac{1}{x} \, dx
\end{align}

* Integrating:

\begin{align}
    \lim_{b \rightarrow \infty} \int_1^b \cfrac{1}{x} \, dx & = \lim_{b \rightarrow \infty} \left. \left( \ln{|x|} \right) \right|_1^b \\
    & = \lim_{b \rightarrow \infty} \left( \ln{|b|} - \ln{1} \right) \\
    & = \left( \infty - 0 \right) \\
    & = \infty
\end{align}

* Therefore, the integral $\boxed{\text{diverges}}$ since the limit does not exist

##### Example 2

Find $\displaystyle \int_{-\infty}^0 x^2 e^{x} \, dx$.

##### Solution:

* Since our lower limit tends to $-\infty$ we have an improper integral and we have

\begin{align}
    \int_{-\infty}^0 x^2 e^{x} \, dx & = \lim_{a \rightarrow -\infty} \int_a^0 x^2 e^{x} \, dx
\end{align}

* To perform the integration we could use tabular method for integration by parts:

|$+$|$u = x^2$|$dv = e^x \, dx$|
|:-:|:-:|:-:|
|$-$|$2x$|$e^x$|
|$+$|$2$|$e^x$|
|$-$|$0$|$e^x$|

* From the table above we have 

\begin{align}
    \lim_{a \rightarrow -\infty} \int_a^0 x^2 e^{x} \, dx & = \lim_{a \rightarrow -\infty} \left. (x^2 e^x - 2x e^x + 2 e^x) \right|_a^0 \\
    & = \lim_{a \rightarrow -\infty} ((0 - 0 + 2e^0) - (a^2 e^a - 2a e^a + 2 e^a)) \\
    & = \lim_{a \rightarrow -\infty} (2 - a^2 e^a + 2a e^a - 2 e^a)
\end{align}

* Distributing the limit:

\begin{align}
    \lim_{a \rightarrow -\infty} (2 - a^2 e^a + 2a e^a - 2 e^a) & = 2 - \lim_{a \rightarrow -\infty} a^2 e^a + \lim_{a \rightarrow -\infty} 2a e^a - \lim_{a \rightarrow -\infty} 2 e^a
\end{align}

* Evaluating these limits we can use the fact that an exponential function's growth/decay is far more dominant than a polynomial's growth/decay $\left( \text{i.e. for $n > 0$, $\displaystyle \lim_{a \rightarrow -\infty} a^n e^a \rightarrow 0$} \right)$:

\begin{align}
    \lim_{a \rightarrow -\infty} a^2 e^a & \rightarrow 0 \\
     \lim_{a \rightarrow -\infty} 2a e^a & \rightarrow 0 \\
     \lim_{a \rightarrow -\infty} 2 e^a & \rightarrow 0
\end{align}

* Therefore,

\begin{align}
    \int_{-\infty}^0 x^2 e^{x} \, dx & = \boxed{2}
\end{align}

##### Example 3

Find $\displaystyle \int_{-\infty}^{\infty} \cfrac{e^x}{1 + e^{2x}} \, dx$

##### Solution:

* Now both our bounds of integration are infinite; so for a real number $c$ we have

\begin{align}
    \int_{-\infty}^{\infty} \cfrac{e^x}{1 + e^{2x}} \, dx & = \int_{-\infty}^{c} \cfrac{e^x}{1 + e^{2x}} \, dx + \int_{c}^{\infty} \cfrac{e^x}{1 + e^{2x}} \, dx
\end{align}

* Since we are integrating over the interval $(-\infty, \infty)$ we can choose the value for $c$; a convenient value is $c = 0$:

\begin{align}
    \int_{-\infty}^{\infty} \cfrac{e^x}{1 + e^{2x}} \, dx & = \int_{-\infty}^{0} \cfrac{e^x}{1 + e^{2x}} \, dx + \int_{0}^{\infty} \cfrac{e^x}{1 + e^{2x}} \, dx
\end{align}

* Since both integrands are the same we will evaluate the indefinite integral and then apply the appropriate limits:

\begin{align}
    \int \cfrac{e^x}{1 + e^{2x}} \, dx & = \int \cfrac{e^x}{1 + \left(e^{x}\right)^2} \, dx
\end{align}

* Performing a $u$ substitution we let $u = e^x$, then $du = e^x \, dx$ and we have

\begin{align}
    \int \cfrac{e^x}{1 + \left(e^{x}\right)^2} \, dx & = \int \cfrac{1}{1 + u^2} \, du
\end{align}

* The integral on the right is now of the form $\displaystyle \int \cfrac{1}{a^2 + u^2} \, du$ with $a = 1$ so we can use a familiar integration identity:

\begin{align}
    \int \cfrac{1}{1 + u^2} \, du & = \tan^{-1}{u} + C
\end{align}

* Back substituting:

\begin{align}
    \tan^{-1}{u} + C & = \tan^{-1}{(e^x)} + C
\end{align}

* Now we are ready to apply our limits:

\begin{align}
    \int_{-\infty}^{\infty} \cfrac{e^x}{1 + e^{2x}} \, dx & = \int_{-\infty}^{0} \cfrac{e^x}{1 + e^{2x}} \, dx + \int_{0}^{\infty} \cfrac{e^x}{1 + e^{2x}} \, dx \\
    & = \lim_{a \rightarrow -\infty} \int_{a}^{0} \cfrac{e^x}{1 + e^{2x}} \, dx + \lim_{b \rightarrow \infty} \int_{0}^{b} \cfrac{e^x}{1 + e^{2x}} \, dx \\
    & = \lim_{a \rightarrow -\infty} \left. \left( \tan^{-1}{(e^x)} \right) \right|_a^0 + \lim_{b \rightarrow \infty} \left. \left( \tan^{-1}{(e^x)} \right) \right|_0^b \\
    & = \lim_{a \rightarrow -\infty} \left( \tan^{-1}{(e^0)} - \tan^{-1}{(e^a)} \right) \\
    & \qquad + \lim_{b \rightarrow \infty} \left( \tan^{-1}{(e^b)} - \tan^{-1}{(e^0)} \right) \\
    & =  \tan^{-1}{(1)} - \lim_{a \rightarrow -\infty} \tan^{-1}{(e^a)} \\
    & \qquad + \lim_{b \rightarrow \infty} \left( \tan^{-1}{(e^b)} - \tan^{-1}{(1)} \right) \\
\end{align}

* Distributing the limit:

\begin{align}
    \int_{-\infty}^{\infty} \cfrac{e^x}{1 + e^{2x}} \, dx & = \tan^{-1}{(1)} - \lim_{a \rightarrow -\infty} \tan^{-1}{(e^a)} \\
    & \qquad + \lim_{b \rightarrow \infty} \tan^{-1}{(e^b)} - \tan^{-1}{(1)} \\
    & = - \lim_{a \rightarrow -\infty} \tan^{-1}{(e^a)} + \lim_{b \rightarrow \infty} \tan^{-1}{(e^b)}
\end{align}

* Note that $\displaystyle \lim_{a \rightarrow -\infty} e^a \rightarrow 0$ and $\displaystyle \lim_{b \rightarrow \infty} e^b \rightarrow \infty$

* Therefore, $\displaystyle \lim_{a \rightarrow - \infty} \tan^{-1}{(e^a)} = \tan^{-1}(0) = 0$ and $\displaystyle \lim_{b \rightarrow \infty} \tan^{-1}{(e^b)} \rightarrow \pi / 2$

    * The latter limit can be determined by the graph of inverse tangent

* Therefore,

\begin{align}
    \int_{-\infty}^{\infty} \cfrac{e^x}{1 + e^{2x}} \, dx & = \boxed{\cfrac{\pi}{2}}
\end{align}

##### Objective 2: Evaluate an improper integral that has an infinite discontinuity

##### Improper Integrals with Infinite Discontinuities

* If $f$ is continous on the interval $[a, b)$ and has an infinite discontinuity at $b$, then

\begin{align}
    \int_a^b f(x) \, dx & = \lim_{c \rightarrow b^{-}}{\int_a^c f(x) \, dx}
\end{align}

* If $f$ is continous on the interval $(a, b]$ and has an infinite discontinuity at $a$, then

\begin{align}
    \int_a^b f(x) \, dx & = \lim_{c \rightarrow a^{+}}{\int_c^b f(x) \, dx}
\end{align}

* If $f$ is continous on the interval $[a, b]$, except for some $c$ in $(a, b)$ at which $f$ has an infinite discontinuity, then

\begin{align}
    \int_a^b f(x) \, dx & = \int_{a}^c f(x) \, dx + \int_c^b f(x) \, dx
\end{align}

For each individual case above, provided *all* of the limits exist on the right hand side we say that the improper integral on the left hand side **converges.** Otherwise, the improper integral **diverges.**

##### Example 4

Find $\displaystyle \int_{0}^{\pi/2} \cfrac{\cot{t}}{1 + \sin{t}} \, dt$.

##### Solution:

* Since we are integrating $f(t) = \cfrac{\cot{t}}{1 + \sin{t}}$ over the interval $\left[ 0, \frac{\pi}{2} \right]$ we note that we have an infinite discontinuity at $c = 0$ (since $\cot{0}$ is has an infinite discontinunity at $t = 0$)

* So we have an improper integral and we have

\begin{align}
    \int_{0}^{\pi/2} \cfrac{\cot{t}}{1 + \sin{t}} \, dt & = \lim_{a \rightarrow 0^{+}} \int_{a}^{\pi/2} \cfrac{\cot{t}}{1 + \sin{t}} \, dt
\end{align}

* One approach would be to convert to sines and cosines:

\begin{align}
     \lim_{a \rightarrow 0^{+}} \int_{a}^{\pi/2} \cfrac{\cot{t}}{1 + \sin{t}} \, dt & = \lim_{a \rightarrow 0^{+}} \int_{a}^{\pi/2} \cfrac{\frac{\cos{t}}{\sin{t}}}{1 + \sin{t}} \, dt \\
     & = \lim_{a \rightarrow 0^{+}} \int_{a}^{\pi/2} \cfrac{\cos{t}}{\sin{t}} \cdot \cfrac{1}{1 + \sin{t}} \, dt \\
     & = \lim_{a \rightarrow 0^{+}} \int_{a}^{\pi/2} \cfrac{\cos{t}}{\sin{t} + \sin^2{t}} \, dt \\
     & = \lim_{a \rightarrow 0^{+}} \int_{a}^{\pi/2} \cfrac{\cos{t}}{\sin^2{t} + \sin{t}} \, dt
\end{align}

* Now we can perform a $u$ substitution here by letting $u = \sin{t} \Rightarrow du = \cos{t} \, dt$
    
    * Substituting for the bounds of integration: if $t \rightarrow 0^{+}$, then $u \rightarrow 0^{+}$ and if $t = \pi / 2$, then $u = 1$

* Substituting:

\begin{align}
    \lim_{a \rightarrow 0^{+}} \int_{a}^{\pi/2} \cfrac{\cos{t}}{\sin^2{t} + \sin{t}} \, dt & = \lim_{a \rightarrow 0^{+}} \int_{a}^{1} \cfrac{1}{u^2 + u} \, du
\end{align}

* Completing the square:

\begin{align}
    \lim_{a \rightarrow 0^{+}} \int_{a}^{1} \cfrac{1}{u^2 + u} \, du & = \lim_{a \rightarrow 0^{+}} \int_{a}^{1} \cfrac{1}{(u^2 + u + 1/4) - 1/4} \, du \\
    & = \lim_{a \rightarrow 0^{+}} \int_{a}^{1} \cfrac{1}{(u + 1/2)^2 - 1/4} \, du \\
\end{align}

* We can perform a second substitution by letting $x = u + 1/2 \Rightarrow dx = du$
    
    * Again, substituting for the bounds of integration: if $u \rightarrow 0^{+}$, then $x \rightarrow (1/2)^{+}$ and if $u = 1$, then $x = 3/2$

* Substituting:

\begin{align}
    \lim_{a \rightarrow 0^{+}} \int_{a}^{1} \cfrac{1}{(u + 1/2)^2 - 1/4} \, du & = \lim_{a \rightarrow (1/2)^{+}} \int_{a}^{3/2} \cfrac{1}{x^2 - 1/4} \, dx
\end{align}

* Factoring:

\begin{align}
    \lim_{a \rightarrow (1/2)^{+}} \int_{a}^{3/2} \cfrac{1}{x^2 - 1/4} \, dx & = \lim_{a \rightarrow (1/2)^{+}} \int_{a}^{3/2} \cfrac{1}{(x + 1/2)(x - 1/2)} \, dx
\end{align}

* Now we're all setup for partial fractions:

\begin{align}
     \lim_{a \rightarrow (1/2)^{+}} \int_{a}^{3/2} \cfrac{1}{(x + 1/2)(x - 1/2)} \, dx & = \lim_{a \rightarrow (1/2)^{+}} \int_{a}^{3/2} \left( \cfrac{A}{(x + 1/2)} + \cfrac{B}{(x - 1/2)} \right) \, dx
\end{align}

* Solving for $A$ and $B$:

\begin{align}
    \cfrac{1}{(x + 1/2)(x - 1/2)} & = \cfrac{A}{(x + 1/2)} + \cfrac{B}{(x - 1/2)} \\
    & \Downarrow \\
    1 & = A(x - 1/2) + B(x + 1/2) \\
    \text{If $x = 1/2$, then } & B = 1 \\
    \text{If $x = -1/2$, then } & A = -1 \\
\end{align}

* Subustituting into our integral:

\begin{align}
     \lim_{a \rightarrow (1/2)^{+}} \int_{a}^{3/2} \cfrac{1}{(x + 1/2)(x - 1/2)} \, dx & = \lim_{a \rightarrow (1/2)^{+}} \int_{a}^{3/2} \left( \cfrac{-1}{(x + 1/2)} + \cfrac{1}{(x - 1/2)} \right) \, dx \\
     & = \lim_{a \rightarrow (1/2)^{+}} \int_{a}^{3/2} \cfrac{-1}{(x + 1/2)} \, dx \\ 
     & \qquad + \lim_{a \rightarrow (1/2)^{+}} \int_{a}^{3/2} \cfrac{1}{(x - 1/2)} \, dx
\end{align}

* For first integral on the right hand side we can perform the substitution $y = x + 1/2 \Rightarrow dy = dx$ and for the second integral on the right hand side we can perform the substitution $z = x - 1/2 \Rightarrow dz = dx$

    * If $x \rightarrow (1/2)^{+}$, then $y \rightarrow 1^{+}$ and $z \rightarrow 0^{+}$
    
    * If $x = 3/2$, then $y = 2$ and $z = 1$

* Subsitituting and integrating:

\begin{align}
     \lim_{a \rightarrow (1/2)^{+}} \int_{a}^{3/2} \cfrac{-1}{(x + 1/2)} \, dx & + \lim_{a \rightarrow (1/2)^{+}} \int_{a}^{3/2} \cfrac{1}{(x - 1/2)} \, dx \\
     & = \lim_{a \rightarrow 1^{+}} \int_{a}^{2} \cfrac{-1}{y} \, dy + \lim_{a \rightarrow 0^{+}} \int_{a}^{1} \cfrac{1}{z} \, dz \\
     & = \lim_{a \rightarrow 1^{+}} \left. \left( -\ln{|y|} \right) \right|_a^2 + \lim_{a \rightarrow 0^{+}} \left. \left( \ln{|z|} \right) \right|_a^1 \\
     & = \lim_{a \rightarrow 1^{+}} \left( -\ln{2} - \left( -\ln{a} \right) \right) + \lim_{a \rightarrow 0^{+}} \left( \ln{1} - \ln{a} \right) \\
     & = \lim_{a \rightarrow 1^{+}} \left( -\ln{2} + \ln{a} \right) + \lim_{a \rightarrow 0^{+}} \left( \ln{1} - \ln{a} \right) \\
     & = -\ln{2} + \ln{1} +  \ln{1} - \lim_{a \rightarrow 0^{+}} \ln{a} \\
     & = -\ln{2} - \lim_{a \rightarrow 0^{+}} \ln{a}
\end{align}

* Since $\lim_{a \rightarrow 0^{+}} \ln{a} \rightarrow - \infty$ the integral $\boxed{\text{diverges.}}$

##### Example 5

Find $\displaystyle \int_{-1}^2 \cfrac{1}{x^3} \, dx$.

##### Solution:

* Since we are integrating $f(x) = \cfrac{1}{x^3}$ over the interval $[-1, 2]$ we note that we have an infinite discontinuity at $c = 0$

* So we have an improper integral and we have

\begin{align}
    \int_{-1}^{2} \cfrac{1}{x^3} \, dx & = \int_{-1}^{0} \cfrac{1}{x^3} \, dx + \int_{0}^{2} \cfrac{1}{x^3} \, dx \\
    & = \lim_{c \rightarrow 0^{-}} \int_{-1}^{c} \cfrac{1}{x^3} \, dx + \lim_{c \rightarrow 0^{+}} \int_{c}^{2} \cfrac{1}{x^3} \, dx \\
    & = \lim_{c \rightarrow 0^{-}} \int_{-1}^{c} x^{-3} \, dx + \lim_{c \rightarrow 0^{+}} \int_{c}^{2} x^{-3} \, dx \\
    & = \lim_{c \rightarrow 0^{-}} \left. \left( \cfrac{x^{-2}}{-2} \right) \right|_{-1}^{c} + \lim_{c \rightarrow 0^{+}} \left. \left( \cfrac{x^{-2}}{-2} \right) \right|_{c}^{2} \\
    & = \lim_{c \rightarrow 0^{-}} \left. \left( -\cfrac{1}{2x^2} \right) \right|_{-1}^{c} + \lim_{c \rightarrow 0^{+}} \left. \left( -\cfrac{1}{2x^2} \right) \right|_{c}^{2} \\
    & = \lim_{c \rightarrow 0^{-}} \left( -\cfrac{1}{2c^2} - \left( -\cfrac{1}{2(-1)^2} \right) \right) \\
    & \qquad + \lim_{c \rightarrow 0^{+}} \left( -\cfrac{1}{2(2)^2} - \left( -\cfrac{1}{2(c)^2} \right) \right) \\
    & = \lim_{c \rightarrow 0^{-}} \left( -\cfrac{1}{2c^2} + \cfrac{1}{2} \right) + \lim_{c \rightarrow 0^{+}} \left( -\cfrac{1}{8} + \cfrac{1}{2c^2} \right)
\end{align}

* Since $\displaystyle \lim_{c \rightarrow 0^{-}} -\cfrac{1}{2c^2} \rightarrow -\infty$ and $\displaystyle \lim_{c \rightarrow 0^{+}} \cfrac{1}{2c^2} \rightarrow \infty$ both of these limits do not exist! So the improper integral $\boxed{\text{diverges.}}$

##### Example 6

For what values of $p$ does the integral converge?

$$
    \int_{1}^{\infty} \cfrac{1}{x^p} \, dx
$$

##### Solution:

We will consider cases:

1. If $p = 1$, then

\begin{align}
    \int_1^\infty \cfrac{1}{x} \, dx & = \lim_{b \rightarrow \infty} \int_1^b \cfrac{1}{x} \, dx \\
    & = \lim_{b \rightarrow \infty} (\ln{|x|}) \big|_1^b \\
    & = \lim_{b \rightarrow \infty} (\ln{|b|} - \ln|1|) \\
    & = \infty - 0 \\
    & = \infty \qquad \text{(diverges)}
\end{align}

2. If $p > 1$, then

\begin{align}
    \int_1^\infty \cfrac{1}{x^p} \, dx & = \int_1^\infty x^{-p} \, dx
\end{align}

and we can use the power rule for integration (since $x^{-p} \neq x^{-1}$). So we have

\begin{align}
    \int_1^\infty \cfrac{1}{x^p} \, dx & = \int_1^\infty x^{-p} \, dx \\
    & = \lim_{b \rightarrow \infty} \int_1^b x^{-p} \, dx \\
    & = \lim_{b \rightarrow \infty} \left. \left( \cfrac{x^{-p+1}}{-p + 1} \right) \right|_1^b \\
    & = \lim_{b \rightarrow \infty} \left( \cfrac{b^{-p+1}}{-p + 1} - \cfrac{1^{-p+1}}{-p + 1}. \right)
\end{align}

Since $p > 1$, then $p - 1 > 0 \Rightarrow -p + 1 < 0 \Rightarrow b^{-p + 1} \rightarrow 0$ as $b \rightarrow \infty$ and so we get

\begin{align}
    \lim_{b \rightarrow \infty} \left( \cfrac{b^{-p+1}}{-p + 1} - \cfrac{1^{-p+1}}{-p + 1} \right) & = 0 -\cfrac{1}{-p + 1} \\
    & = -\cfrac{1}{-p + 1} \\
    & = \cfrac{1}{p - 1} \qquad \text{(converges)}
\end{align}

3. If $p < 1$, then

\begin{align}
    \int_1^\infty \cfrac{1}{x^p} \, dx & = \int_1^\infty x^{-p} \, dx
\end{align}

and we can again use the power rule for integration (again, since $x^{-p} \neq x^{-1}$). So we get

\begin{align}
    \int_1^\infty \cfrac{1}{x^p} \, dx & = \int_1^\infty x^{-p} \, dx \\
    & = \lim_{b \rightarrow \infty} \int_1^b x^{-p} \, dx \\
    & = \lim_{b \rightarrow \infty} \left. \left( \cfrac{x^{-p+1}}{-p + 1} \right) \right|_1^b \\
    & = \lim_{b \rightarrow \infty} \left( \cfrac{b^{-p+1}}{-p + 1} - \cfrac{1^{-p+1}}{-p + 1}. \right)
\end{align}

as before. But now, since $p < 1$, then $p - 1 < 0 \Rightarrow -p + 1 > 0 \Rightarrow b^{-p + 1} \rightarrow \infty$ as $b \rightarrow \infty$ and so we get

\begin{align}
    \lim_{b \rightarrow \infty} \left( \cfrac{b^{-p+1}}{-p + 1} - \cfrac{1^{-p+1}}{-p + 1} \right) & = \infty -\cfrac{1}{-p + 1} \\
    & = \infty \qquad \text{(diverges)}
\end{align}

##### Theorem 8.7

If $p$ is a real number, then

\begin{align}
    \int_1^\infty \cfrac{1}{x^p} \, dx =\begin{cases} \cfrac{1}{p - 1}, \; p > 1 \\ \text{diverges,} \; p \leq 1 \end{cases}
\end{align}
