##### Theorem 9.22: The Form of a Convergent Power Series

If $f$ is represented by a power series $\displaystyle f(x) = \sum_{k = 0}^\infty a_k (x - c)^k$ for all $x$ in an open interval $I$ containing $c$, then

\begin{align}
    a_k & = \frac{f^{(k)}(c)}{k!}
\end{align}

and

\begin{align}
    f(x) & = f(c) + f'(c)(x - c) + \frac{f''(c)}{2!}(x - c)^2 + \cdots + \frac{f^{(k)}(c)}{k!}(x - c)^k + \cdots
\end{align}

##### Taylor and Maclaurin Series

If a function $f$ has derivatives of *all orders* at $x = c$, then the series

\begin{align}
    \sum_{k = 0}^\infty \frac{f^{(k)}(c)}{k!} (x - c)^k
\end{align}

is called the **Taylor series for $f$ centered at $c$.** If $c = 0$, then the series is the **Maclaurin series for $f$.**

##### Interpretation of Theorem 9.22

* By theorem 9.22, the generalized coefficients of *any* convergent power series must be $a_k = \frac{f^{(k)}(c)}{k!}$ and these are precisely the coefficients of a Taylor series

* Given the definition of a Taylor series, theorem 9.22 could be phrased as a convergent power series centered at $c$ is the Taylor series centered at $c$

* A *convergent power series* is always a *Taylor series*

##### Example 1

Find the Maclaurin series representation for $f(x) = \sin{x}$ and determine the radius and interval of convergence.

##### Solution:

* The approach is very much the same as that of finding a Maclaurin polynomial, but with the added step of generalizing the sequence of terms that define the terms of the series

* Therefore, we find the first several derivatives evaluated at $c = 0$ (since we have a *Maclaurin* series):

\begin{align}
    f(x) & = \sin{x} \Rightarrow f(0) = 0 \\
    f'(x) & = \cos{x} \Rightarrow f'(0) = 1 \\
    f''(x) & = -\sin{x} \Rightarrow f''(0) = 0 \\
    f'''(x) & = -\cos{x} \Rightarrow f'''(0) = -1 \\
    f^{(4)}(x) & = \sin{x} \Rightarrow f^{(4)}(0) = 0 \\
    f^{(5)}(x) & = \cos{x} \Rightarrow f^{(5)}(0) = 1 \\
    f^{(6)}(x) & = -\sin{x} \Rightarrow f^{(6)}(0) = 0 \\
    f^{(7)}(x) & = -\cos{x} \Rightarrow f^{(7)}(0) = -1 \\
    & \vdots \\
    f^{(k)}(x) & = \cdots \\
    & \vdots
\end{align}

* Here we have noted that although we found the first few derivatives, we have *all orders* of differentiation and therefore we never actually stop

* Writing out the first few terms of the series:

\begin{align}
    \sum_{k = 0}^\infty \frac{f^{(k)}(0)}{k!} x^k & = \frac{f(0)}{0!} + \frac{f'(0)}{1!} x + \frac{f''(0)}{2!} x^2 + \frac{f'''(0)}{3!} x^3 + \frac{f^{(4)}(0)}{4!} x^4 \\
    & \qquad \qquad \qquad \qquad \qquad + \frac{f^{(5)}(0)}{5!} x^5 + \frac{f^{(6)}(0)}{6!} x^6 + \frac{f^{(7)}(0)}{7!} x^7 + \cdots \\
    & = 0 + x + 0 x^2 - \frac{1}{3!} x^3 + 0 x^4 + \frac{1}{5!} x^5 + 0 x^6 - \frac{1}{7!} x^7 + \cdots \\
    & = x - \frac{1}{3!} x^3 + \frac{1}{5!} x^5 - \frac{1}{7!} x^7 + \cdots \quad (\text{only the odd terms "survive"}) \\
    & = \sum_{k = 0}^\infty \frac{(-1)^k x^{2k + 1}}{(2k + 1)!}
\end{align}

* Therefore, the Taylor series reperesentation for $\sin{x}$ is $\displaystyle \sum_{k = 0}^\infty \frac{(-1)^k x^{2k + 1}}{(2k + 1)!}$ for which we need to determine the radius and interval of convergence

* We can determine the radius of convergence in the process of determining the interval of convergence

* To determine the interval of convergence we first apply the Ratio Test and then test the endpoints

* Applying the Ratio Test with $u_k = \frac{(-1)^k x^{2k + 1}}{(2k + 1)!}$ and $u_{k + 1} = \frac{(-1)^{k + 1} x^{2k + 3}}{(2k + 3)!}$:

\begin{align}
    \lim_{k \rightarrow \infty} \left| \frac{u_{k + 1}}{u_k} \right| & = \lim_{k \rightarrow \infty} \left| \frac{\frac{(-1)^{k + 1} x^{2k + 3}}{(2k + 3)!}}{\frac{(-1)^k x^{2k + 1}}{(2k + 1)!}} \right| \\
    & = \lim_{k \rightarrow \infty} \left| \frac{x^{2k + 3}}{(2k + 3)!} \cdot \frac{(2k + 1)!}{x^{2k + 1}} \right| \\
    & = \lim_{k \rightarrow \infty} \left| \frac{(2k + 1)!}{(2k + 3)!} \cdot \frac{x^{2k + 3}}{x^{2k + 1}} \right| \\
    & = \lim_{k \rightarrow \infty} \left| \frac{(2k + 1)!}{(2k + 1)! \cdot (2k + 3)(2k + 2)} \cdot \frac{x^2 x^{2k + 1}}{x^{2k + 1}} \right| \\
    & = \lim_{k \rightarrow \infty} \frac{x^2}{(2k + 3)(2k + 2)} \\
    & = x^2 \lim_{k \rightarrow \infty} \frac{1}{(2k + 3)(2k + 2)} \\
    & = x^2 \cdot 0 \\
    & = 0 < 1 \Rightarrow \text{series converges for all $x$} \\
\end{align}

* Therefore, the interval of convergence is $(-\infty, \infty)$ and the radius of convergence is $R = \infty$

* So we have a series representation for *all* real numbers $x$ for $\sin{x}$

* That is, the series $\displaystyle \sum_{k = 0}^\infty \frac{(-1)^k x^{2k + 1}}{(2k + 1)!}$ represents the function $f(x) = \sin{x}$ with interval of convergence $(-\infty, \infty)$

##### Theorem 9.23: Convergence of Taylor Series

If $\displaystyle \lim_{k \rightarrow \infty} R_k = 0$ for all $x$ in the interval $I$, then the Taylor series for $f$ converges and equals $f(x)$,

\begin{align}
    f(x) & = \sum_{k = 0}^\infty \frac{f^{(k)}(c)}{k!} (x - c)^k.
\end{align}

Here $R_k = \frac{f^{(k + 1)}(z)}{(k + 1)!} (x - c)^{k + 1}$ for $z$ between $x$ and $c$ (the remainder from Taylor's Remainder Theorem).

##### Example 2

Show that the Maclaurin series for $f(x) = \sin{x}$ converges to $\sin{x}$ for all $x$.

##### Solution:

* From the first example we know that the Taylor series generated by $\sin{x}$ is

\begin{align}
    \sum_{k = 0}^\infty \frac{(-1)^k x^{2k + 1}}{(2k + 1)!}
\end{align}

* Applying the Ratio Test to determine the interval of convergence for all $x$ tells us that the series converges for all $x$, but this does not tell us what the series *converges to*

    * Therefore we need to apply theorem 9.23 to achieve this

* We know that for this particular series and for $z$ between $x$ and $0$ (or equivalently, between $0$ and $x$) the remainder is

\begin{align}
    R_k(x) = \frac{f^{(k+1)}(z)}{(k + 1)!} x^{k + 1}
\end{align}

* Either $f^{(k + 1)}(x) = \pm \sin{x}$ or $f^{(k + 1)}(x) = \pm \cos{x}$, all of which oscillate between $-1$ and $1$

* Noting that for a constant $a$, $\displaystyle \lim_{k \rightarrow \infty} \frac{a^k}{k!} = 0$, then for any specified value $x$ we have

\begin{align}
    \lim_{k \rightarrow \infty} | R_k | & = \lim_{k \rightarrow \infty} \left| \frac{f^{(k+1)}(z)}{(k + 1)!} x^{k + 1} \right| \\
    & = \lim_{k \rightarrow \infty} \frac{\left| f^{(k+1)}(z) \right|}{(k + 1)!} \left| x^{k + 1} \right| \\
    & \leq \lim_{k \rightarrow \infty} \frac{1}{(k + 1)!} \left| x^{k + 1} \right| \\
    & = \lim_{k \rightarrow \infty} \frac{\left| x^{k + 1} \right|}{(k + 1)!} \\
    & = 0
\end{align}

* Therefore, $\displaystyle \sin{x} = \sum_{k = 0}^\infty \frac{(-1)^k x^{2k + 1}}{(2k + 1)!}$ for all $x$

##### Steps for Finding Taylor Series

1. Differentiate $f$ several times and evaluate at $c$
    
    * Find $f(c), f'(c), f''(c), f'''(c), \dots$

2. Apply the formula for finding Taylor series and write in expanded form

3. Use the expanded form of the series from step 2 to identify sequential patterns in order to determine the general terms of the series and write in summation notation

4. Determine the interval of convergence for the series found

5. Within the interval of convergence found, determine whether the series converges to $f(x)$

##### Example 3

Find the Maclaurin series for the hyperbolic function $f(x) = \cosh{x}$.

##### Solution:

* Recall by definition $\cosh{x} = \frac{e^x + e^{-x}}{2}$ and $\sinh{x} = \frac{e^x - e^{-x}}{2}$

* Finding $f'(x)$:

\begin{align}
    f'(x) & = \frac{d}{dx}\left( \frac{e^x + e^{-x}}{2} \right) \\
    & = \frac{\frac{d}{dx} (e^x) + \frac{d}{dx} (e^{-x})}{2} \\
    & = \frac{e^x - e^{-x}}{2} \\
    & = \sinh{x}
\end{align}

* Finding $f''(x)$:

\begin{align}
    f''(x) & = \frac{d}{dx}\left( \frac{e^x - e^{-x}}{2} \right) \\
    & = \frac{\frac{d}{dx} (e^x) - \frac{d}{dx} (e^{-x})}{2} \\
    & = \frac{e^x - (-e^{-x})}{2} \\
    & = \frac{e^x + e^{-x}}{2} \\
    & = \cosh{x}
\end{align}

* Differentiating several times and evaluating at $c = 0$:

\begin{align}
    f(x) & = \cosh{x} \Rightarrow f(0) = 1 \\
    f'(x) & = \sinh{x} \Rightarrow f(0) = 0 \\
    f''(x) & = \cosh{x} \Rightarrow f(0) = 1 \\
    f'''(x) & = \sinh{x} \Rightarrow f(0) = 0 \\
    f^{(4)}(x) & = \cosh{x} \Rightarrow f(0) = 1 \\
    & \vdots \\
    f^{(k)}(x) & = \cdots \\
    & \vdots
\end{align}

* Writing the Taylor series in expanded form:

\begin{align}
    \sum_{k = 0}^\infty \frac{f^{(k)}(0)}{k!} x^k & = \frac{f(0)}{0!} + \frac{f'(0)}{1!} x + \frac{f''(0)}{2!} x^2 + \frac{f'''(0)}{3!} x^3 + \frac{f^{(4)}(0)}{4!} + \cdots \\
    & = 1 + 0 + \frac{x^2}{2!} + 0 + \frac{x^4}{4!} + \cdots
\end{align}

* Identifying the pattern:

    * We know that all odd order derivatives evaluate to zero and all even order derivatives evaluate to $1$
    
    * Therefore, only the *even* terms "survive"

* Therefore, we have the series

\begin{align}
    \sum_{k = 0}^\infty \frac{x^{2k}}{(2k)!}
\end{align}

* The interval of convergence is determined by the Ratio Test:

    * Let $u_k = \frac{x^{2k}}{(2k)!}$, then $u_{k+1} = \frac{x^{2k + 2}}{(2k + 2)!}$ and

\begin{align}
    \lim_{k \rightarrow \infty} \left| \frac{\frac{x^{2k + 2}}{(2k + 2)!}}{\frac{x^{2k}}{(2k)!}} \right| & = \lim_{k \rightarrow \infty} \left| \frac{x^{2k + 2}}{(2k + 2)!} \cdot \frac{(2k)!}{x^{2k}} \right| \\
    & = \lim_{k \rightarrow \infty} \frac{(2k)!}{(2k + 2)!} \cdot \frac{x^{2k + 2}}{x^{2k}} \\
    & = \lim_{k \rightarrow \infty} \frac{(2k)!}{(2k)! (2k + 2) (2k + 1)} \cdot x^2 \\
    & = \lim_{k \rightarrow \infty} \frac{1}{(2k + 2) (2k + 1)} \cdot x^2 \\
    & = x^2 \lim_{k \rightarrow \infty} \frac{1}{(2k + 2) (2k + 1)} \\
    & = x^2 \cdot 0 \\
    & = 0 < 1 \Rightarrow \text{converges for all $x$ by Ratio Test}
\end{align}

* Therefore, the interval of convergence is $(-\infty, \infty)$ and the radius of convergence is $R = \infty$

* We will omit the proof that $\displaystyle \cosh{x} = \sum_{k = 0}^\infty \frac{x^{2k}}{(2k)!}$, but this is true

##### Example 4

Find a power series for $f(x) = \sin{x^2}$.

##### Solution:

* From  the table of elementary functions we have

\begin{align}
    g(x) & = \sin{x} \\
    & = \sum_{k = 0}^\infty \frac{(-1)^k x^{2k + 1}}{(2k + 1)!}
\end{align}

* Using composition $f(x) = g(x^2)$:

\begin{align}
    f(x) & = \sin{x^2} \\
    & = \sum_{k = 0}^\infty \frac{(-1)^k \left( x^2 \right)^{2k + 1}}{(2k + 1)!} \\
    & = \sum_{k = 0}^\infty \frac{(-1)^k x^{4k + 2}}{(2k + 1)!}
\end{align}

* This power series has an interval of convergence $(-\infty, \infty)$ and radius of convergence $R = \infty$

##### Example 5

Find a power series for $\cosh{x}$ using power series of elementary functions.

##### Solution:

* This is an alternative solution to findning the power series found in example 3

* Using the definition $\cosh{x} = \frac{e^x + e^{-x}}{2}$

\begin{align}
    \cosh{x} & = \frac{e^x + e^{-x}}{2} \\
    & = \frac{1}{2} \left( e^x + e^{-x} \right) \\
\end{align}

* From the table of elementary functions

\begin{align}
    e^x & = \sum_{k = 0}^\infty \frac{x^k}{k!}
\end{align}

and by composition

\begin{align}
    e^{-x} & = \sum_{k = 0}^\infty \frac{(-x)^k}{k!}
\end{align}

* Therefore,

\begin{align}
    \cosh{x} & = \frac{e^x + e^{-x}}{2} \\
    & = \frac{1}{2} \left( e^x + e^{-x} \right) \\
    & = \frac{1}{2} \left( \sum_{k = 0}^\infty \frac{x^k}{k!} + \sum_{k = 0}^\infty \frac{(-x)^k}{k!}  \right)
\end{align}

* By properties of series we have

\begin{align}
    \cosh{x} & = \frac{1}{2} \left( \sum_{k = 0}^\infty \frac{x^k}{k!} + \sum_{k = 0}^\infty \frac{(-x)^k}{k!}  \right) \\
    & = \frac{1}{2} \sum_{k = 0}^\infty \left( \frac{x^k}{k!} + \frac{(-x)^k}{k!}  \right) \\
    & = \frac{1}{2} \sum_{k = 0}^\infty \left( \frac{x^k}{k!} + \frac{(-1)^k x^k}{k!}  \right) \\
    & = \frac{1}{2} \sum_{k = 0}^\infty \left( \frac{x^k + (-1)^k x^k}{k!}  \right)
\end{align}

* Expanding the series and recognizing a pattern:

\begin{align}
    \cosh{x} & = \frac{1}{2} \sum_{k = 0}^\infty \left( \frac{x^k + (-1)^k x^k}{k!}  \right) \\
    & = \frac{1}{2} \left( \frac{2}{0!} + \frac{x - x}{1!} + \frac{x^2 + x^2}{2!} + \frac{x^3 - x^3}{3!} + \frac{x^4 + x^4}{4!} + \cdots \right) \\
    & = \frac{1}{2} \left( 2 + 0 + \frac{2x^2}{2!} + 0 + \frac{2x^4}{4!} + \cdots \right) \\
    & = 1 + 0 + \frac{x^2}{2!} + 0 + \frac{x^4}{4!} + \cdots \\
    & = \sum_{k = 0}^\infty \frac{x^{2k}}{(2k)!}
\end{align}

* This is in agreement with example 3

##### Example 6

Find a power series for $f(x) = \frac{\cos{x}}{x^2}$.

##### Solution:

* Rewriting:

\begin{align}
    f(x) & = \frac{\cos{x}}{x^2} \\
    & = \frac{1}{x^2} \cos{x} \\
\end{align}

* From the table of elementary functions

\begin{align}
    \cos{x} & = \sum_{k = 0}^\infty \frac{(-1)^k x^{2k}}{(2k)!}
\end{align}

* Therefore,

\begin{align}
    f(x) & = \frac{1}{x^2} \cos{x} \\
    & = \frac{1}{x^2} \sum_{k = 0}^\infty \frac{(-1)^k x^{2k}}{(2k)!} \\
    & = x^{-2} \sum_{k = 0}^\infty \frac{(-1)^k x^{2k}}{(2k)!} \\
    & = \sum_{k = 0}^\infty x^{-2} \frac{(-1)^k x^{2k}}{(2k)!} \\
    & = \sum_{k = 0}^\infty \frac{(-1)^k x^{2k-2}}{(2k)!}
\end{align}

* The interval of convergence is $(-\infty, 0) \cup (0, \infty)$ (the function is undefined at $x = 0$) and has radius of convergence is $R = \infty$

##### Example 7

Use power series to approximate $\displaystyle \int_0^1 e^{-x^2} \, dx$ to within an error of less than $0.01$.

##### Solution:

* First we use composition with the power series representation of $e^{x}$ to obtain a power series for $e^{-x^2}$:

\begin{align}
    e^{-x^2} & = \sum_{k = 0}^\infty \frac{(-x^2)^k}{k!} \\
    & = \sum_{k = 0}^\infty \frac{(-1)^k (x^2)^k}{k!} \\
    & = \sum_{k = 0}^\infty \frac{(-1)^k x^{2k}}{k!} \\
\end{align}

* Therefore,

\begin{align}
    \int_0^1 e^{-x^2} \, dx & = \int_0^1 \sum_{k = 0}^\infty \frac{(-1)^k x^{2k}}{k!} \, dx \\
    & = \sum_{k = 0}^\infty \int_0^1 \frac{(-1)^k x^{2k}}{k!} \, dx \\
    & = \sum_{k = 0}^\infty \left. \left( \frac{(-1)^k x^{2k + 1}}{k! \cdot (2k + 1)} \right) \right|_0^1 \\
    & = \sum_{k = 0}^\infty \left( \frac{(-1)^k (1)^{2k + 1}}{k! \cdot (2k + 1)} \right) - \left( \frac{(-1)^k (0)^{2k + 1}}{k! \cdot (2k + 1)} \right) \\
    & = \sum_{k = 0}^\infty \frac{(-1)^k}{k! \cdot (2k + 1)}
\end{align}

* This series is an alternating series, which has remainder $|S - S_N| = |R_N| \leq a_{N + 1}$

* We require that $|R_N| \leq 0.01$ and therefore we need to know the smallest number $N$ such that $|R_N| = a_{N + 1} < 0.01$

* Finding $a_{N + 1}$:

\begin{align}
    a_{N + 1} & = \frac{1}{(N + 1)! \cdot (2(N + 1) + 1)} \\
    & = \frac{1}{(N + 1)! \cdot (2N + 3)}
\end{align}

* Solving for the minimum $N$:

\begin{align}
    a_{N + 1} & < 0.01 \\
    \frac{1}{(N + 1)! \cdot (2N + 3)} & < 0.01 \\
    \frac{1}{(N + 1)! \cdot (2N + 3)} \cdot (N + 1)! \cdot (2N + 3) & < 0.01 \cdot (N + 1)! \cdot (2N + 3) \\
    1 & < 0.01 \cdot (N + 1)! \cdot (2N + 3) \\
    \frac{1}{0.01} & < \frac{0.01 \cdot (N + 1)! \cdot (2N + 3)}{0.01} \\
    100 & < (N + 1)! \cdot (2N + 3) \\
    (N + 1)! \cdot (2N + 3)  & > 100
\end{align}

* The lowest value for which this inequality is true is $N = 3$

* Therefore,

\begin{align}
    \int_0^1 e^{-x^2} \, dx & \approx \sum_{k = 0}^3 \frac{(-1)^k}{k! \cdot (2k + 1)} \\
    & = \frac{(-1)^0}{0! \cdot (2 \cdot 0 + 1)} + \frac{(-1)^1}{1! \cdot (2 \cdot 1 + 1)} + \frac{(-1)^2}{2! \cdot (2 \cdot 2 + 1)} + \frac{(-1)^3}{3! \cdot (2 \cdot 3 + 1)} \\
    & = 1 - \frac{1}{3} + \frac{1}{10} - \frac{1}{42} \\
    & \approx 0.742867
\end{align}
