##### Objective 1: Understand the definition of a convergent infinite series

##### Infinite Series

* Let $\{a_n\}$ be an infinite sequence. The **infinite series** associated with $\{a_n\}$ is the summation

\begin{align}
    \sum_{n = 1}^\infty a_n & = a_1 + a_2 + a_3 + \cdots + a_n + \cdots.
\end{align}

* When the starting index is known (in this case $n$ is the index of summation) the summation is often written $\sum a_n$

* The numbers $a_1, a_2, a_3, \dots$ are called the **terms** of the series

* The **$N$th partial sum** of an infinite series is defined by

\begin{align}
    S_N & = \sum_{n = 1}^N a_n \\
    S_N & = a_1 + a_2 + a_3 + \cdots + a_N
\end{align}

* If the *sequence of partial sums converges*, then the (infinite) series **converges**

    * i.e. when $\{S_n\} = \{ S_1, S_2, S_3, \dots \}$ converges, then $\displaystyle S = \sum_{n = 1}^\infty a_n$ converges
    
    * Note that
    
\begin{align}
    S_1 & = a_1 \\
    S_2 & = a_1 + a_2 \\
    S_3 & = a_1 + a_2 + a_3 \\
    & \vdots \\
    S_n & = a_1 + a_2 + a_3 + \cdots + a_n
\end{align}

* If $\displaystyle \lim_{n \rightarrow \infty} S_n = S$, where $S$ is a real number, then $\displaystyle S = \sum_{n = 1}^\infty a_n$

* If $\{S_n\}$ diverges, then the series **diverges**

##### Working with Infinite Series

Generally we are concerned with two important questions regarding series:

1. Does a series converge or diverge?

2. If a series converges, what does the series converge to?

##### Example 1

Determine if the series $\displaystyle \sum_{n = 1}^\infty \cfrac{1}{2^n} = \cfrac{1}{2} + \cfrac{1}{4} + \cfrac{1}{8} + \cfrac{1}{16} + \cdots$ converges or diverges.

##### Solution:

* The partial sums of the series in question are given below:

\begin{align}
    S_1 & = \cfrac{1}{2} \\
    S_2 & = \cfrac{1}{2} + \cfrac{1}{4} = \cfrac{3}{4} \\
    S_3 & = \cfrac{1}{2} + \cfrac{1}{4} + \cfrac{1}{8} = \cfrac{7}{8} \\
    & \vdots \\
    S_n & = \cfrac{1}{2} + \cfrac{1}{4} + \cfrac{1}{8} + \cdots + \cfrac{1}{2^n} = \cfrac{2^n - 1}{2^n}
\end{align}

* We need to take the limit of the general term:

\begin{align}
    \lim_{n \rightarrow \infty} S_n & = \lim_{n \rightarrow \infty} \cfrac{2^n - 1}{2^n} \\
    & = \boxed{1}
\end{align}

* Since the limit of the sequence of partial sums converges, the series converges and in this case the series converges to $1$

##### Example 2

Determine if the series $\displaystyle \sum_{n = 1}^\infty \left( \cfrac{1}{n} - \cfrac{1}{n + 1} \right)$ converges or diverges.

##### Solution:

* First we write out the general form of the series term by term:

\begin{align}
    \sum_{n = 1}^\infty \left( \cfrac{1}{n} - \cfrac{1}{n + 1} \right) & = \left( \cfrac{1}{1} - \cfrac{1}{2} \right) + \left( \cfrac{1}{2} - \cfrac{1}{3} \right) + \left( \cfrac{1}{3} - \cfrac{1}{4} \right) \\
    & \qquad \qquad + \cdots + \left( \cfrac{1}{n} - \cfrac{1}{n + 1} \right) + \cdots
\end{align}

\begin{align}
    \require{cancel}
    \Rightarrow S_n = \sum_{n = 1}^\infty \left( \cfrac{1}{n} - \cfrac{1}{n + 1} \right) & = \left( \cfrac{1}{1} - \cancel{\cfrac{1}{2}} \right) + \left( \cancel{\cfrac{1}{2}} - \cancel{\cfrac{1}{3}} \right) + \left( \cancel{\cfrac{1}{3}} - \cancel{\cfrac{1}{4}} \right) \\
    & \qquad \qquad + \cdots + \left( \cancel{\cfrac{1}{n}} - \cfrac{1}{n + 1} \right) \\
    S_n & = 1 - \cfrac{1}{n + 1}
\end{align}

* Taking the limit:

\begin{align}
    \lim_{n \rightarrow \infty} S_n & = \lim_{n \rightarrow \infty} \left( 1 - \cfrac{1}{n + 1} \right)  \\
    & = \boxed{1}
\end{align}

* Since the limit of the sequence of partial sums converges to $1$, the series converges and in this case the sum is $1$

##### Telescoping Series

* The series in the previous example is called a telescoping series (the terms collapse like an old school telescope)

* A **telescoping series** is a series of the form $(a_1 - a_2) + (a_2 - a_3) + (a_3 - a_4) + (a_4 - a_5) + \cdots$

##### Example 3

Determine if the series $\displaystyle \sum_{n = 1}^{\infty} 1$ converges or diverges.

##### Solution:

* Since $\displaystyle \sum_{n = 1}^\infty 1 = 1 + 1 + 1 + 1 + \cdots$ the $n$th partial sum can be expressed as $S_n = n$

* Taking the limit of the general term for this sequence of partial sums:

\begin{align}
    \lim_{n \rightarrow \infty} S_n & = \lim_{n \rightarrow \infty} n \\
    & = \infty
\end{align}

* Since the limit does not exist, the series $\boxed{\text{diverges}}$

##### Example 2

Find the sum of the series $\displaystyle \sum_{n = 1}^\infty \cfrac{2}{4n^2 - 1}$.

##### Solution:

* First, we factor the denominator:

\begin{align}
    \sum_{n = 1}^\infty \cfrac{2}{4n^2 - 1} & = \sum_{n = 1}^\infty \cfrac{2}{(2n + 1)(2n - 1)}
\end{align}

* Next, we use partial fractions to decompose the fraction in the summation:

\begin{align}
    \cfrac{2}{(2n + 1)(2n - 1)} & = \cfrac{A}{2n + 1} + \cfrac{B}{2n - 1}
\end{align}

* Multiplying both sides by $(2n + 1)(2n - 1)$:

\begin{align}
    2 & = A(2n - 1) + B(2n + 1) \\
    2 & = 2nA + 2nB - A + B \\
    2 & = 2(A + B)n + (-A + B)
\end{align}

* Comparing coefficients:

\begin{align}
    -A + B & = 2 \\
    2(A + B) & = 0
\end{align}

* Solving:

\begin{align}
    A & = -B \quad \text{(from the second equation)} \\
    \Rightarrow -A + (-A) & = 2 \\
    \Rightarrow  -2A & = 2 \\
    \Rightarrow A & = -1 \\
    \Rightarrow B & = 1
\end{align}

* Therefore, the series can be represented as

\begin{align}
    \sum_{n = 1}^\infty \cfrac{2}{4n^2 - 1} & = \sum_{n = 1}^\infty \left( \cfrac{-1}{2n + 1} + \cfrac{1}{2n - 1} \right) \\
    & = \sum_{n = 1}^\infty \left( \cfrac{1}{2n - 1} - \cfrac{1}{2n + 1} \right)
\end{align}

* Expanding this series we get

\begin{align}
    \sum_{n = 1}^\infty \cfrac{2}{4n^2 - 1} & = \sum_{n = 1}^\infty \left( \cfrac{1}{2n - 1} - \cfrac{1}{2n + 1} \right) \\
    & = \left( \cfrac{1}{1} - \cfrac{1}{3} \right) + \left( \cfrac{1}{3} - \cfrac{1}{5} \right)
    + \left( \cfrac{1}{5} - \cfrac{1}{7} \right) \\
    & \qquad \qquad + \left( \cfrac{1}{7} - \cfrac{1}{9} \right) + \cdots + \left( \cfrac{1}{2n - 1} - \cfrac{1}{2n + 1} \right) + \cdots
\end{align}

* The $n$th partial sum is then determined by

\begin{align}
    S_n & = \left( \cfrac{1}{1} - \cancel{\cfrac{1}{3}} \right) + \left( \cancel{\cfrac{1}{3}} - \cancel{\cfrac{1}{5}} \right)
    + \left( \cancel{\cfrac{1}{5}} - \cancel{\cfrac{1}{7}} \right) \\
    & \qquad \qquad + \left( \cancel{\cfrac{1}{7}} - \cancel{\cfrac{1}{9}} \right) + \cdots + \left( \cancel{\cfrac{1}{2n - 1}} - \cfrac{1}{2n + 1} \right) \\
    S_n & = 1 - \cfrac{1}{2n + 1}
\end{align}

* Taking the limit:

\begin{align}
    \lim_{n \rightarrow \infty} S_n & = \lim_{n \rightarrow \infty} \left( 1 - \cfrac{1}{2n + 1} \right) \\
    & = \boxed{1}
\end{align}

* So the series converges to $1$

##### Objective 2: Use properties of infinite geometric series

##### Geometric Series

* A **geometric series** with ratio $r \neq 0$ and initial term $a \neq 0$ is a series of the form

\begin{align}
    \displaystyle \sum_{n = 0}^\infty a r^n = a + ar + ar^2 + \cdots + ar^n + \cdots
\end{align}

##### Theorem 9.6: Convergence of a Geometric Series

* A geometric series with ratio $r$ diverges when $|r| \geq 1$. If $|r| < 1$, then the series converges to the sum

\begin{align}
    \sum_{n = 0}^\infty a r^n & = \cfrac{a}{1 - r}, \quad |r| < 1.
\end{align}

##### Proof:

When $r = \pm 1$ we have $\displaystyle \sum_{n = 0}^\infty a (\pm 1)^n$, which is a series that either oscillates between $-a$ and $a$ or diverges to $\infty$. When $r \neq \pm 1$, then

\begin{align}
    S_n & = a + ar + ar^2 + \cdots + ar^{n - 1}.
\end{align}

Multiplying by $r$ gives us

\begin{align}
    r S_n & = ar + ar^2 + ar^3 + \cdots + ar^n.
\end{align}

Subtracting the second equation from the first:

\begin{align}
    S_n - rS_n & = (a - ar) + (ar - ar^2) + (ar^2 - ar^3) \\
    & \qquad \qquad + \cdots + (ar^{n-1} - ar^n) \\
    \Rightarrow S_n - rS_n & = a - ar^n
\end{align}

Factoring:

\begin{align}
    S_n (1 - r) & = a(1 - r^n) \\
    S_n& = \cfrac{a(1 - r^n)}{1 - r} \\
\end{align}

Taking the limit:

\begin{align}
    S & = \lim_{n \rightarrow \infty} S_n \\
    & = \lim_{n \rightarrow \infty} \cfrac{a(1 - r^n)}{1 - r}
\end{align}

If $|r| > 1$, then the sequence of partial sums diverges ($|r|^n \rightarrow \infty$ as $n \rightarrow \infty$). If $|r| < 1$, then the sequence of partial sums converges to $S = \cfrac{a}{1 - r}$ ($r^n \rightarrow 0$ as $r \rightarrow \infty$) $\blacksquare$

##### Example 3

Determine whether the series $\displaystyle \sum_{n = 0}^\infty \cfrac{3}{2^n}$ converges or diverges.

##### Solution:

* Rewriting:

\begin{align}
    \sum_{n = 0}^\infty \cfrac{3}{2^n} & = \sum_{n = 0}^\infty 3 \left( \cfrac{1}{2} \right)^n
\end{align}

* So we have a geometric series with $a = 3$ and $r = 1/2 < 1 \Rightarrow$ the geometric series converges

* Applying theorem 9.6:

\begin{align}
    S & = \cfrac{a}{1 - r}\\
    & = \cfrac{3}{1 - 1/2} \\
    & = \cfrac{3}{1/2} \\
    & = \boxed{6}
\end{align}

* So the geometric series converges to $6$

##### Example 5

Use geometric series to show that $1 = 0.\bar{9}$

##### Solution:

* Rewriting the repeating decimal $0.\bar{9}$:

\begin{align}
    0.\bar{9} & = 0.9 + 0.09 + 0.009 + 0.0009 + \cdots \\
    & = \cfrac{9}{10} + \cfrac{9}{100} + \cfrac{9}{1000} + \cdots \\
    & = 9 \left( \cfrac{1}{10} + \cfrac{1}{100} + \cfrac{1}{1000} + \cdots \right) \\
    & = 9 \left( \cfrac{1}{10} + \cfrac{1}{10^2} + \cfrac{1}{10^3} + \cdots \right) \\
    & = \cfrac{9}{10} + \cfrac{9}{10} \left( \cfrac{1}{10} \right)^1 + \cfrac{9}{10} \left( \cfrac{1}{10} \right)^2 + \cdots \\
    & = \sum_{n = 0}^\infty \cfrac{9}{10} \left( \cfrac{1}{10} \right)^n
\end{align}

* So $0.\bar{9}$ is a geometric series with $a = 9/10$ and $r = 1/10 < 1 \Rightarrow$ the geometric series converges

* We apply the formula $S = \cfrac{a}{1 - r}$:

\begin{align}
    S & = \cfrac{a}{1 - r} \\
    & = \cfrac{9/10}{1 - 1/10} \\
    & = \cfrac{9/10}{9/10} \\
    & = \boxed{1}
\end{align}

* Therefore, we can conclude that $0.\bar{9} = 1$ since the geometric series representing the repeating decimal converges to $1$

##### Properties of Infinite Series

Let $\sum a_n$ and $\sum b_n$ be convergent series, and let $A, B,$ and $c$ be real numbers. If $\sum a_n = A$ and $\sum b_n = B$, then the following series converge to the indicated sums.

1. $\displaystyle \sum_{n = 1}^\infty c a_n = c A$

2. $\displaystyle \sum_{n = 1}^\infty (a_n \pm b_n) = A \pm B$

##### Objective 3: Use the $n$th-Term Test for Divergence of an infinite series

##### Theorem 9.8: Limit of the $n$th Term of a Convergent Series

* If $\displaystyle \sum_{n = 1}^\infty a_n$ converges, then $\displaystyle \lim_{n \rightarrow \infty} a_n = 0$.

##### Theorem 9.9: $n$th-Term Test for Divergence

* If $\displaystyle \lim_{n \rightarrow \infty} a_n \neq 0$, then $\displaystyle \sum_{n = 1}^\infty a_n$ diverges.

##### Example 6

Determine whether the sereies $\displaystyle \sum_{n = 0}^\infty 2^n$ converges or diverges.

##### Solution:

* We take the limit of $a_n = 2^n$:

\begin{align}
    \lim_{n \rightarrow \infty} 2^n & = \infty \neq 0
\end{align}

* Since $\displaystyle \lim_{n \rightarrow \infty} a_n \neq 0$ the series $\boxed{\text{diverges.}}$

##### Example 7

Determine whether the series $\displaystyle \sum_{n = 1}^\infty \cfrac{n!}{2n! + 1}$ converges or diverges.

##### Solution:

* We take the limit of $a_n = \cfrac{n!}{2n! + 1}$:

\begin{align}
    \lim_{n \rightarrow \infty} \cfrac{n!}{2n! + 1} & = \lim_{n \rightarrow \infty} \cfrac{n!}{2n! + 1} \cdot \cfrac{1/n!}{1/n!} \\
    & = \lim_{n \rightarrow \infty} \cfrac{1}{2 + 1/n!} \\
    & = \cfrac{1}{2}
\end{align}

* Since $\displaystyle \lim_{n \rightarrow \infty} a_n \neq 0$ the sereies $\boxed{\text{diverges.}}$

##### Example 8

Determine whether the series $\displaystyle \sum_{n = 1}^\infty \cfrac{1}{n}$ converges or diverges.

##### Solution:

* We take the limit of $a_n = \cfrac{1}{n}$:

\begin{align}
    \lim_{n \rightarrow \infty} \cfrac{1}{n} & = 0
\end{align}

* Since $\lim_{n \rightarrow \infty} a_n = 0$ *we cannot draw any conclusion* about the convergence or divergence of the series

* The best we can do based on this information alone is say that the series *might* converge

* This is what is known as the harmonic series and is a very famous *divergent* series, but we do not yet have the required tools to determine this fact
