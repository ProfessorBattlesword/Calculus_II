##### Objective 1: Use the Integral Test to determine whether an infinite series converges or diverges

##### Theorem 9.10: The Integral Test

If $f$ is positive, continuous, and decreasing for $x \geq N \geq 1$ and $a_n = f(n)$, then

\begin{align}
    \sum_{n = N}^\infty a_n
\end{align}

and

\begin{align}
    \int_N^\infty f(x) \, dx
\end{align}

either both converge or both diverge.

**Note:** In most cases, $N = 1$, but the above theorem applies for any starting index satisfying $N \geq 1$

##### Example 1

Use the Integral Test to determine the convergence or divergence of the series $\displaystyle \sum_{n = 1}^\infty \cfrac{n}{n^2 + 1}$.

##### Solution:

* We let $f(x) = \cfrac{x}{x^2 + 1}$ and check the necessary criteria for the Integral Test:

    * $f(x)$ is *positive* over the interval $[1, \infty)$
    
    * $f(x)$ is *continous* over the interval $[1, \infty)$
    
    * $f'(x) = \cfrac{1 - x^2}{(x^2 + 1)^2} < 0$ (i.e. $f(x)$ is decreasing on its entire domain)

* Now integrate the improper integral:

\begin{align}
    \int_1^\infty \cfrac{x}{x^2 + 1} \, dx
\end{align}

* To integrate, we perform a $u$ substitution by letting $u = x^2 + 1$. Then $du = 2x \, dx \Rightarrow \cfrac{1}{2} \, du = x \, dx$.
    
    * Addressing the limits of integration:
    
        * If $x = 1$, then $u = 2$
        
        * If $x \rightarrow \infty$, then $u \rightarrow \infty$

* So we get

\begin{align}
    \int_1^\infty \cfrac{x}{x^2 + 1} \, dx & = \cfrac{1}{2} \int_2^\infty \cfrac{1}{u} \, du \\
    & = \cfrac{1}{2} \lim_{b \rightarrow \infty} \int_2^b \cfrac{1}{u} \, du \\
    & = \cfrac{1}{2} \lim_{b \rightarrow \infty} \left. \ln{|u|} \right|_2^b \\
    & = \cfrac{1}{2} \lim_{b \rightarrow \infty} (\ln{|b|} - \ln{2}) \\
    & = \infty
\end{align}

* Therefore, since the integral diverges, so too does the series $\displaystyle \sum_{n = 1}^\infty \cfrac{n}{n^2 + 1}$

##### Example 2

Use the Integral Test to determine the convergence or divergence of the series $\displaystyle \sum_{n = 1}^\infty \cfrac{1}{n^2 + 1}$.

##### Solution:

* We let $f(x) = \cfrac{1}{x^2 + 1}$ and check the necessary criteria for the Integral Test:

    * $f(x)$ is *positive* over the interval $[1, \infty)$
    
    * $f(x)$ is *continous* over the interval $[1, \infty)$
    
    * $f'(x) = \cfrac{-2x}{(x^2 + 1)^2} < 0$ (i.e. $f(x)$ is decreasing on the interval $(1, \infty)$)

* Now integrate the improper integral:

\begin{align}
    \int_1^\infty \cfrac{1}{x^2 + 1} \, dx
\end{align}

* We have an identity for this integral:

\begin{align}
    \int_1^\infty \cfrac{1}{x^2 + 1} \, dx & = \lim_{b \rightarrow \infty} \int_1^\infty \cfrac{1}{x^2 + 1} \, dx \\
    & = \lim_{b \rightarrow \infty} \left. \tan^{-1}{x} \right|_1^b \\
    & = \lim_{b \rightarrow \infty} \left( \tan^{-1}{b} - \tan^{-1}{1} \right) \\
    & = \cfrac{\pi}{2} - \cfrac{\pi}{4} \\
    & = \cfrac{\pi}{4}
\end{align}

* Since the integral converges, so too does the series $\displaystyle \sum_{n = 1}^\infty \cfrac{1}{n^2 + 1}$

**Important Note:** The sereies $\displaystyle \sum_{n = 1}^\infty \frac{1}{n^2 + 1}$ does *NOT* converge to $\frac{\pi}{4}$! This is what the integral converges to, but the series does not necessarily converge the same value as the improper integral! The only information we know for sure is that the series also converges to some finite value.

##### Objective 2: Use properties of $p$-series and harmonic series

##### $p$-Series

* For a positive constant $p$, a **$p$-series** is a series of the form

\begin{align}
    \sum_{n = 1}^\infty \cfrac{1}{n^p} & = \cfrac{1}{1^p} + \cfrac{1}{2^p} + \cfrac{1}{3^p} + \cdots.
\end{align}

* If $p = 1$, then we have what we call the **harmonic series**

\begin{align}
    \sum_{n = 1}^\infty \cfrac{1}{n}
\end{align}

##### Theorem 9.11: Convergence of $p$-Series

The $p$-series 

\begin{align}
    \sum_{n = 1}^\infty \cfrac{1}{n^p} & = \cfrac{1}{1^p} + \cfrac{1}{2^p} + \cfrac{1}{3^p} + \cdots
\end{align}

converges for $p > 1$ and diverges for $0 < p \leq 1$.

##### Proof:

This follows immediately by applying the Integral Test to Theorem 8.7.

Let $f(x) = \cfrac{1}{x^p}$. Theorem 8.7 states that if $p$ is a real number, then

\begin{align}
    \int_1^\infty \cfrac{1}{x^p} \, dx =\begin{cases} \cfrac{1}{p - 1}, \; p > 1 \\ \text{diverges,} \; p \leq 1 \end{cases}.
\end{align}

By the Integral Test, a $p$-series converges if $p > 1$ and a $p$-series diverges if $0 < p \leq 1$. $\blacksquare$

##### Example 3

Determine the convergence or divergence of the harmonic series.

##### Solution:

* Since the harmonic series is given by $\displaystyle \sum_{n = 1}^\infty \cfrac{1}{n}$, then we have a $p$-series with $p = 1$ and by theorem 9.11 the series $\boxed{\text{diverges.}}$

##### Example 4

Determine the convergence or divergence of the series $\displaystyle \sum_{n = 1}^\infty \cfrac{1}{n^2}$.

##### Solution:

* Since this is a $p$-series with $p = 2 > 1$, the series $\boxed{\text{converges.}}$
