##### Theorem 9.12: Direct Comparison Test

Let $0 < a_n \leq b_n$ for all $n$.

1. If $\displaystyle \sum_{n = 1}^\infty b_n$ converges, then $\displaystyle \sum_{n = 1}^\infty a_n$ converges.

2. If $\displaystyle \sum_{n = 1}^\infty a_n$ diverges, then $\displaystyle \sum_{n = 1}^\infty b_n$ diverges.

##### Example 1

Determine if the series $\displaystyle \sum_{n = 1}^\infty \cfrac{1}{n + 1}$ converges or diverges by the Direct Comparison Test.

##### Solution:

* The first question to ask is what do we *think* the series will do based on behavior of previous series we've seen

* This series looks quite similar to the harmonic series: $\displaystyle \sum_{n = 1}^\infty \cfrac{1}{n}$

* So we might be a bit sus that the series diverges (yes, that's an Among Us reference)

* Since we're sus that the series diverges, we need to determine a general term for a second series that we *know* diverges with general term that is *smaller* for all $n \geq N$ for some positive integer $N$

* With this in mind, let $b_n = \cfrac{1}{n + 1}$ (the general term of our series above) and $a_n = \cfrac{1}{2n}$
    
    * Since we know that the harmonic series diverges, then $\displaystyle \cfrac{1}{2} \sum_{n = 1}^\infty \cfrac{1}{n} = \sum_{n = 1}^\infty \cfrac{1}{2n} = \sum_{n = 1}^\infty a_n$ also diverges
    
    * Furthermore, $a_n \leq b_n$, for $n \geq 1$
    
    * Therefore, by the Direct Comparison Test, $\displaystyle \sum_{n = 1}^\infty b_n = \sum_{n = 1}^\infty \cfrac{1}{n + 1}$ diverges

##### Example 2

Determine if the series $\displaystyle \sum_{n = 1}^\infty \cfrac{1}{2^n + 3}$ converges or diverges by the Direct Comparison Test.

##### Solution:

* Again, we ask ourselves what do we *think* the series will do based on behavior of previous series we've seen

* This series looks quite similar to the geometric series: $\displaystyle \sum_{n = 1}^\infty \cfrac{1}{2^n}$

* So we might sus that the series converges

* Since we suspect that the series converges, we need to determine a general term for a second series that we *know* converges with general term that is *larger* for all $n \geq N$ for some positive integer $N$

* With this in mind, let $a_n = \cfrac{1}{2^n + 3}$ (the general term of our series above) and $b_n = \left(\cfrac{1}{2}\right)^n = \cfrac{1}{2^n}$
    
    * We know that the $b_n$ describes a convergent geometric series $(|r| < 1)$ with $a = \cfrac{1}{2}$ and $r = \cfrac{1}{2}$
    
    * Furthermore, $a_n \leq b_n$, for $n \geq 1$
    
    * Therefore, by the Direct Comparison Test, $\displaystyle \sum_{n = 1}^\infty a_n = \sum_{n = 1}^\infty \cfrac{1}{2^n + 3}$ converges

##### Example 3

Determine if the the series $\displaystyle \sum_{n = 1}^\infty \cfrac{1}{2 + \sqrt{n}}$ converges or diverges.

##### Solution:

* What do we *think* the series will do based on behavior of previous series we've seen?

* If we teak the series a bit, the series looks somewhat like a $p$-sereies: $\displaystyle \sum_{n = 1}^\infty \cfrac{1}{2 + n^{1/2}}$

* So we might suspect that the series diverges since it looks similar to $\displaystyle \sum_{n = 1}^\infty \cfrac{1}{n^{1/2}}$, which is a divergent $p$-series with $p = \cfrac{1}{2}$ ($p$-series diverges when $p \leq 1$)

* Since we suspect that the series diverges, we need to determine a general term for a second series that we *know* diverges with general term that is *smaller* for all $n \geq N$ for some positive integer $N$

* With this in mind, let $a_n = \cfrac{1}{n}$ and $b_n  = \cfrac{1}{2 + n^{1/2}}$
    
    * We know that $a_n$ is the harmonic series, which we know diverges
    
    * Furthermore, $a_n \leq b_n$, for $n \geq 4$
    
    * Therefore, by the Direct Comparison Test, $\displaystyle \sum_{n = 1}^\infty b_n = \sum_{n = 1}^\infty \cfrac{1}{2 + \sqrt{n}}$ diverges

##### Example 4

Determine if the series $\displaystyle \sum_{n = 1}^\infty \cfrac{1}{n^2 - \frac{1}{2}}$ converges or diverges.

##### Solution:

* What do we *think* the series will do based on behavior of previous series we've seen?

* We might suspect that the series converges since it looks similar to $\displaystyle \sum_{n = 1}^\infty \cfrac{1}{n^2}$, which is a convergent $p$-series with $p = 2$ ($p$-series converges when $p > 1$)

* Since we suspect that the series converges, we need to determine a general term for a second series that we *know* converges with general term that is *larger* for all $n \geq N$ for some positive integer $N$

* With this in mind, let $a_n = \cfrac{1}{n^2 - \frac{1}{2}}$ and $b_n  = \cfrac{2}{n^2}$
    
    * We know that $b_n$ is a convergent $p$-series
    
    * Furthermore, $a_n \leq b_n$, for $n \geq 2$
    
    * Therefore, by the Direct Comparison Test, $\displaystyle \sum_{n = 1}^\infty a_n = \sum_{n = 1}^\infty \cfrac{1}{n^2 - \frac{1}{2}}$ converges

##### Objective 2: Use the Limit Comparison Test to determine whether a series converges or diverges

##### Theorem 9.13: Limit Comparison Test

If $a_n > 0, b_n > 0$, and $\displaystyle \lim_{n \rightarrow \infty} \cfrac{a_n}{b_n} = L$ where $L$ is a *finite* and *positive* real number, then $\displaystyle \sum_{n = 1}^\infty a_n$ and $\displaystyle \sum_{n = 1}^\infty b_n$ either *both converge* or *both diverge.*

##### Example 5

Use the limit comparison test to show that the sum $\displaystyle \sum_{n = 1}^\infty \cfrac{1}{2 + \sqrt{n}}$ diverges.

##### Solution:

* Once again, we first want to ask ourselves what do we *think* the series will do: converge or diverge?

* We already know from Example 3 that the series diverges so we need to look for a general term that we know diverges, but this time we don't need to compare the two general terms

* All that matters is that when we take the limit of the ratio of the two general terms, we need to end up with a positive, finite value

* So we let $a_n = \cfrac{1}{2 + \sqrt{n}}$ and $b_n = \cfrac{1}{\sqrt{n}}$

    * Then $a_n > 0$ and $b_n > 0$ as required in Theorem 9.13

* Taking the limit:

\begin{align}
    \lim_{n \rightarrow \infty} \cfrac{a_n}{b_n} & = \lim_{n \rightarrow \infty} \cfrac{\frac{1}{2 + \sqrt{n}}}{\frac{1}{\sqrt{n}}} \\
    & = \lim_{n \rightarrow \infty} \cfrac{1}{2 + \sqrt{n}} \cdot \cfrac{\sqrt{n}}{1} \\
    & = \lim_{n \rightarrow \infty} \cfrac{\sqrt{n}}{2 + \sqrt{n}}
\end{align}

* Now we could use either L'&ocirc;pital's Rule or multiply top and bottom by $1/\sqrt{n}$

* Multiplying top and bottom by $1/\sqrt{n}$:

\begin{align}
    \lim_{n \rightarrow \infty} \cfrac{a_n}{b_n} & = \lim_{n \rightarrow \infty} \cfrac{\sqrt{n}}{2 + \sqrt{n}} \cdot \cfrac{\frac{1}{\sqrt{n}}}{\frac{1}{\sqrt{n}}} \\
    & = \lim_{n \rightarrow \infty} \cfrac{1}{\frac{2}{\sqrt{n}} + 1} \\
    & = 1 = L > 0
\end{align}

* Since $L = 1 > 0$ is positive and finite, both $\sum a_n$ and $\sum b_n$ behave the same in terms of convergence and therefore, since $b_n$ diverges, by the Limit Comparison Test the series $\displaystyle \sum_{n = 1}^\infty \cfrac{1}{2 + \sqrt{n}}$ also diverges

##### Example 6

Show that the **general harmonic series** defined as $\displaystyle \sum_{n = 1}^\infty \cfrac{1}{an + b}$, $a > 0, b > 0$, diverges.

##### Solution:

* Since the series resembles the harmonic series and we know the harmonic series diverges, we might try comparing the given series with the harmonic series

* So we let $a_n = \cfrac{1}{an + b}$ and $b_n = \cfrac{1}{n}$

    * Then $a_n > 0$ and $b_n > 0$ as required

* Taking the limit of the ratio of the two general terms:

\begin{align}
    \lim_{n \rightarrow \infty} \cfrac{a_n}{b_n} & = \lim_{n \rightarrow \infty} \cfrac{\frac{1}{an + b}}{\frac{1}{n}} \\
    & = \lim_{n \rightarrow \infty} \cfrac{1}{an + b} \cdot \cfrac{n}{1} \\
    & = \lim_{n \rightarrow \infty} \cfrac{n}{an + b} \\
    & = \lim_{n \rightarrow \infty} \cfrac{n}{an + b} \cdot \cfrac{\frac{1}{n}}{\frac{1}{n}} \\
    & = \lim_{n \rightarrow \infty} \cfrac{1}{a + b/n} \\
    & = \cfrac{1}{a} = L > 0 \\
\end{align}

* Since $L = 1/a > 0$ is positive and finite, both series diverge, and therefore, the general harmonic series diverges by the Limit Comparison Test

##### Example 7

Determine if the series $\displaystyle \sum_{n = 1}^\infty \cfrac{1}{5n^2 - 5n + 1}$ converges or diverges.

##### Solution:

* Since the denominator is a polynomial of order $2$, we might compare this series with a convergent $p$-series with $p = 2$

* So we let $a_n = \cfrac{1}{5n^2 - 5n + 1}$ and $b_n = \cfrac{1}{n^2}$

    * Then $a_n > 0$ and $b_n > 0$ as required

* Taking the limit of the ratio of the two general terms:

\begin{align}
    \lim_{n \rightarrow \infty} \cfrac{a_n}{b_n} & = \lim_{n \rightarrow \infty} \cfrac{\frac{1}{5n^2 - 5n + 1}}{\frac{1}{n^2}} \\
    & = \lim_{n \rightarrow \infty} \cfrac{1}{5n^2 - 5n + 1} \cdot \cfrac{n^2}{1} \\
    & = \lim_{n \rightarrow \infty} \cfrac{n^2}{5n^2 - 5n + 1} \\
    & = \lim_{n \rightarrow \infty} \cfrac{n^2}{5n^2 - 5n + 1} \cdot \cfrac{\frac{1}{n^2}}{\frac{1}{n^2}} \\
    & = \lim_{n \rightarrow \infty} \cfrac{1}{5 - 5/n + 1/n^2} \\
    & = \cfrac{1}{5} = L > 0 \\
\end{align}

* Since $L = 1/5 > 0$ is positive and finite, both series converge, and therefore, the series $\displaystyle \sum_{n = 1}^\infty \cfrac{1}{5n^2 - 5n + 1}$ converges by the Limit Comparison Test

##### Example 8

Determine if the series $\displaystyle \sum_{n = 1}^\infty \cfrac{\sqrt{n}}{n^2 + 1}$ converges or diverges.

##### Solution:

* The series might not immediately have an obvious choice for comparison

* Rewriting:

\begin{align}
    \sum_{n = 1}^\infty \cfrac{\sqrt{n}}{n^2 + 1} & = \sum_{n = 1}^\infty \cfrac{n^{1/2}}{n^2 + 1}
\end{align}

* If we look at the ratio of only the leading powers in the numerator and denominator we get $\cfrac{n^{1/2}}{n^2} = \cfrac{1}{n^{3/2}}$

* So we might consider comparing the original series with a convergent $p$-series with $p = 3/2 > 1$

* So we let $a_n = \cfrac{\sqrt{n}}{n^2 + 1}$ and $b_n = \cfrac{1}{n^{3/2}}$

    * Then $a_n > 0$ and $b_n > 0$ as required

* Taking the limit of the ratio of the two general terms:

\begin{align}
    \lim_{n \rightarrow \infty} \cfrac{a_n}{b_n} & = \lim_{n \rightarrow \infty} \cfrac{\frac{\sqrt{n}}{n^2 + 1}}{\frac{1}{n^{3/2}}} \\
    & = \lim_{n \rightarrow \infty} \cfrac{n^{1/2}}{n^2 + 1} \cdot \cfrac{n^{3/2}}{1} \\
    & = \lim_{n \rightarrow \infty} \cfrac{n^2}{n^2 + 1} \\
    & = \lim_{n \rightarrow \infty} \cfrac{n^2}{n^2 + 1} \cdot \cfrac{\frac{1}{n^2}}{\frac{1}{n^2}} \\
    & = \lim_{n \rightarrow \infty} \cfrac{1}{1 + 1/n^2} \\
    & = 1 = L > 0 \\
\end{align}

* Since $L = 1 > 0$ is positive and finite, both series converge, and therefore, the series $\displaystyle \sum_{n = 1}^\infty \cfrac{\sqrt{n}}{n^2 + 1}$ converges by the Limit Comparison Test

##### Example 9

Determine if the series $\displaystyle \sum_{n = 0}^\infty e^{-n^2}$ converges or diverges.

##### Solution:

* First we'll rewrite:

\begin{align}
    \sum_{n = 0}^\infty e^{-n^2} & = \sum_{n = 0}^\infty \cfrac{1}{e^{n^2}} \\
    & = \sum_{n = 0}^\infty \left( \cfrac{1}{e} \right)^{n^2}
\end{align}

* This resembles a convergent geometric series with $a = 1$ and $r = \cfrac{1}{e} < 1$

* So we might let $a_n = \left( \cfrac{1}{e} \right)^{n^2}$ and $b_n = \left( \cfrac{1}{e} \right)^n$
    
    * Then $a_n > 0$ and $b_n > 0$ as required

* Taking the limit of the ratio of the two general terms:

\begin{align}
    \lim_{n \rightarrow \infty} \cfrac{a_n}{b_n} & = \lim_{n \rightarrow \infty} \cfrac{\left( \frac{1}{e} \right)^{n^2}}{\left( \frac{1}{e} \right)^n} \\
    & = \lim_{n \rightarrow \infty} \left( \frac{1}{e} \right)^{n^2 - n} \\
    & = \lim_{n \rightarrow \infty} \frac{1}{e^{n^2 - n}} \\
    & = 0
\end{align}

* But although $L = 0$ is finite, $L$ is *not positive* and therefore we cannot draw any conclusions from the Limit Comparison Test with the comparison $b_n = \left( \cfrac{1}{e} \right)^n$

* This means we either need to find a different comparison that gives us a finite $L >0$, or we need to use a different test...

* What about the Direct Comparison Test with the same choice for $b_n$?

* With the same choices for $a_n$ and $b_n$ as used above, we have $0 < a_n \leq b_n$ as required by the Direct Comparison Test

* Then since $\sum b_n$ is a convergent geometric series, $a_n$ must also converge by the Direct Comparison Test
