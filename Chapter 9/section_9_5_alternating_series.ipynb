##### Objective 1: Use the Alternating Series Test to determine whether an infinite series converges

##### Alternating Series

* A series whose terms alternate in sign is called an **alternating series**

* An alternating series has the form

\begin{align}
    \sum_{n = 0}^\infty (-1)^n a_n \\
    \text{ or } \\
    \sum_{n = 0}^\infty (-1)^{n + 1} a_{n}
\end{align}

* One example of an alternating series:

\begin{align}
    \sum_{n = 0}^\infty \left( -\cfrac{1}{3} \right)^n & = 1 - \cfrac{1}{3} + \cfrac{1}{9} - \cfrac{1}{27} + \cfrac{1}{81} -+ \cdots \\
    & = \sum_{n = 0}^\infty (-1)^n \left( \cfrac{1}{3} \right)^n
\end{align}

##### Theorem 9.14: Alternating Series Test

Let $a_n > 0$. The alternating series

\begin{align}
    \sum_{n = 1}^\infty (-1)^n a_n \\
    \text{and} \\
    \sum_{n = 1}^\infty (-1)^{n + 1} a_n
\end{align}

converge when the following two conditions are met.

1. $\displaystyle \lim_{n \rightarrow \infty} a_n = 0$

2. $a_{n+1} \leq a_n,$ for all $n \geq N$ for some positive integer $N$

##### Example 1

Determine if the **alternating harmonic series** defined by $\displaystyle \sum_{n = 1}^\infty (-1)^{n+1} \cfrac{1}{n}$ converges or diverges.

##### Solution:

* First, we determine the non-alternating portion of the general term: $a_n = \cfrac{1}{n}$

* Now we take the limit to test the first condition of the Alternating Series Test:

\begin{align}
    \lim_{n \rightarrow \infty} a_n = \lim_{n \rightarrow \infty} \cfrac{1}{n} \\
    & = 0
\end{align}
    
* So the first condition is satisfied
    
* Next we need to check the second condition of the Alternating Series Test:

\begin{align}
    a_{n + 1} & = \cfrac{1}{n + 1} \\
    & \leq \cfrac{1}{n} \\
    & = a_n
\end{align}

* So *both* condintions of the Alternating Series Test are satisfied and therefore, the alternating harmonic series $\displaystyle \sum_{n = 1}^\infty (-1)^{n + 1} \cfrac{1}{n}$ converges by the Alternating Series Test

##### Example 2

Determine the convergence or divergence of the series $\displaystyle \sum_{n = 1}^\infty \cfrac{n}{(-2)^{n - 1}}$.

##### Solution:

* We do not have a factor of $(-1)^n$ or the like to easily set aside so we will rewrite:

\begin{align}
    \sum_{n = 1}^\infty \cfrac{n}{(-2)^{n - 1}} & = \sum_{n = 1}^\infty \cfrac{n}{(-1)^{n - 1}2^{n - 1}} \\
    & = \sum_{n = 1}^\infty (-1)^{n - 1} \cfrac{n}{2^{n - 1}}
\end{align}

* Now we can determine the non-alternating portion of the general term: $a_n = \cfrac{n}{2^{n - 1}}$

* Taking the limit:

\begin{align}
    \lim_{n \rightarrow \infty} a_n & = \lim_{n \rightarrow \infty} \cfrac{n}{2^{n - 1}} \rightarrow \cfrac{\infty}{\infty} \\
    & = \lim_{n \rightarrow \infty} \cfrac{1}{\left( \ln{2} \right) 2^{n - 1}} \quad (\text{by L'H$\hat{\text{o}}$pital's Rule}) \\
    & = 0
\end{align}

* Furthermore, noting that $a_{n + 1} = \cfrac{n + 1}{2^n}$ and $a_n = \cfrac{n}{2^{n - 1}}$ we need to show that $a_{n + 1} \leq a_n$ for $n \geq 1$

    * We can do this in a couple of ways:

\begin{align}
    \cfrac{a_{n + 1}}{a_n} & = \cfrac{\frac{n + 1}{2^n}}{\frac{n}{2^{n - 1}}} \\
    & = \cfrac{n + 1}{2^n} \cdot \cfrac{2^{n - 1}}{n} \\
    & = \cfrac{n + 1}{1} \cdot \cfrac{2^{-1}}{n} \\
    & = \cfrac{n + 1}{2n} \\
    & \leq 1 \\
    & \Downarrow \\
    \cfrac{a_{n + 1}}{a_n} \leq 1 & \Rightarrow a_{n + 1} \leq a_n
\end{align}

* Alternatively, we could show $a_{n + 1} \leq a_n$ for all $n \geq 1$ by starting with a known inequality and manipulating from there:

\begin{align}
    \cfrac{1}{2} & \leq \cfrac{n}{n + 1} \quad (\text{true for $n \geq 1$}) \\
    \cfrac{1}{2} \cdot \cfrac{2^{n - 1}}{2^{n - 1}} & \leq \cfrac{n}{n + 1} \\
    \cfrac{2^{n - 1}}{2^n} & \leq \cfrac{n}{n + 1} \\
    \cfrac{2^{n - 1}}{2^n} \cdot (n + 1) & \leq \cfrac{n}{n + 1} \cdot (n + 1) \\
    \cfrac{(n + 1)2^{n - 1}}{2^n} \cdot \cfrac{1}{2^{n - 1}} & \leq n \cdot \cfrac{1}{2^{n - 1}} \\
    \cfrac{(n + 1)}{2^n} & \leq \cfrac{n}{2^{n - 1}} \\
    a_{n + 1} & \leq a_n
\end{align}

* Doesn't matter which method you use; sometimes one approach may be easier than the other

* Now we have shown both conditions for the Alternating Series Test are satisfied, and therefore, the series $\displaystyle \sum_{n = 1}^\infty \cfrac{n}{(-2)^{n - 1}}$ converges by the Alternating Series Test

##### Example 3

Determine the convergence or divergence of the series $\displaystyle \sum_{n = 1}^\infty \cfrac{(-1)^{n + 1}(n + 1)}{n}$.

##### Solution:

* Identifying the non-alternation portion of the general term: $a_n = \cfrac{n + 1}{n}$

* Taking the limit:

\begin{align}
    \lim_{n \rightarrow \infty} a_n & = \lim_{n \rightarrow \infty} \cfrac{n + 1}{n} \rightarrow \cfrac{\infty}{\infty} \\
    & = \lim_{n \rightarrow \infty} \cfrac{1}{1} \quad (\text{by L'H$\hat{\text{o}}$pital's Rule}) \\
    & = 1 \neq 0
\end{align}

* Since the limit of $a_n$ is *not* zero, the series will diverge by the $n$th-Term Test for Divergence

* This is true in generaly, if $\displaystyle \lim_{n \rightarrow \infty} a_n \neq 0$ (the first condition of the Alternating Series Test fails), then the series diverges by the $n$th-Term Test for Divergence

##### Objective 2: Use the Alternating Series Remainder to approximate the sum of an alternating series

##### Theorem 9.15: Alternating Series Remainder

If a convergent alternating series satisfies the condition $a_{n+1} \leq a_n$, then the absolute value fo the remainder $R_N$ involved in approximating the sum $S$ by $S_N$ is less than or equal to the first neglected term. That is,

\begin{align}
    |S - S_N| & = |R_N| \leq a_{N + 1}.
\end{align}

##### Example 4

Approximate the sum of the series

\begin{align}
    \sum_{n = 1}^\infty (-1)^{n + 1} \left( \cfrac{1}{n!} \right) & = \cfrac{1}{1!} - \cfrac{1}{2!} + \cfrac{1}{3!} - \cfrac{1}{4!} + \cfrac{1}{5!} - \cfrac{1}{6!} +- \cdots
\end{align}

by its first six terms.

##### Solution:

* The first condition for Theorem 9.15 to apply is that the series must converge, so we apply the Alternating Series Test for convergence

* Identifying the non-alternating portion of the general term: $a_n = \cfrac{1}{n!}$

* Taking the limit: $\displaystyle \lim_{n \rightarrow \infty} a_n = \lim_{n \rightarrow \infty} \cfrac{1}{n!} = 0$

* Since $a_{n + 1} = \cfrac{1}{(n + 1)!} \leq \cfrac{1}{n!} = a_n$, the second condition of the Alternating Series Test is satisfied, and therefore, by the Alternating Series Test, the series converges

* Now we apply the Alternating Series Remainder theorem with $N = 6$:

\begin{align}
    |S - S_6| = |R_6| & \leq a_{6 + 1} \\
    & = a_7 \\
    & = \cfrac{1}{7!} \\
    & = \cfrac{1}{5040} \\
    & \approx 0.0002
\end{align}

* Summing over the first six terms:

\begin{align}
    S_6 & = 1 - \cfrac{1}{2} + \cfrac{1}{6} - \cfrac{1}{24} + \cfrac{1}{120} - \cfrac{1}{720} \\
    S_6 & \approx 0.63194
\end{align}

* Combining this these results we have

\begin{align}
    0.63194 - 0.0002 & \leq S \leq 0.63194 + 0.0002 \\
    0.63192 & \leq S \leq 0.63196
\end{align}

* Therefore, the *actual* sum $S$ of the infinite series lies within the interval $[0.61392, 0.63196]$

##### Example 5

Determine the number of terms required to approximate the series $\displaystyle \sum_{n = 1}^\infty \cfrac{(-1)^{n + 1}}{n^4}$ within an error of $0.001$.

##### Solution:

* First, we need to determine if the series converges:

    * The non-alternating portion of the general term: $a_n = \cfrac{1}{n^4}$
        
        * The limit: $\displaystyle \lim_{n \rightarrow \infty} a_n = \lim_{n \rightarrow \infty} \cfrac{1}{n^4} = 0$
        
    * The general term is a decreasing function since $a_{n + 1} = \cfrac{1}{(n + 1)^4} \leq \cfrac{1}{n^4} = a_n$
    
    * So the Alternating Series Test tells us the series converges and we can proceed in finding the minimum number of terms needed to approximate the series to within an error of $0.001$
    
* The remainder for the series tells us

\begin{align}
    |S - S_N| & = |R_N| \leq a_{N + 1} \leq 0.001 \\
    \Rightarrow a_{N + 1} & = \cfrac{1}{(N + 1)^4} \leq 0.001 \\
    \Rightarrow 1 & \leq 0.001 (N + 1)^4 \\
    \Rightarrow (N + 1)^4 & \geq \cfrac{1}{0.001} \\
    \Rightarrow (N + 1)^4 & \geq 1000 \\
    \Rightarrow N + 1 & \geq (1000)^{1/4} \\
    \Rightarrow N & \geq (1000)^{1/4} - 1 \approx 4.62341 \\
    \Rightarrow N & \geq 5 \\
\end{align}

* So we need *at least* $5$ terms to approximate the series to within an error of $0.001$

* Now performing the actual approximation:

\begin{align}
    S_5 & = \cfrac{1}{1^4} - \cfrac{1}{2^4} + \cfrac{1}{3^4} - \cfrac{1}{4^4} + \cfrac{1}{5^4} \\
    S_5 & \approx 0.94754
\end{align}

* So the infinite series can be *approximated* as $0.94754$ to within an error of less than $0.001$

##### Objective 3: Classify a convergent series as absolutely or conditionally convergent

##### Absolute and Conditional Convergence

* The series $\displaystyle \sum a_n$ is **absolutely convergent** when $\displaystyle \sum |a_n|$ converges

* The series $\displaystyle \sum a_n$ is **conditionally convergent** when $\displaystyle \sum |a_n|$ diverges, but $\displaystyle \sum a_n$ converges

##### Theorem 9.16: Absolute Convergence

If the series $\displaystyle \sum |a_n|$ converges, then the series $\displaystyle \sum a_n$ also converges

##### Example 6

Determine the convergence or divergence of the series $\displaystyle \sum_{n = 0}^\infty \cfrac{(-1)^n n!}{2^n}$. If the series converges, determine if the series is absolutely convergent or conditionally convergent.

##### Solution:

* First, we check the conditions for the Alternating Series Test

* The non-alternating part of the general term is $a_n = \cfrac{n!}{2^n}$

* Recall that $\displaystyle \lim_{n \rightarrow \infty} \cfrac{n!}{k^n} \rightarrow \infty$ for $k > 1$ $\left( \text{and } \lim_{n \rightarrow \infty} \frac{k^n}{n!} \rightarrow 0 \right)$

* Therefore, the limit does not exist and the series diverges by the $n$th-Term Test for Divergence

##### Example 7

Determine the convergence or divergence of the series $\displaystyle \sum_{n = 1}^\infty \cfrac{(-1)^n}{\sqrt{n}}$. If the series converges, determine if the series is absolutely convergent or conditionally convergent.

##### Solution:

* First, we checkt he conditions for the Alternating Series Test

* The non-alternating part of the general term is $a_n = \cfrac{1}{\sqrt{n}} = \cfrac{1}{n^{1/2}}$

* Taking the limit:

\begin{align}
    \lim_{n \rightarrow \infty} a_n & = \lim_{n \rightarrow \infty} \cfrac{1}{n^{1/2}} \\
    & = 0
\end{align}

* Furthermore, $a_{n + 1} = \cfrac{1}{\sqrt{n + 1}} \leq \cfrac{1}{\sqrt{n}} = a_n$

* Therefore, the series converges by the Alternating Series Test

* Now we need to determine if the series is absolutely convergent or conditionally convergent

* Now if we let $a_n = \cfrac{(-1)^n}{\sqrt{n}}$, then $\displaystyle \sum_{n = 1}^\infty |a_n| = \sum_{n = 1}^\infty \left| \cfrac{(-1)^n}{\sqrt{n}} \right| = \sum_{n = 1}^\infty \cfrac{1}{\sqrt{n}} = \sum_{n = 1}^\infty \cfrac{1}{n^{1/2}}$

* This is a $p$-series with $p = \cfrac{1}{2} < 1 \Rightarrow \sum |a_n|$ diverges, while $\sum a_n$ converges $\Rightarrow$ the series is conditionally convergent
