##### Objective 1: Use the Ratio Test to determine whether a series converges or diverges

##### Theorem 9.17: Ratio Test

Let $\sum a_n$ be a series with nonzero terms.

1. The series $\sum a_n$ converges absolutely when $\displaystyle \lim_{n \rightarrow \infty} \left| \frac{a_{n + 1}}{a_n} \right| < 1$

2. The series $\sum a_n$ diverges when $\displaystyle \lim_{n \rightarrow \infty} \left| \frac{a_{n + 1}}{a_n} \right| > 1$ or $\displaystyle \lim_{n \rightarrow \infty} \left| \frac{a_{n + 1}}{a_n} \right| = \infty$

3. The Ratio Test is *inconclusive* when $\displaystyle \lim_{n \rightarrow \infty} \left| \frac{a_{n + 1}}{a_n} \right| = 1$

**Note:** For any $p$-series, the Ratio Test is *always* inconclusive

##### Example 1

Determine the convergence or divergence of $\displaystyle \sum_{n = 0}^\infty \frac{2^n}{n!}$.

##### Solution:

* Let $a_n = \cfrac{2^n}{n!}$, then $a_{n + 1} = \cfrac{2^{n + 1}}{(n + 1)!}$

* Taking the limit of the ratio $\left| \cfrac{a_{n + 1}}{a_n} \right|$ and simplifying:

\begin{align}
    \lim_{n \rightarrow \infty} \left| \frac{a_{n + 1}}{a_n} \right| & = \lim_{n \rightarrow \infty} \left| \cfrac{\frac{2^{n + 1}}{(n + 1)!}}{\frac{2^n}{n!}} \right| \\
    & = \lim_{n \rightarrow \infty} \cfrac{2^{n + 1}}{(n + 1)!} \cdot \cfrac{n!}{2^n} \\
    & = \lim_{n \rightarrow \infty} \cfrac{n!}{(n + 1)!} \cdot \cfrac{2^{n + 1}}{2^n} \\
    & = \lim_{n \rightarrow \infty} \cfrac{n!}{(n + 1) \cdot (n!)} \cdot \cfrac{2 \cdot 2^n}{2^n} \\
    & = \lim_{n \rightarrow \infty} \cfrac{2}{n + 1} \\
    & = 0 < 1 \Rightarrow \text{ converges by the Ratio Test}
\end{align}

##### Example 2

Determine whether each series converges or diverges.

**(a)** $\displaystyle \sum_{n = 0}^\infty \frac{n^2 2^{n + 1}}{3^n}$

**(b)** $\displaystyle \sum_{n = 1}^\infty \frac{n^n}{n!}$

##### Solution:

**(a)** $\displaystyle \sum_{n = 0}^\infty \frac{n^2 2^{n + 1}}{3^n}$

* Let $a_n = \cfrac{n^2 2^{n + 1}}{3^n}$, then $a_{n + 1} = \cfrac{(n + 1)^2 2^{n + 2}}{3^{n + 1}}$

* Taking the limit of the ratio $\left| \cfrac{a_{n + 1}}{a_n} \right|$ and simplifying:

\begin{align}
    \lim_{n \rightarrow \infty} \left| \frac{a_{n + 1}}{a_n} \right| & = \lim_{n \rightarrow \infty} \left| \cfrac{\frac{(n + 1)^2 2^{n + 2}}{3^{n + 1}}}{\frac{n^2 2^{n + 1}}{3^n}} \right| \\
    & = \lim_{n \rightarrow \infty} \cfrac{(n + 1)^2 2^{n + 2}}{3^{n + 1}} \cdot \cfrac{3^n}{n^2 2^{n + 1}} \\
    & = \lim_{n \rightarrow \infty} \cfrac{(n + 1)^2}{n^2} \cdot \cfrac{3^n}{3^{n + 1}} \cdot \cfrac{2^{n + 2}}{2^{n + 1}} \\
    & = \lim_{n \rightarrow \infty} \cfrac{n^2 + 2n + 1}{n^2} \cdot \cfrac{3^n}{3 \cdot 3^n} \cdot \cfrac{2^2 \cdot 2^n}{2 \cdot 2^n} \\
    & = \lim_{n \rightarrow \infty} \cfrac{n^2 + 2n + 1}{n^2} \cdot \cfrac{1}{3} \cdot 2 \\
    & = \cfrac{2}{3} \lim_{n \rightarrow \infty} \cfrac{n^2 + 2n + 1}{n^2} \\
    & = \cfrac{2}{3} < 1 \Rightarrow \text{ converges by the Ratio Test}
\end{align}

**(b)** $\displaystyle \sum_{n = 1}^\infty \frac{n^n}{n!}$

* Let $a_n = \cfrac{n^n}{n!}$, then $a_{n + 1} = \cfrac{(n + 1)^{n + 1}}{(n + 1)!}$

* Applying the Ratio Test:

\begin{align}
    \lim_{n \rightarrow \infty} \left| \cfrac{a_{n + 1}}{a_n} \right| & = \lim_{n \rightarrow \infty} \left| \cfrac{\frac{(n + 1)^{n + 1}}{(n + 1)!}}{\frac{n^n}{n!}} \right| \\
    & = \lim_{n \rightarrow \infty} \frac{(n + 1)^{n + 1}}{(n + 1)!} \cdot \frac{n!}{n^n} \\
    & = \lim_{n \rightarrow \infty} \frac{n!}{(n + 1)!} \cdot \frac{(n + 1)^{n + 1}}{n^n} \\
    & = \lim_{n \rightarrow \infty} \frac{n!}{n! \cdot (n + 1)} \cdot \frac{(n + 1) (n + 1)^n}{n^n} \\
    & = \lim_{n \rightarrow \infty} \frac{1}{n + 1} \cdot (n + 1) \left( \frac{n + 1}{n} \right)^n \\
    & = \lim_{n \rightarrow \infty} \left( 1 + \cfrac{1}{n} \right)^n \\
    & = e > 1 \Rightarrow \text{ diverges by the Ratio Test}
\end{align}

##### Example 3

Determine the convergence or divergence of $\displaystyle \sum_{n = 1}^\infty (-1)^n \frac{\sqrt{n}}{n + 1}.$

##### Solution:

* Let $a_n = (-1)^n \cfrac{\sqrt{n}}{n + 1}$, then $a_{n + 1} = (-1)^{n + 1} \cfrac{\sqrt{n + 1}}{n + 2}$

* Applying the Ratio Test:

\begin{align}
    \lim_{n \rightarrow \infty} \left| \cfrac{a_{n + 1}}{a_n} \right| & = \lim_{n \rightarrow \infty} \left| \cfrac{(-1)^{n + 1} \frac{\sqrt{n + 1}}{n + 2}}{\frac{(-1)^n \sqrt{n}}{n + 1}} \right| \\
    & = \lim_{n \rightarrow \infty} \cfrac{\frac{\sqrt{n + 1}}{n + 2}}{\frac{\sqrt{n}}{n + 1}} \\
    & = \lim_{n \rightarrow \infty} \frac{\sqrt{n + 1}}{n + 2} \cdot \frac{n + 1}{\sqrt{n}} \\
    & = \lim_{n \rightarrow \infty} \frac{n + 1}{n + 2} \cdot \frac{\sqrt{n + 1}}{\sqrt{n}} \\
    & = \lim_{n \rightarrow \infty} \frac{n + 1}{n + 2} \cdot \sqrt{\frac{n + 1}{n}} \\
    & = \lim_{n \rightarrow \infty} \frac{n + 1}{n + 2} \cdot \frac{1/n}{1/n} \cdot \sqrt{\frac{n + 1}{n} \cdot \frac{1/n}{1/n}} \\
    & = \lim_{n \rightarrow \infty} \frac{1 + 1/n}{1 + 2/n} \cdot \sqrt{\frac{1 + 1/n}{1}} \\
    & = 1 \cdot 1 \\
    & = 1 \Rightarrow \text{ Ratio Test is $\underline{\text{inconclusive}}$}
\end{align}

* So we need to apply a *different* test for convergence/divergence

* One possible test we could use is the Alternating Series Test for convergence:

* Let $a_n = \cfrac{\sqrt{n}}{n + 1}$, then

\begin{align}
    \lim_{n \rightarrow \infty} a_n & = \lim_{n \rightarrow \infty} \cfrac{\sqrt{n}}{n + 1} \\
    & = \lim_{n \rightarrow \infty} \cfrac{\sqrt{n}}{n + 1} \cdot \cfrac{1/\sqrt{n}}{1/\sqrt{n}} \\
    & = \lim_{n \rightarrow \infty} \cfrac{1}{n^{1/2} + \frac{1}{\sqrt{n}}} \\
    & = 0
\end{align}

* Furthermore, $a_{n} = f(n) = \cfrac{\sqrt{n}}{n + 1} \Rightarrow f(x) = \cfrac{\sqrt{x}}{x + 1}$

* Therefore, applying the quotient rule for differentiation:

\begin{align}
    f'(x) & = \cfrac{(x + 1) \cdot 1/2 x^{-1/2} - \sqrt{x} \cdot (1)}{(x + 1)^2} \\
    & = \cfrac{1/2 x^{1/2} + 1/2 x^{-1/2} - \sqrt{x}}{(x + 1)^2} \\
    & = \cfrac{\frac{\sqrt{x}}{2} + \frac{1}{2\sqrt{x}} - \sqrt{x}}{(x + 1)^2} \\
    & = \cfrac{-\frac{\sqrt{x}}{2} + \frac{1}{2\sqrt{x}}}{(x + 1)^2} \\
    & = \cfrac{-\frac{x}{2\sqrt{x}} + \frac{1}{2\sqrt{x}}}{(x + 1)^2} \\
    & = \cfrac{\frac{1}{2\sqrt{x}} (-x + 1)}{(x + 1)^2} \\
    & = \cfrac{-x + 1}{2 \sqrt{x}(x + 1)^2} < 0 \text{ for all $x \geq 2$}
\end{align}

which means that $a_n$ is decreasing for all $n \geq 2$ (i.e. $a_{n + 1} \leq a_n$)

* Therefore, the series converges by the Alternating Series Test

##### Objective 2: Use the Root Test to determine whether a series converges or diverges

##### Theorem 9.18: Root Test

1. The series $\sum a_n$ converges absolutely when $\displaystyle \lim_{n \rightarrow \infty} \sqrt[n]{|a_n|} < 1$

2. The series $\sum a_n$ diverges when $\lim_{n \rightarrow \infty} \sqrt[n]{|a_n|} > 1$ or $\displaystyle \lim_{n \rightarrow \infty} \sqrt[n]{|a_n|} = \infty$

3. The Root Test is *inconclusive* when $\displaystyle \lim_{n \rightarrow \infty} \sqrt[n]{|a_n|} = 1$

**Note:** Just as with the Ratio Test, the Root Test is also *always* inconclusive for any $p$-series

##### Example 4

Determine the convergence or divergence of $\displaystyle \sum_{n = 1}^\infty \frac{e^{2n}}{n^n}.$

##### Solution:

* Let $a_n = \cfrac{e^{2n}}{n^n}$

* Applying the Root Test:

\begin{align}
    \lim_{n \rightarrow \infty} \sqrt[n]{\frac{e^{2n}}{n^n}} & = \lim_{n \rightarrow \infty} \left( \frac{e^{2n}}{n^n} \right)^{\frac{1}{n}} \\
    & = \lim_{n \rightarrow \infty} \frac{\left( e^{2n} \right)^{\frac{1}{n}}}{\left( n^n \right)^{\frac{1}{n}}} \\
    & = \lim_{n \rightarrow \infty} \frac{e^{\frac{2n}{n}}}{n^{\frac{n}{n}}} \\
    & = \lim_{n \rightarrow \infty} \frac{e^2}{n^1} \\
    & = 0 < 1 \Rightarrow \text{ converges by Root Test}
\end{align}

##### Example 5

Determine the convergence or divergence of each series.

**(a)** $\displaystyle \sum_{n = 1}^\infty \frac{n + 1}{3n + 1}$

**(b)** $\displaystyle \sum_{n = 1}^\infty \left( \frac{\pi}{6} \right)^n$

**(c)** $\displaystyle \sum_{n = 1}^\infty n e^{-n^2}$

**(d)** $\displaystyle \sum_{n = 1}^\infty \frac{1}{3n + 1}$

**(e)** $\displaystyle \sum_{n = 1}^\infty (-1)^n \frac{3}{4n + 1}$

**(f)** $\displaystyle \sum_{n = 1} \frac{n!}{4n + 1}$

**(g)** $\displaystyle \sum_{n = 1}^\infty \left( \frac{n + 1}{2n + 1} \right)^n$

##### Answers:

**Note:** Only one type of test is given in the answers below, but more than one test may be applicable!

**(a)** $\displaystyle \sum_{n = 1}^\infty \frac{n + 1}{3n + 1}$: Diverges by the $n$th Term Test

**(b)** $\displaystyle \sum_{n = 1}^\infty \left( \frac{\pi}{6} \right)^n$: Convergent geometric series with $|r| = \frac{\pi}{6} < 1$

**(c)** $\displaystyle \sum_{n = 1}^\infty n e^{-n^2}$: Converges by the Root Test

**(d)** $\displaystyle \sum_{n = 1}^\infty \frac{1}{3n + 1}$: Diverges by Limit Comparison Test with $b_n = \frac{1}{n}$

**(e)** $\displaystyle \sum_{n = 1}^\infty (-1)^n \frac{3}{4n + 1}$: Converges by the Alternating Series Test

**(f)** $\displaystyle \sum_{n = 1} \frac{n!}{4n + 1}$: Diverges by Ratio Test

**(g)** $\displaystyle \sum_{n = 1}^\infty \left( \frac{n + 1}{2n + 1} \right)^n$: Converges by Root Test
