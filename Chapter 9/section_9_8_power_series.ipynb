##### Objective 1: Understand the definition of a power series

##### Motivating Example

Consider the function $f(x) = e^x$.

* If we note that $f^{(n)}(x) = e^x \Rightarrow f^{(n)}(0) = 1$ for any $n \geq 1$, then the $n$th Maclaurin polynomial that approximates $f(x) = e^x$ is determined by

\begin{align}
    f(x) = e^x \approx P_n(x) & = \frac{f(0)}{0!} + \frac{f'(0)}{1!} \cdot x + \frac{f''(0)}{2!} \cdot x^2 \\
    & \qquad + \frac{f'''(0)}{3!} \cdot x^3 + \cdots + \frac{f^{(n)}(0)}{n!} \cdot x^n \\
    & = \frac{1}{0!} + \frac{x}{1!} + \frac{x^2}{2!} + \frac{x^3}{3!} + \cdots + \frac{x^n}{n!} \\
    & = \sum_{k = 0}^n \frac{x^k}{k!} \\
\end{align}

* But what happens if we let $n \rightarrow \infty$?

* We then obtain an *infinite series:*

\begin{align}
    f(x) = e^x = \lim_{n \rightarrow \infty} \sum_{k = 0}^n \frac{f^{(k)}(0)}{k!} \cdot x^k \\
    \Rightarrow f(x) = e^x = \sum_{k = 0}^\infty \frac{x^k}{k!}
\end{align}

* We call such a series a *power series*

##### Power Series

If $x$ is a variable, then an infinite series of the form

\begin{align}
    \sum_{k = 0}^\infty a_k x^k & = a_0 + a_1 x + a_2 x^2 + a_3 x^3 + \cdots + a_k x^k + \cdots
\end{align}

is called a **power series.** More generally, an infinite series of the form

\begin{align}
    \sum_{k = 0}^\infty a_k (x - c)^k & = a_0 + a_1 (x - c) + a_2 (x - c)^2 + a_3 (x - c)^3 + \cdots + a_k (x - c)^k + \cdots
\end{align}

is called a **power series centered at $c$**, where $c$ is a constant.

**Note:** For simplicity, we assume that $(x - c)^0 = 1$ even when $x = c$ for any power series

##### Examples of Power Series

* The following power series is centered at $c = 0$:

\begin{align}
    \sum_{k = 0}^\infty \frac{x^k}{k!} & = 1 + x + \frac{x^2}{2!} + \frac{x^3}{3!} + \cdots
\end{align}

* The following power series is centered at $c = -1$:

\begin{align}
    \sum_{k = 0}^\infty (-1)^k (x + 1)^k & = 1 - (x + 1) + (x + 1)^2 - (x + 1)^3 + \cdots
\end{align}

* The following power series is centered at $c = 1$:

\begin{align}
    \sum_{k = 1}^\infty \frac{1}{k} (x - 1)^k & = (x - 1) + \frac{1}{2} (x - 1)^2 + \frac{1}{3} (x - 1)^3 + \cdots
\end{align}

##### Objective 2: Find the radius and interval of convergence of a power series

##### Radius and Interval of Convergence

A power series in $x$ is a function of $x$

\begin{align}
    f(x) & = \sum_{k = 0}^\infty a_k (x - c)^k
\end{align}

where the **domain of $f$** must be only those values of $x$ for which the series converges.

Every power series converges at its center $c$ since $f(c) = a_0$ for any power series centered at $c$.

##### Theorem 9.20: Convergence of a Power Series

For a power series centered at $c$, precisely one of the following is true:

1. The series converges only at $c$

2. There exists a real number $R > 0$ such that the series converges absoutely for

\begin{align}
    |x - c| < R
\end{align}

and diverges for

\begin{align}
    |x - c| > R
\end{align}

3. The series converges absolutely for all $x$

The number $R$ is the **radius of convergence** of the power series

* If condition 1 is true, then $R = 0$
    
* If condition 3 is true, then $R = \infty$
    
* The set of all values $x$ for which the sereies converges is the **interval of convergence** of the power series

##### Example 2

Find the radius of convergence of $\displaystyle \sum_{k = 0}^\infty k! x^k$.

##### Solution:

* This power series is centered at $c = 0$ which means

\begin{align}
    f(0) = \sum_{k = 0}^\infty k! \cdot 0^k = 1 + 0 + 0 + \cdots = 1
\end{align}

**Note:** Recall that we assume $(x - c)^0 = 1$ for simplicity for any power series

* We apply the Ratio Test to determine the values of $x$ for which the series converges:

* Let $u_k = k! x^k$, then
    
\begin{align}
    \lim_{k \rightarrow \infty} \left| \frac{u_{k + 1}}{u_k} \right| & = \lim_{k \rightarrow \infty} \left| \frac{(k + 1)! x^{k + 1}}{k! x^k} \right|, x \neq 0 \\
    & = \lim_{k \rightarrow \infty} \frac{(k + 1)!}{k!} \cdot \left| \frac{x^{k + 1}}{x^k} \right| \\
    & = \lim_{k \rightarrow \infty} \frac{(k + 1) \cdot k!}{k!} \cdot \left| \frac{x \cdot x^k}{x^k} \right| \\
    & = \lim_{k \rightarrow \infty} \frac{(k + 1) \cdot k!}{k!} \cdot \frac{|x| \cdot |x^k|}{|x^k|} \\
    & = |x| \lim_{k \rightarrow \infty} (k + 1) \\
\end{align}

* Since $x \neq 0, |x| > 0 \Rightarrow |x| \lim_{k \rightarrow \infty} (k + 1) = \infty$

* Therefore, by the Ratio Test, the series *diverges for all $x \neq 0$* and converges *only* at its center $x = 0$

* The radius of convergence is therefore $R = 0$ (there is no interval of convergence; the convergence is at a single point)

##### Example 3

Find the radius of convergence of $\displaystyle \sum_{k = 0}^\infty 3 (x - 2)^k$.

##### Solution:

* We know the series converges for $x = c = 2$

* We apply the Ratio Test to determine the values of $x$ for which the series converges:

* Let $u_k = 3 (x - 2)^k$, then
    
\begin{align}
    \lim_{k \rightarrow \infty} \left| \frac{u_{k + 1}}{u_k} \right| & = \lim_{k \rightarrow \infty} \left| \frac{3 (x - 2)^{k + 1}}{3 (x - 2)^k} \right|, x \neq 2 \\
    & = \lim_{k \rightarrow \infty} \left| \frac{(x - 2) (x - 2)^k}{(x - 2)^k} \right| \\
    & = \lim_{k \rightarrow \infty} | x - 2 | \\
    & = | x - 2 | \lim_{k \rightarrow \infty} 1 \\
    & = | x - 2 |
\end{align}

* For the series to converge by the Ratio Test, we require the limit to be less than $1$:

\begin{align}
    |x - 2| & < 1 \\
\end{align}

* Therefore the radius of convergence is $R = 1$

##### Example 4

Find the radius of convergence of $\displaystyle \sum_{k = 0}^\infty \frac{(-1)^{2k + 1} x^{2k + 1}}{(2k + 1)!}$.

##### Solution:

* We know the series converges for $x = c = 0$

* We apply the Ratio Test to determine the values of $x$ for which the series converges:

* Let $u_k = \cfrac{(-1)^{2k + 1}x^{2k + 1}}{(2k + 1)!}$. First, we'll determine $u_{k + 1}$ and simplifiy:

\begin{align}
    u_{k + 1} & = \frac{(-1)^{2(k + 1) + 1}x^{2(k + 1) + 1}}{(2(k + 1) + 1)!} \\
    & = \frac{(-1)^{2k + 2 + 1}x^{2k + 2 + 1}}{(2k + 2 + 1)!} \\
    & = \frac{(-1)^{2k + 3}x^{2k + 3}}{(2k + 3)!}
\end{align}

* Therefore we get

\begin{align}
    \lim_{k \rightarrow \infty} \left| \frac{u_{k + 1}}{u_k} \right| & = \lim_{k \rightarrow \infty} \left| \cfrac{\frac{(-1)^{2k + 3}x^{2k + 3}}{(2k + 3)!}}{\frac{(-1)^{2k + 1}x^{2k + 1}}{(2k + 1)!}} \right|, x \neq 0 \\
    & = \lim_{k \rightarrow \infty} \left| \frac{(-1)^{2k + 3}x^{2k + 3}}{(2k + 3)!} \cdot \frac{(2k + 1)!}{(-1)^{2k + 1}x^{2k + 1}} \right| \\
    & = \lim_{k \rightarrow \infty} \left| \frac{x^{2k + 3}}{(2k + 3)!} \cdot \frac{(2k + 1)!}{x^{2k + 1}} \right| \\
    & = \lim_{k \rightarrow \infty} \left| \frac{(2k + 1)!}{(2k + 3)!} \cdot \frac{x^{2k + 3}}{x^{2k + 1}} \right| \\
    & = \lim_{k \rightarrow \infty} \left| \frac{(2k + 1)!}{(2k + 1)! \cdot (2k + 2) \cdot (2k + 3)} \cdot \frac{x^2 \cdot x^{2k + 1}}{x^{2k + 1}} \right| \\
    & = \lim_{k \rightarrow \infty} \left| \frac{x^2}{(2k + 2) \cdot (2k + 3)} \right| \\
    & = x^2 \lim_{k \rightarrow \infty} \frac{1}{(2k + 2) \cdot (2k + 3)} \\
    & = x^2 \cdot 0 \\
    & = 0 < 1
\end{align}

* So for *every* real number $x$, the series will converge $\Rightarrow R = \infty$ (the series converges for all values on the entire real number line)

##### Objective 3: Determine the endpoint convergence of a power series

##### Intervals of Convergence

There are a total of six types of convergence:

* Let $a$ and $b$ be real numbers on the real number line such that $a < b$ and let $c$ be the midpoint of $a$ and $b$

* Then the six types of convergence are as follows

1. The power series converges for precisely one real number $x = c$ at which the power series is centered
    
    * i.e. The power series converges at a single point

2. Convergence for all real numbers $x$ in the interval $(a, b)$
        
    * i.e. The power series has interval of convergence $(a, b)$

3. Convergence for all real numbers $x$ in the half open interval $(a, b]$
        
    * i.e. The power series has interval of convergence $(a, b]$

4. Convergence for all real numbers $x$ in the half open interval $[a, b)$
        
    * i.e. The power series has interval of convergence $[a, b)$

5. Convergence for all real numbers $x$ in the closed interval $[a, b]$
        
    * i.e. The power sereies has interval of convergence $[a, b]$

6. Convergence for all real numbers $x$ in the open interval $(-\infty, \infty)$
        
    * i.e. The power series converges for *all* real numbers

**Important Notes:**

* Every interval of convergence with finite radius of convergence requires the left and right endpoints to be tested for convergence

    * This is due to the fact that the Ratio Test is inconclusive when $\displaystyle \lim_{k \rightarrow \infty} \left| \frac{u_{k + 1}}{u_k} \right| = 1$
    
    * The endpoints will *always* result in $\displaystyle \lim_{k \rightarrow \infty} \left| \frac{u_{k + 1}}{u_k} \right| = 1$ when applying the Ratio Test to a power series
    
    * This means we must evaluate the power series at the points $x = a$ and $x = b$ individually and determine the convergence/divergence using any previously discussed tests for convergence/divergence

##### Example 5

Find the interval of convergence of $\displaystyle \sum_{k = 1}^\infty \frac{x^k}{k}$.

##### Solution:

* We know the series converges for $x = c = 0$

* We apply the Ratio Test to determine the values of $x$ for which the series converges:

* Let $u_k = \cfrac{x^k}{k}$, then $u_{k + 1} = \cfrac{x^{k + 1}}{k + 1}$

* Applying the Ratio Test:

\begin{align}
    \lim_{k \rightarrow \infty} \left| \frac{u_{k + 1}}{u_k} \right| & = \lim_{k \rightarrow \infty} \left| \cfrac{\frac{x^{k + 1}}{k + 1}}{\frac{x^k}{k}} \right|, x \neq 0 \\
    & = \lim_{k \rightarrow \infty} \left| \frac{x^{k + 1}}{k + 1} \cdot \frac{k}{x^k} \right| \\
    & = \lim_{k \rightarrow \infty} \left| \frac{k}{k + 1} \cdot \frac{x^{k + 1}}{x^k} \right| \\
    & = \lim_{k \rightarrow \infty} \left| \frac{k}{k + 1} \cdot \frac{x \cdot x^k}{x^k} \right| \\
    & = \lim_{k \rightarrow \infty} \left| \frac{k}{k + 1} \cdot x \right| \\
    & = |x| \lim_{k \rightarrow \infty} \frac{k}{k + 1} \\
    & = |x| \cdot 1 \\
    & = |x|
\end{align}

* For the power series to converge the Ratio Test requires $|x| < 1$:

\begin{align}
    \Rightarrow -1 & < x < 1
\end{align}

* This means we currently have the open interval $(a, b) = (-1, 1)$ and we know that at the very least the series converges for all $x$ within this interval

* Now we must check the endpoints where $x = -1$ and $x = 1$ and determine convergence or divergence for these series

* For $x = -1$ we have $\displaystyle \sum_{k = 1}^\infty \frac{(-1)^k}{k}$ which converges by the Alternating Series Test:

    1. Let $a_k = \cfrac{1}{k}$, then $\displaystyle \lim_{k \rightarrow \infty} a_k = \lim_{k \rightarrow \infty} \frac{1}{k} = 0$
    
    2. $a_{k + 1} = \cfrac{1}{k + 1} \leq \cfrac{1}{k} = a_k$
    
* Therefore, the power series converges for $x = -1$ so we'll include this endpoint in our interval of convergence

* For $x = 1$ we have $\displaystyle \sum_{k = 1}^\infty \frac{1}{k}$ which is the harmonic series and therefore diverges ($p$-series with $p = 1 \Rightarrow$ diverges)

* Since the power series diverges for $x = 1$ we will *not* include this in our interval

* Therefore, the interval of convergence is $[-1, 1)$ and $R = 1$

    * That is, the series will converge for all real numbers $x$ such that $-1 \leq x < 1$

##### Example 6

Find the interval of convergence of $\displaystyle \sum_{k = 0}^\infty \frac{(-1)^k (x + 1)^k}{2^k}$.

##### Solution:

* We know the series converges for $x = c = -1$

* We apply the Ratio Test to determine the values of $x$ for which the series converges:

* Let $u_k = \cfrac{(-1)^k (x + 1)^k}{2^k}$, then $u_{k + 1} = \cfrac{(-1)^{k + 1} (x + 1)^{k + 1}}{2^{k + 1}}$

* Applying the Ratio Test:

\begin{align}
    \lim_{k \rightarrow \infty} \left| \frac{u_{k + 1}}{u_k} \right| & = \lim_{k \rightarrow \infty} \left| \cfrac{\frac{(-1)^{k + 1} (x + 1)^{k + 1}}{2^{k + 1}}}{\frac{(-1)^k (x + 1)^k}{2^k}} \right|, x \neq -1 \\
    & = \lim_{k \rightarrow \infty} \left| \frac{(x + 1)^{k + 1}}{2^{k + 1}} \cdot \frac{2^k}{(x + 1)^k} \right| \\
    & = \lim_{k \rightarrow \infty} \left| \frac{2^k}{2^{k + 1}} \cdot \frac{(x + 1)^{k + 1}}{(x + 1)^k} \right| \\
    & = \lim_{k \rightarrow \infty} \left| \frac{2^k}{2^{k + 1}} \cdot \frac{(x + 1) \cdot (x + 1)^k}{(x + 1)^k} \right| \\
    & = \lim_{k \rightarrow \infty} \left| \frac{2^k}{2^{k + 1}} \cdot (x + 1) \right| \\
    & = |x + 1| \lim_{k \rightarrow \infty} \frac{2^k}{2^{k + 1}}  \\
    & = |x + 1| \lim_{k \rightarrow \infty} \frac{2^k}{2 \cdot 2^{k}} \\
    & = |x + 1| \lim_{k \rightarrow \infty} \frac{1}{2} \\
    & = \frac{|x + 1|}{2}
\end{align}

* For the power series to converge the Ratio Test requires $\displaystyle \frac{|x + 1|}{2} < 1$:

\begin{align}
    \Rightarrow -1 & < \frac{x + 1}{2} < 1 \\
    \Rightarrow -2 & < x + 1 < 2 \\
    \Rightarrow -3 & < x < 1 \\
\end{align}

* This means we currently have the open interval $(a, b) = (-3, 1)$ and we know that at the very least the series converges for all $x$ within this interval

* Now we must check the endpoints where $x = -3$ and $x = 1$ and determine convergence or divergence for these series

* For $x = -3$ we have

\begin{align}
    \sum_{k = 0}^\infty \frac{(-1)^k (-2)^k}{2^k} & = \sum_{k = 0}^\infty \frac{(-1)^k (-1)^k 2^k}{2^k} \\
    & = \sum_{k = 0}^\infty \frac{(-1)^{2k} 2^k}{2^k} \\
    & = \sum_{k =0}^\infty 1
\end{align}

which diverges by the $n$th Term Test for divergence
    
* Therefore, the power series diverges for $x = -3$ so we'll exclude this endpoint in our interval of convergence

* For $x = 1$ we have

\begin{align}
    \sum_{k = 0}^\infty \frac{(-1)^k (1 + 1)^k}{2^k} & = \sum_{k = 0}^\infty \frac{(-1)^k 2^k}{2^k} \\
    & = \sum_{k = 0}^\infty (-1)^k
\end{align}

which is a divergent geometric series ($r = -1 \Rightarrow |r| = 1$)

* Since the series diverges for $x = 1$ we do not include this in the interval of convergence

* Therefore, the interval of convergence is $(-3, 1)$ and $R = 2$

    * That is, the series will converge for all real numbers $x$ such that $-3 < x < 1$

##### Example 7

Find the interval of convergence of $\displaystyle \sum_{k = 1}^\infty \frac{x^k}{k^2}$.

##### Solution:

* We know the series converges for $x = c = 0$

* We apply the Ratio Test to determine the values of $x$ for which the series converges:

* Let $u_k = \cfrac{x^k}{k^2}$, then $u_{k + 1} = \cfrac{x^{k + 1}}{(k + 1)^2}$

* Applying the Ratio Test:

\begin{align}
    \lim_{k \rightarrow \infty} \left| \frac{u_{k + 1}}{u_k} \right| & = \lim_{k \rightarrow \infty} \left| \cfrac{\frac{x^{k + 1}}{(k + 1)^2}}{\frac{x^k}{k^2}} \right|, x \neq 0 \\
    & = \lim_{k \rightarrow \infty} \left| \frac{x^{k + 1}}{(k + 1)^2} \cdot \frac{k^2}{x^k} \right| \\
    & = \lim_{k \rightarrow \infty} \left| \frac{k^2}{(k + 1)^2} \cdot \frac{x^{k + 1}}{x^k} \right| \\
    & = \lim_{k \rightarrow \infty} \left| \frac{k^2}{(k + 1)^2} \cdot \frac{x \cdot x^k}{x^k} \right| \\
    & = \lim_{k \rightarrow \infty} \left| \frac{k^2}{(k + 1)^2} \cdot x \right| \\
    & = |x| \lim_{k \rightarrow \infty} \frac{k^2}{(k + 1)^2} \\
    & = |x| \cdot 1 \\
    & = |x|
\end{align}

* For the power series to converge the Ratio Test requires $|x| < 1$:

\begin{align}
    \Rightarrow -1 & < x < 1
\end{align}

* This means we currently have the open interval $(a, b) = (-1, 1)$ and we know that at the very least the series converges for all $x$ within this interval

* Now we must check the endpoints where $x = -1$ and $x = 1$ and determine convergence or divergence for these series

* For $x = -1$ we have $\displaystyle \sum_{k = 1}^\infty \frac{(-1)^k}{k^2}$ which converges by the Alternating Series Test:

    1. Let $a_k = \frac{1}{k^2}$, then $\displaystyle \lim_{k \rightarrow \infty} \frac{1}{k^2} = 0$
    
    2. $a_{k + 1} = \frac{1}{(k + 1)^2} \leq \frac{1}{k^2} = a_k$
    
* Therefore, the power series converges for $x = -1$ so we'll include this endpoint in our interval of convergence

* For $x = 1$ we have $\displaystyle \sum_{k = 1}^\infty \frac{1}{k^2}$ which is a convergent $p$-series with $p = 2 > 1$

* Therefore, the power series converges for $x = 1$ and we include this in the interval of convergence

* Therefore, the interval of convergence is $[-1, 1]$ and $R = 1$

    * That is, the series will converge for all real numbers $x$ such that $-1 \leq x \leq 1$

##### Objective 4: Differentiate and integrate a power series

##### Theorem 9.21: Differentiation and Integration of Power Series

* If a power series

\begin{align}
    f(x) & = \sum_{k = 0}^\infty a_k (x - c)^k \\
    & = a_0 + a_1 (x - c) + a_2 (x - c)^2 + a_3 (x - c)^3 + \cdots
\end{align}

has a radius of convergences of $R > 0$, then $f$ is differentiable on the interval $(c - R, c + R)$

* The derivative of $f$ is given by

\begin{align}
    f'(x) & = \frac{d}{dx} \left( a_0 + a_1 (x - c) + a_2 (x - c)^2 + a_3 (x - c)^3 + \cdots \right) \\
    & = a_1 + 2 a_2 (x - c) + 3 a_3 (x - c)^2 + \cdots \\
    & = \sum_{k = 1}^\infty k a_k (x - c)^{k - 1}
\end{align}

* The integral of $f$ is given by

\begin{align}
    \int f(x) \, dx & = C + \sum_{k = 0}^\infty a_k \frac{(x - c)^{k + 1}}{k + 1} \\
    & = C + a_0 (x - c) + a_1 \frac{(x - c)^2}{2} + a_2 \frac{(x - c)^3}{3} + \cdots 
\end{align}

**Important Notes:**

* The radius of convergence of the series obtained via differentiation and integration is the same as that of the original power series

* The interval of convergence may differ due to the behavior at the endpoints

* Note the change in the start of the index $k$:

    * For $f(x)$ the starting index is $k = 0$
    
    * For $f'(x)$ the starting index is $k = 1$
    
    * This means that the index of summation shifts up by one unit for differentiation due to the constant term $a_0$; if the first term of the series is not constant, then the index will *not* shift
    
* The starting index does not change for integration

##### Example 8

For the power series $\displaystyle f(x) = \sum_{k = 1}^\infty \frac{x^k}{k}$ find the interval of convergence for the following.

**(a)** $f(x)$

**(b)** $f'(x)$

**(c)** $\int f(x) \, dx$

##### Solution:

**(a)** For $f(x)$ we have

\begin{align}
    f(x) & = \sum_{k = 1}^\infty \frac{x^k}{k}
\end{align}

* Applying the Ratio Test:

\begin{align}
    \lim_{k \rightarrow \infty} \left| \cfrac{\frac{x^{k + 1}}{k + 1}}{\frac{x^k}{k}} \right| & = \lim_{k \rightarrow \infty} \left| \frac{x^{k + 1}}{k + 1} \cdot \frac{k}{x^k} \right| \\
    & = \lim_{k \rightarrow \infty} \left| \frac{k}{k + 1} \cdot \frac{x^{k + 1}}{x^k} \right| \\
    & = \lim_{k \rightarrow \infty} \left| \frac{k}{k + 1} \cdot \frac{x \cdot x^k}{x^k} \right| \\
    & = \lim_{k \rightarrow \infty} \left| \frac{k}{k + 1} \cdot x \right| \\
    & = |x| \lim_{k \rightarrow \infty} \frac{k}{k + 1} \\
    & = |x|
\end{align}

* Therefore, we require that $|x| < 1 \Rightarrow -1 < x < 1$ and we need to test the endpoints

* For $x = -1$ we have $\displaystyle \sum_{k = 1}^\infty \frac{(-1)^k}{k}$ which is the alternating harmonic series which converges by the Alternating Series Test

* For $x = 1$ we have $\displaystyle \sum_{k = 1}^\infty \frac{1}{k}$ which is the harmonic series and therefore diverges

* Therefore, the interval of convergence is $[-1, 1)$ and $R = 1$

**(b)** For $f'(x)$ we have

\begin{align}
    f'(x) & = \frac{d}{dx} \left( \sum_{k = 1}^\infty \frac{x^k}{k} \right) \\
    & = \frac{d}{dx} \left( x + \frac{x^2}{2} + \frac{x^3}{3} + \cdots \right) \\
    & = 1 + \frac{2 x}{2} + \frac{3 x^2}{3} + \cdots \\
    & = 1 + x + x^2 + \cdots \\
    & = \sum_{k = 1}^\infty \frac{k x^{k - 1}}{k} \\
    & = \sum_{k = 1}^\infty x^{k - 1} \\
\end{align}

* Applying the Ratio Test:

\begin{align}
    \lim_{k \rightarrow \infty} \left| \frac{x^k}{x^{k - 1}} \right| & = \lim_{k \rightarrow \infty} | x^{k - (k - 1)} | \\
    & = \lim_{k \rightarrow \infty} |x^{k - k + 1}| \\
    & = \lim_{k \rightarrow \infty} |x| \\
    & = |x|
\end{align}

* The Ratio Test converges provided $|x| < 1$ and therefore $-1 < x < 1$

* For $x = -1$ we have $\displaystyle \sum_{k = 1}^\infty (-1)^{k - 1}$ which diverges by the $n$th Term Test since $\displaystyle \lim_{k \rightarrow \infty} (-1)^{k - 1}$ does not exist

* For $x = 1$ we have $\displaystyle \sum_{k = 1}^\infty 1$ which diverges also by the $n$th Term Test since $\displaystyle \lim_{k \rightarrow \infty} 1 = 1 \neq 0$

* Therefore, the interval of convergence is $(-1, 1)$ and $R = 1$

**(c)** For $\int f(x) \, dx$ we have

\begin{align}
    \int f(x) \, dx & = \int \sum_{k = 1}^\infty \frac{x^k}{k} \, dx \\
    & = \sum_{k = 1}^\infty \int \frac{x^k}{k} \, dx \\
    & = \sum_{k = 1}^\infty \frac{1}{k} \int x^k \, dx \\
    & = C + \sum_{k = 1}^\infty \frac{x^{k + 1}}{k (k + 1)} \\
\end{align}

* Applying the Ratio Test:

\begin{align}
    \lim_{k \rightarrow \infty} \left| \cfrac{\frac{x^{k + 2}}{(k + 1)(k + 2)}}{\frac{x^{k + 1}}{k(k + 1)}} \right| & = \lim_{k \rightarrow \infty} \left| \frac{x^{k + 2}}{(k + 1)(k + 2)} \cdot \frac{k(k + 1)}{x^{k + 1}} \right| \\
    & = \lim_{k \rightarrow \infty} \left| \frac{k(k + 1)}{(k + 1)(k + 2)} \cdot \frac{x \cdot x^{k + 1}}{x^{k + 1}} \right| \\
    & = |x| \lim_{k \rightarrow \infty} \frac{k}{k + 1} \\
    & = |x| \\
\end{align}

* The Ratio Test requires $|x| < 1$ and therefore $-1 < x < 1$

* For $x = -1$ we have $\displaystyle \sum_{k \rightarrow \infty} \frac{(-1)^{k}}{k(k + 1)} = \sum_{k \rightarrow \infty} \frac{(-1)^{k}}{k^2 + k}$ which converges by the Alternating Series Test:

    1. Let $a_k = \cfrac{1}{k^2 + k}$, then $\displaystyle \lim_{k \rightarrow \infty} \frac{1}{k^2 + k} = 0$
    
    2. $a_{k + 1} = \cfrac{1}{(k + 1)^2 + k + 1} \leq \cfrac{1}{k^2 + k} = a_k$
    
* Therefore, we include $x = -1$ as part of the interval of convergence

* For $x = 1$ we have $\displaystyle \sum_{k = 1}^\infty \frac{1}{k^2 + k}$ which also converges by the limit comparison test with $b_k = \frac{1}{k^2}$ (which is a convergent $p$-series with $p = 2 > 1$):

\begin{align}
    \lim_{k \rightarrow \infty} \cfrac{\frac{1}{k^2 + k}}{\frac{1}{k^2}} & = \lim_{k \rightarrow \infty} \frac{1}{k^2 + k} \cdot \frac{k^2}{1} \\
    & = \lim_{k \rightarrow \infty} \frac{k^2}{k^2 + k} \\
    & = 1 > 0
\end{align}

* Therfore, both series converge by the Limit Comparison Test we include $x = 1$ in the interval of convergence

* Therefore, we have an interval of convergence of $[-1, 1]$ and $R = 1$
