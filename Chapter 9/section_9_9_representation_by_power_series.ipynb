##### Objective 1: Find a geometric power series that represents a function

##### Review of Geometric Series

* Recall that a geometric series is a series of the form

\begin{align}
    \sum_{k = 0}^{\infty} a_0 r^k
\end{align}

* Recall, also, that the sum of a convergent geometric series ($|r| < 1$) is determined by

\begin{align}
    \sum_{k = 0}^{\infty} a_0 r^k = S = \frac{a_0}{1 - r}
\end{align}

##### Functions Represented as Power Series

* Suppose we define a geometric series with $a_0 = 1$ and $r = x$, where $x$ is a real number such that $|x| < 1$

* Then we would have a geometric series that is in the form of a power series:

\begin{align}
    f(x) & = \sum_{k = 0}^\infty x^k
\end{align}

* Furthermore, since $S = \frac{a_0}{1 - r}$ with $a_0 = 1$ and $r = x$ such that $|r| = |x| < 1$ we have $f(x) = \frac{1}{1 - x}$

* The interval of convergence is $(-1, 1)$ since the series diverges for $|r| = |x| \geq 1$

* In fact, we can express any function of the form $f(x) = \frac{a}{b - cx}$ with nonzero constants $a, b,$ and $c$ as a power series, provided $\left| \frac{c x}{b} \right| < 1$ with interval of convergence to be determined

##### Example 1

Find a power series for $f(x) = \frac{4}{x + 2}$, centered at $c = 0$.

* We rewrite the function:

\begin{align}
    f(x) & = \frac{4}{x + 2} \\
    & = \frac{4}{2 + x} \\
    & = \frac{4}{2 - (-x)} \\
    & = \frac{4}{2 \left( 1 - \left( -\frac{x}{2} \right) \right)} \\
    & = \frac{\frac{4}{2}}{1 - \left( -\frac{x}{2} \right)} \\
    & = \frac{2}{1 - \left( -\frac{x}{2} \right)} \\
\end{align}

* This can be seen as a geometric series with $a_0 = 2$ and $r = -\frac{x}{2}$

* Therefore, we can express the function as a power series in the form

\begin{align}
    f(x) & = \sum_{k = 0}^\infty 2 \left( - \frac{x}{2} \right)^k \\
    & = \sum_{k = 0}^\infty 2 \left( - \frac{1}{2} \right)^k x^k 
\end{align}

* The interval of convergence is determined by $|r| = \left| -\frac{x}{2} \right| < 1$

* Therefore, the interval of convergence is

\begin{align}
    & \left| - \frac{x}{2} \right| = \left| \frac{x}{2} \right| < 1 \\
    & \Rightarrow |x| < 2 \\
    & \Rightarrow -2 < x < 2
\end{align}

* Or in interval notation: $(-2, 2)$

##### Example 2 

Find a power series for $f(x) = \frac{1}{x}$, centered at $c = 1$.

##### Solution:

* Rewriting:

\begin{align}
    f(x) & = \frac{1}{x} \\
    & = \frac{1}{x + 1 - 1} \\
    & = \frac{1}{1 + x - 1} \\
    & = \frac{1}{1 + (x - 1)} \\
    & = \frac{1}{1 - (-x + 1)} \\
    & = \frac{1}{1 - (1 - x)} \\
\end{align}

* Then $f(x)$ can be represented as a geometric series with $a_0 = 1$ and $r = 1 - x$

* Therefore, we can express the function as a power series in the form

\begin{align}
    f(x) & = \sum_{k = 0}^\infty \left( 1 - x \right)^k \\
    & = \sum_{k = 0}^\infty \left( -(x - 1) \right)^k \\
    & = \sum_{k = 0}^\infty (-1)^k \left( x - 1 \right)^k \\
\end{align}

* The interval of convergence is determined by $|r| = |x - 1| < 1$

* Therefore, the interval of convergence is

\begin{align}
    & |x - 1| < 1 \\
    & \Rightarrow -1 < x - 1 < 1 \\
    & \Rightarrow 0 < x < 2 \\
\end{align}

* Or in interval notation: $(0, 2)$

##### Objective 2: Construct a power series using series operations

##### Operations with Power Series

Let $\displaystyle f(x) = \sum_{k = 0}^\infty a_k x^k$ and $\displaystyle g(x) = \sum_{k = 0}^\infty b_k x^k$. Then for constants $h$ and $N$, the following properties hold.

1. $\displaystyle f(h x) = \sum_{k = 0}^\infty a_k h^k x^k$

2. $\displaystyle f(x^N) = \sum_{k = 0}^\infty a_k x^{kN}$

3. $\displaystyle f(x) \pm g(x) = \sum_{k = 0}^\infty (a_k \pm b_k) x^k$

**Important Note:** Performing these operations may change the interval of convergence of the series that results!

##### Example 3

Find a power series for $f(x) = \frac{3x - 1}{x^2 - 1}$ centered at $c = 0$.

##### Solution:

* General approach: factor the denominator and perform partial fraction decomposition

* Factoring:

\begin{align}
    f(x) & = \frac{3x - 1}{(x + 1)(x - 1)}
\end{align}

* Partial fraction decomposition is performed below:

\begin{align}
    f(x) = \frac{3x - 1}{(x + 1)(x - 1)} & = \frac{A}{x + 1} + \frac{B}{x - 1}
\end{align}

* Mulitplying both sides by the common denominator $(x + 1)(x - 1)$:

\begin{align}
    \frac{3x - 1}{(x + 1)(x - 1)} \cdot (x + 1)(x - 1) & = \left( \frac{A}{x + 1} + \frac{B}{x - 1} \right) (x + 1)(x - 1)
\end{align}

* Simplifying the left-hand side and distributing on the right-hand side:

\begin{align}
    3x - 1 & = \frac{A}{x + 1} \cdot (x + 1)(x - 1) + \frac{B}{x - 1} \cdot (x + 1)(x - 1)
\end{align}

* Simplifying:

\begin{align}
    3x - 1 & = A (x - 1) + B (x + 1)
\end{align}

* If $x = 1$, then

\begin{align}
    3(1) - 1 & = A(1 - 1) + B(1 + 1) \\
    2 & = 2B \\
    B & = 1
\end{align}

* If $x = -1$, then

\begin{align}
    3(-1) - 1 & = A (-1 - 1) + B (-1 + 1) \\
    -4 & = -2A \\
    A & = 2
\end{align}

* Therefore,

\begin{align}
    f(x) & = \frac{3x - 1}{x^2 - 1} \\
    & = \frac{2}{x + 1} + \frac{1}{x - 1}
\end{align}

* Now we need to rewrite each fraction as a power series and add the two:

* For $\frac{2}{x + 1}$:

\begin{align}
    \frac{2}{x + 1} & = \frac{2}{1 + x} \\
    & = \frac{2}{1 - (-x)}
\end{align}

* This can be expresses as a geometric series with $a_0 = 2$ and $r = -x$

\begin{align}
    \sum_{k = 0}^\infty 2 (-x)^k & = \sum_{k = 0}^\infty 2 (-1)^k x^k
\end{align}

* The above series has interval of convergence $|r| = |-x| = |x| < 1 \Rightarrow -1 < x < 1$

* For $\frac{1}{x - 1}$:

\begin{align}
    \frac{1}{x - 1} & = \frac{1}{-1 + x} \\
    & \frac{1}{-(1 - x)} \\
    & = \frac{-1}{1 - x}
\end{align}

* This can be expressed as a geometric series with $a_0 = -1$ and $r = x$

\begin{align}
    \sum_{k = 0}^\infty -x^k & = - \sum_{k = 0}^\infty x^k
\end{align}

* The above series has interval of convergence $|r| = |x| < 1 \Rightarrow -1 < x < 1$

* Therefore,

\begin{align}
    f(x) & = \sum_{k = 0}^\infty 2 (-1)^k x^k - \sum_{k = 0}^\infty x^k \\
    & = \sum_{k = 0}^\infty \left( 2 (-1)^k x^k - x^k \right) \\
    & = \sum_{k = 0}^\infty \left( 2 (-1)^k - 1 \right) x^k \\
\end{align}

* The interval of convergence for this series is $(-1, 1)$

##### Example 4

Find a power series for $f(x) = \ln x$ centered at $c = 1$.

##### Solution:

* From example 2 we know that

\begin{align}
    g(x) = \frac{1}{x} = \sum_{k = 0}^\infty (-1)^k (x - 1)^k
\end{align}

* We also know that

\begin{align}
    \int \frac{1}{x} \, dx = \ln x + C_1
\end{align}

for some constant of integration $C_1$

* Therefore,

\begin{align}
    \ln x & = C_2 + \int \frac{1}{x} \, dx \\
    & = C_2 + \int \sum_{k = 0}^\infty (-1)^k (x - 1)^k \, dx \\
    & = C_2 + \sum_{k = 0}^\infty (-1)^k \int (x - 1)^k \, dx \\
    & = C + \sum_{k = 0}^\infty (-1)^k \frac{(x - 1)^{k + 1}}{k + 1} \\
\end{align}

* If $x = 1$, then

\begin{align}
    \ln (1) & = 0 = C + 0 \Rightarrow C = 0
\end{align}

* Therefore,

\begin{align}
    \ln x & = \sum_{k = 0}^\infty (-1)^k \frac{(x - 1)^{k + 1}}{k + 1}
\end{align}

* To determine the interval of convergence we need apply the Ratio Test by letting $u_k = (-1)^k \frac{(x - 1)^{k + 1}}{k + 1} \Rightarrow u_{k + 1} = (-1)^{k + 1} \frac{(x - 1)^{k + 2}}{k + 2}$

* The Ratio Test gives us

\begin{align}
    \lim_{k \rightarrow \infty} \left| \frac{u_{k + 1}}{u_k} \right| & = \lim_{k \rightarrow \infty} \left| \frac{(-1)^{k + 1} \frac{(x - 1)^{k + 2}}{k + 2}}{(-1)^k \frac{(x - 1)^{k + 1}}{k + 1}} \right| \\
    & = \lim_{k \rightarrow \infty} \left| \frac{\frac{(x - 1)^{k + 2}}{k + 2}}{\frac{(x - 1)^{k + 1}}{k + 1}} \right| \\
    & = \lim_{k \rightarrow \infty} \left| \frac{(x - 1)^{k + 2}}{k + 2} \cdot \frac{k + 1}{(x - 1)^{k + 1}} \right| \\
    & = \lim_{k \rightarrow \infty} \left| \frac{k + 1}{k + 2} \cdot \frac{(x - 1)(x - 1)^{k + 1}}{(x - 1)^{k + 1}} \right| \\
    & = \lim_{k \rightarrow \infty} \left| \frac{k + 1}{k + 2} \cdot (x - 1) \right| \\
    & = \lim_{k \rightarrow \infty} \frac{k + 1}{k + 2} \cdot \left| x - 1 \right| \\
    & = \left| x - 1 \right| \lim_{k \rightarrow \infty} \frac{k + 1}{k + 2} \\
    & = \left| x - 1 \right| \cdot 1 \\
    & = \left| x - 1 \right|
\end{align}

* For the Ratio Test to determine convergence we require $|x - 1| < 1$:

\begin{align}
    & |x - 1| < 1 \\
    & \Rightarrow -1 < x - 1 < 1 \\
    & \Rightarrow 0 < x < 2
\end{align}

* If $x = 0$, then

\begin{align}
    \sum_{k = 0}^\infty (-1)^k \frac{(0 - 1)^{k + 1}}{k + 1} & = \sum_{k = 0}^\infty (-1)^k \frac{(- 1)^{k + 1}}{k + 1} \\
    & = \sum_{k = 0}^\infty \frac{(- 1)^{2k + 1}}{k + 1} \\
    & = \sum_{k = 0}^\infty \frac{- 1}{k + 1} \qquad (\text{$2k + 1$ is always odd}) \\
    & = -\sum_{k = 0}^\infty \frac{1}{k + 1}
\end{align}

* The above series is a general harmonic series, which diverges (or you could apply whichever appropriate test)

* If $x = 2$, then

\begin{align}
    \sum_{k = 0}^\infty (-1)^k \frac{(2 - 1)^{k + 1}}{k + 1} & = \sum_{k = 0}^\infty (-1)^k \frac{ 1^{k + 1}}{k + 1} \\
    & = \sum_{k = 0}^\infty \frac{(- 1)^k}{k + 1}
\end{align}

* The above series converges by the Alternating Series Test:

    * $\displaystyle \lim_{k \rightarrow \infty} \frac{1}{k + 1} = 0$
    
    * $a_{k + 1} = \frac{1}{k + 2} \leq \frac{1}{k + 1} = a_k$

* Therefore, the inteval of convergence for the series is $(0, 2]$

* To restate our findings: $\displaystyle f(x) = \ln x = \sum_{k = 0}^\infty (-1)^k \frac{(x - 1)^{k + 1}}{k + 1}$ with interval of convergence $(0, 2]$
